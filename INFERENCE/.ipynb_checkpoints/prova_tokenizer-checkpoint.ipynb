```python
from transformers import AutoTokenizer

# 1. Caricamento del tokenizzatore
model_name = "InstaDeepAI/nucleotide-transformer-2.5B-multi-species"
print(f"[*] Caricamento tokenizzatore: {model_name}...")
tokenizer = AutoTokenizer.from_pretrained(model_name)

# 2. Creazione di due sequenze di prova (lunghezza 60 bp)
# Sequenza A: DNA puro (perfettamente divisibile per 6)
seq_pulita = "ATGCGT" * 33 + 'AGA'  

# Sequenza B: Stessa lunghezza, ma con 18 'N' al centro (simulazione masking)
seq_mascherata = seq_pulita[:18] + ('N' * 18) + seq_pulita[36:162] + ('N' * 18) + seq_pulita[180:]

# 3. Tokenizzazione COMPLETA (inclusi i token speciali del modello)
# tokenizer.encode() converte il testo in numeri aggiungendo [CLS] e [SEP]
ids_pulita = tokenizer.encode(seq_pulita)
ids_mascherata = tokenizer.encode(seq_mascherata)

# Riconvertiamo i numeri in stringhe leggibili per vederli con i nostri occhi
tokens_pulita = tokenizer.convert_ids_to_tokens(ids_pulita)
tokens_mascherata = tokenizer.convert_ids_to_tokens(ids_mascherata)
# 4. Stampa dei risultati per confronto
print("\n" + "="*50)
print(" 🧪 TEST: SEQUENZA PULITA")
print("="*50)
print(f"Lunghezza stringa fisica: {len(seq_pulita)} bp")
print(f"Numero di token generati: {len(tokens_pulita)}")
print(f"Lista dei token:\n{tokens_pulita}")
print("-> Come vedi, il modello legge blocchi perfetti da 6-mer.")

print("\n" + "="*50)
print(" 💣 TEST: SEQUENZA MASCHERATA (CON LE 'N')")
print("="*50)
print(f"Lunghezza stringa fisica: {len(seq_mascherata)} bp")
print(f"Numero di token generati: {len(tokens_mascherata)}")
print(f"Lista dei token:\n{tokens_mascherata}")
print("-> ECCO IL PROBLEMA: Guarda come ogni singola 'N' è diventata un token separato!")
print(f"Differenza di token per la stessa lunghezza di DNA: {len(tokens_mascherata) - len(tokens_pulita)} token in più.")
```
