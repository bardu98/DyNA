```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (roc_curve, roc_auc_score, precision_recall_curve, 
                             average_precision_score, confusion_matrix, classification_report,
                             f1_score, accuracy_score, matthews_corrcoef, recall_score, precision_score)

def plot_threshold_analysis(y_true, y_probs, ax):
    """
    Analizza come cambiano le metriche al variare della threshold
    e plotta i risultati sull'asse (ax) fornito.
    """
    thresholds = np.linspace(0, 1, 200)
    
    # Liste per salvare le metriche
    f1_scores = []
    accuracies = []
    precisions = []
    recalls = []
    mccs = []

    # Calcolo metriche per ogni threshold
    for t in thresholds:
        y_pred_t = (y_probs >= t).astype(int)
        
        # Gestione warning: zero division zero
        f1_scores.append(f1_score(y_true, y_pred_t, zero_division=0))
        accuracies.append(accuracy_score(y_true, y_pred_t))
        mccs.append(matthews_corrcoef(y_true, y_pred_t))
        recalls.append(recall_score(y_true, y_pred_t, zero_division=0))
        precisions.append(precision_score(y_true, y_pred_t, zero_division=0))

    # Trova la threshold migliore basata sul massimo F1-Score
    best_idx = np.argmax(f1_scores)
    best_threshold = thresholds[best_idx]
    best_f1 = f1_scores[best_idx]

    # Plotting
    ax.plot(thresholds, precisions, linestyle='--', color='blue', label='Precision', alpha=0.6)
    ax.plot(thresholds, recalls, linestyle='--', color='green', label='Recall', alpha=0.6)
    ax.plot(thresholds, accuracies, linestyle=':', color='gray', label='Accuracy', alpha=0.6)
    ax.plot(thresholds, f1_scores, linewidth=3, color='red', label='F1-Score')
    # ax.plot(thresholds, mccs, linestyle='-.', color='purple', label='MCC', alpha=0.6) # Decommenta se vuoi anche MCC

    # Linea verticale per la Best Threshold
    ax.axvline(best_threshold, color='black', linestyle='--', alpha=0.5)
    ax.text(best_threshold + 0.02, 0.05, f'Best Thresh: {best_threshold:.2f}\nMax F1: {best_f1:.2f}', 
            rotation=0, verticalalignment='bottom')

    ax.set_title('Performance Metrics vs Decision Threshold')
    ax.set_xlabel('Threshold')
    ax.set_ylabel('Score')
    ax.legend(loc='lower left')
    ax.grid(True, alpha=0.3)
    
    print(f"\n✅ Analisi Soglie Completata.")
    print(f"   -> Miglior F1-Score ({best_f1:.3f}) ottenuto con soglia: {best_threshold:.2f}")

def analyze_results_from_csv(csv_path="../data/predictions.csv"):
    # 1. Caricamento Dati
    try:
        df = pd.read_csv(csv_path)
        print(f"✅ Dati caricati da: {csv_path} ({len(df)} righe)")
    except FileNotFoundError:
        print(f"❌ Errore: Il file {csv_path} non esiste.")
        return

    y_true = df['y_true']
    y_probs = df['y_pred_probs']
    
    # Predizione standard a 0.5 per il report testuale
    y_pred_default = (y_probs >= 0.5).astype(int)

    # Impostazioni grafiche generali
    sns.set_style("whitegrid")
    # Creiamo una figura con layout complesso: 2 righe
    # Riga 1: 3 grafici (Confusion, ROC, PR)
    # Riga 2: 2 grafici (Distribuzione Prob, Analisi Threshold)
    fig = plt.figure(figsize=(18, 12))
    gs = fig.add_gridspec(2, 6)

    # ==========================================
    # A. CLASSIFICATION REPORT (Console)
    # ==========================================
    print("\n--- Classification Report (Threshold 0.5) ---")
    print(classification_report(y_true, y_pred_default))

    # ==========================================
    # B. MATRICE DI CONFUSIONE (Row 1, Col 1-2)
    # ==========================================
    ax1 = fig.add_subplot(gs[0, 0:2])
    cm = confusion_matrix(y_true, y_pred_default)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, ax=ax1,
                xticklabels=['Pred 0', 'Pred 1'], yticklabels=['True 0', 'True 1'])
    ax1.set_title("Confusion Matrix (Thresh=0.5)")
    
    # ==========================================
    # C. ROC CURVE (Row 1, Col 2-4)
    # ==========================================
    ax2 = fig.add_subplot(gs[0, 2:4])
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    auc_score = roc_auc_score(y_true, y_probs)
    ax2.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {auc_score:.3f}')
    ax2.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    ax2.set_title('ROC Curve')
    ax2.legend(loc="lower right")

# ==========================================
    # D. PR CURVE SMOOTH (Row 1, Col 4-6)
    # ==========================================
    ax3 = fig.add_subplot(gs[0, 4:6])
    
    # 1. Calcolo curve grezze
    precision, recall, _ = precision_recall_curve(y_true, y_probs)
    pr_auc = average_precision_score(y_true, y_probs)
    
    # 2. INTERPOLAZIONE PER SMOOTHING
    # Creiamo 200 punti equispaziati tra 0 e 1 per la Recall
    recall_smooth = np.linspace(0, 1, 200)
    
    # Interpoliamo la precisione su questi punti
    # Nota: np.interp richiede che l'asse x (recall) sia crescente. 
    # La funzione precision_recall_curve spesso restituisce recall decrescente, quindi usiamo [::-1] per invertire
    precision_smooth = np.interp(recall_smooth, recall[::-1], precision[::-1])
    
    # Opzionale: Assicuriamoci che la precisione sia monotona (step-down) per una visualizzazione più pulita
    # Questa tecnica si chiama "Interpolated Precision" (usata spesso in Information Retrieval)
    precision_smooth = np.maximum.accumulate(precision_smooth[::-1])[::-1]

    baseline = y_true.mean()

    # Plot della versione smooth
    ax3.plot(recall_smooth, precision_smooth, color='green', lw=2, label=f'PR AUC = {pr_auc:.3f} (Smooth)')
    
    # (Opzionale) Plot della versione originale in trasparenza per confronto
    # ax3.plot(recall, precision, color='green', alpha=0.2, lw=1) 

    ax3.plot([0, 1], [baseline, baseline], color='gray', linestyle='--', label=f'Base ({baseline:.2f})')
    
    ax3.set_title('Precision-Recall Curve (Smoothed)')
    ax3.set_xlabel('Recall')
    ax3.set_ylabel('Precision')
    ax3.legend(loc="best")

    # ==========================================
    # E. PROBABILITY DISTRIBUTION (Row 2, Col 1-3)
    # ==========================================
    ax4 = fig.add_subplot(gs[1, 0:3])
    sns.histplot(data=df, x='y_pred_probs', hue='y_true', bins=50, kde=True, element="step", 
                 palette={0: "red", 1: "blue"}, alpha=0.3, ax=ax4)
    ax4.set_title('Probability Distribution')
    ax4.set_xlim(0, 1)

    # ==========================================
    # F. THRESHOLD ANALYSIS (Row 2, Col 3-6) -> NUOVO!
    # ==========================================
    ax5 = fig.add_subplot(gs[1, 3:6])
    plot_threshold_analysis(y_true, y_probs, ax5)

    plt.tight_layout()
    plt.show()


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, average_precision_score

# ==========================================
# 1. CONFIGURAZIONE STILE "NATURE" MIGLIORATO
# ==========================================
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial', 'Helvetica', 'DejaVu Sans'],

    'font.size': 9,
    'axes.labelsize': 10,
    'axes.titlesize': 11,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
    'legend.fontsize': 8,

    'figure.dpi': 300,
    'axes.linewidth': 0.7,
    'lines.linewidth': 1.8,

    # Spines e tick stile Nature
    'axes.spines.top': False,
    'axes.spines.right': False,
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    'xtick.major.size': 3,
    'ytick.major.size': 3,

    'legend.frameon': False,
})

# Colori NPG puliti e coerenti
COLOR_ACT = "#F49F0A"   # Arancio/rosso elegante
COLOR_REST = "#00A6A6"  # Blu intenso
COLOR_BASE = "#666666"  # Grigio neutro

def get_smooth_pr_curve_custom(y_true, y_probs, points=200):
    """Interpolazione Precision-Recall più liscia, stile editoriale."""
    precision, recall, _ = precision_recall_curve(y_true, y_probs)

    recall_smooth = np.linspace(0, 1, points)
    precision_smooth = np.interp(recall_smooth, recall[::-1], precision[::-1])
    precision_smooth = np.maximum.accumulate(precision_smooth[::-1])[::-1]

    return recall_smooth, precision_smooth

def plot_nature_comparison(file_act, file_rest, save_path="../data/nature_figure_comparison.pdf"):
    
    # ============== CARICAMENTO ==============
    try:
        df_act = file_act
        df_rest = file_rest
    except FileNotFoundError as e:
        print(f"❌ Errore: {e}")
        return

    y_true_act, y_probs_act = df_act['y_true'], df_act['y_pred_probs']
    y_true_rest, y_probs_rest = df_rest['y_true'], df_rest['y_pred_probs']

    auc_act = roc_auc_score(y_true_act, y_probs_act)
    auc_rest = roc_auc_score(y_true_rest, y_probs_rest)
    ap_act = average_precision_score(y_true_act, y_probs_act)
    ap_rest = average_precision_score(y_true_rest, y_probs_rest)

    base_act = y_true_act.mean()
    base_rest = y_true_rest.mean()

    # ============== FIGURA stile Nature ==============
    fig, axes = plt.subplots(
        1, 2,
        figsize=(8.6, 3.0),   # proporzioni tipiche Nature
        constrained_layout=False
    )

    plt.subplots_adjust(wspace=0.32, left=0.08, right=0.98, top=0.92, bottom=0.18)

    # ------------------------------------------
    # PANNELLO A: ROC
    # ------------------------------------------
    ax_roc = axes[0]

    fpr_act, tpr_act, _ = roc_curve(y_true_act, y_probs_act)
    fpr_rest, tpr_rest, _ = roc_curve(y_true_rest, y_probs_rest)
    
    ax_roc.plot(fpr_act, tpr_act, color=COLOR_ACT, label=f'Activated cells (AUC={auc_act:.2f})')
    ax_roc.plot(fpr_rest, tpr_rest, color=COLOR_REST, label=f'Resting cells (AUC={auc_rest:.2f})')

    ax_roc.plot([0, 1], [0, 1], linestyle='--', color=COLOR_BASE, linewidth=1, alpha=0.5)

    ax_roc.set_xlabel("False Positive Rate")
    ax_roc.set_ylabel("True Positive Rate")
    ax_roc.set_xlim(0, 1)
    ax_roc.set_ylim(0, 1)

    ax_roc.legend(loc="lower right")

    ax_roc.text(-0.18, 1.05, "a", transform=ax_roc.transAxes,
                fontsize=13, fontweight="bold", ha="left", va="top")
    ax_roc.set_title("ROC Curves")


    # ------------------------------------------
    # PANNELLO B: PR
    # ------------------------------------------
    ax_pr = axes[1]

    rec_act, prec_act = get_smooth_pr_curve_custom(y_true_act, y_probs_act)
    rec_rest, prec_rest = get_smooth_pr_curve_custom(y_true_rest, y_probs_rest)

    ax_pr.plot(rec_act, prec_act, color=COLOR_ACT, label=f'Activated cells (AP={ap_act:.2f})')
    ax_pr.plot(rec_rest, prec_rest, color=COLOR_REST, label=f'Resting cells (AP={ap_rest:.2f})')

    ax_pr.axhline(base_act, linestyle=':', color=COLOR_ACT, alpha=0.4)
    ax_pr.axhline(base_rest, linestyle=':', color=COLOR_REST, alpha=0.4)

    ax_pr.set_xlabel("Recall")
    ax_pr.set_ylabel("Precision")
    ax_pr.set_xlim(0, 1)
    ax_pr.set_ylim(0.5, 1)

    ax_pr.legend(loc="upper right")

    ax_pr.text(-0.18, 1.05, "b", transform=ax_pr.transAxes,
               fontsize=13, fontweight="bold", ha="left", va="top")
    ax_pr.set_title("Precision–Recall Curves")

    # ============== SALVATAGGIO ==============
    plt.savefig(save_path, format="pdf")
    plt.savefig(save_path.replace(".pdf", ".png"), format="png", dpi=300)
    plt.show()

    print(f"Figura generata:\n → {save_path}")




######
#DATI
###
import pickle
from Bio.Seq import Seq

with open('../data/data_pkl/CD4T_h19_Act_tot_99_8_percentile.pkl', 'rb') as f:   
    dataset_sapiens_cd4t_act = pickle.load(f)
if True:
    for d1, d2 in zip(dataset_sapiens_cd4t_act, dataset_sapiens_cd4t_act):
        d1['sequence_rev'] = str(Seq(d1['sequence']).reverse_complement())


with open('../data/data_pkl/CD4T_h19_Rest_tot_99_8_percentile.pkl', 'rb') as f:   
    dataset_sapiens_cd4t_rest = pickle.load(f)
if True:
    for d1, d2 in zip(dataset_sapiens_cd4t_rest, dataset_sapiens_cd4t_rest):
        d1['sequence_rev'] = str(Seq(d1['sequence']).reverse_complement())


df_act = pd.DataFrame([np.array([i['label'] for i in dataset_sapiens_cd4t_act]),pd.read_pickle("preds_act_model_sum_folds_16_02_26.pkl")]).T

df_rest = pd.DataFrame([np.array([i['label'] for i in dataset_sapiens_cd4t_rest]),pd.read_pickle("preds_rest_model_sum_folds_16_02_26.pkl")]).T

df_act = df_act.rename(columns={0:'y_true',1:'y_pred_probs'})
df_rest = df_rest.rename(columns={0:'y_true',1:'y_pred_probs'})

###########
#FINE ESTRAZIONE DATI
########



# Esegui
if __name__ == "__main__":
    plot_nature_comparison(
        file_act=df_act, 
        file_rest=df_rest  #FAI RIGIRARE QUESTO SOTTO
    )
```
