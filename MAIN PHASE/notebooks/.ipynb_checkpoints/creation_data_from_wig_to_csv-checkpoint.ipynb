```python
import pandas as pd
import numpy as np
import gzip
import os
from io import StringIO
import glob # Utile per trovare i file

#################################################################################################
# LA TUA FUNZIONE PER IL PARSING (copiata per comodità)
#################################################################################################

def parse_wig_file(wig_file_path, max_regions=100000):
    """
    Parsifica file WIG con dati di occupancy dei nucleosomi
    """
    print(f"  Parsing file WIG: {os.path.basename(wig_file_path)}...")
    
    regions = []
    current_chrom = None
    current_start = None
    current_step = None
    current_span = 1
    position = 0
    lines_processed = 0
    
    try:
        # Gestisci file compressi o no
        opener = gzip.open if wig_file_path.endswith('.gz') else open
        
        with opener(wig_file_path, 'rt') as f:
            for line in f:
                line = line.strip()
                
                if not line or line.startswith('track') or line.startswith('browser'):
                    continue
                
                lines_processed += 1
                if lines_processed % 50000000 == 0: # Logga ogni milione di righe
                    print(f"    ... Processate {lines_processed:,} linee, trovate {len(regions):,} regioni...")
                
                # Stop se abbiamo raggiunto il limite
                if len(regions) >= max_regions:
                    print(f"    ⚠ Raggiunto limite di {max_regions:,} regioni, stop parsing")
                    break
                
                # fixedStep format
                if line.startswith('fixedStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('start='):
                            current_start = int(part.split('=')[1])
                            position = current_start
                        elif part.startswith('step='):
                            current_step = int(part.split('=')[1])
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                
                # variableStep format
                elif line.startswith('variableStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                    current_step = None
                
                # Data line
                else:
                    try:
                        if current_step:
                            # fixedStep: solo valore
                            value = float(line)
                            if value > 0: # Solo regioni con segnale
                                regions.append({
                                    'chr': current_chrom,
                                    'start': position,
                                    'end': position + current_span,
                                    'score': value
                                })
                            position += current_step
                        else:
                            # variableStep: posizione e valore
                            pos, value = line.split()
                            pos = int(pos)
                            value = float(value)
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': pos,
                                    'end': pos + current_span,
                                    'score': value
                                })
                    except Exception:
                        # Salta righe dati malformate
                        continue
            
        print(f"    ✓ Trovate {len(regions):,} regioni con segnale")
        return pd.DataFrame(regions)
    
    except Exception as e:
        print(f"    ✗ Errore parsing WIG: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

#################################################################################################
# SCRIPT DI ANALISI
#################################################################################################

def analizza_file_wig():
    print("="*60)
    print("ANALISI STATISTICA FILE WIG")
    print("="*60)

    # Aggiorna questo percorso se necessario
    base_dir = os.getcwd()
    data_dir = os.path.abspath(os.path.join(base_dir, "..", "data"))
    wig_extract_dir = os.path.join(data_dir, "wig_files")

    # Trova tutti i file WIG 147bp (come nel tuo script)
    wig_files = glob.glob(os.path.join(wig_extract_dir, "*_147.wig.gz"))

    if not wig_files:
        print(f"Nessun file WIG trovato in {wig_extract_dir}")
        return

    # Imposta un limite MOLTO ALTO per leggere i file interi
    # 5 Milioni dovrebbe essere sufficiente, aumenta se vedi ancora
    # il messaggio "Raggiunto limite..."
    LIMITE_REGIONI_PER_FILE = 5000000 

    for wig_file in sorted(wig_files):
        print(f"\n--- Analizzando: {os.path.basename(wig_file)} ---")
        
        df_wig = parse_wig_file(wig_file, max_regions=LIMITE_REGIONI_PER_FILE)
        
        if df_wig is None or df_wig.empty:
            print("Nessun dato da analizzare.")
            continue

        # Calcola e stampa le statistiche della colonna 'score'
        print(f"\nStatistiche 'score' per {os.path.basename(wig_file)}:")
        
        # Usiamo .describe() di pandas, che è perfetto per questo
        stats = df_wig['score'].describe(percentiles=[.25, .5, .75, .9, .95, .99])
        print(stats)
        print("-"
*60)

if __name__ == "__main__":
    analizza_file_wig()
```

```python
import pandas as pd
import numpy as np
import requests
import gzip
import os
from Bio import SeqIO
from io import StringIO
import tarfile
import random
import matplotlib.pyplot as plt
import seaborn as sns
#from intervaltree import IntervalTree # Dipendenza opzionale ma consigliata per l'efficienza

#################################################################################################
# FUNZIONI HELPER
#################################################################################################

def calc_gc(seq):
    """Calcola il contenuto GC di una sequenza."""
    if not isinstance(seq, str) or len(seq) == 0:
        return 0
    gc = seq.count('G') + seq.count('C')
    return (gc / len(seq)) * 100

#################################################################################################
# STEP 1: GESTIONE GENOMA HG18
#################################################################################################

def download_hg18_genome(output_dir="hg18_genome"):
    """
    Scarica il genoma umano hg18 da UCSC
    """
    print("Scaricando genoma hg18 da UCSC...")
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # URL base UCSC per hg18
    base_url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg18/chromosomes/"
    
    # Scarica cromosomi principali (1-22, X, Y)
    chromosomes = [f"chr{i}" for i in range(1, 23)] + ["chrX", "chrY"]
    
    genome_dict = {}
    
    for chrom in chromosomes:
        file_name = f"{chrom}.fa.gz"
        file_path = os.path.join(output_dir, file_name)
        
        # Percorso del file decompresso
        unzipped_file_path = file_path.replace('.gz', '')

        if os.path.exists(unzipped_file_path):
            print(f"✓ {chrom} già scaricato")
            # Carica sequenza già presente
            try:
                with open(unzipped_file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        genome_dict[chrom] = str(record.seq).upper() # Salva in maiuscolo
                continue
            except Exception as e:
                print(f"✗ Errore nel caricamento di {unzipped_file_path}: {e}. Tento di riscaricarlo.")

        
        url = base_url + file_name
        print(f"Scaricando {chrom}...")
        
        try:
            response = requests.get(url, stream=True, timeout=300)
            if response.status_code == 200:
                # Salva file compresso
                with open(file_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                
                # Decomprimi e leggi
                sequence_content = ""
                with gzip.open(file_path, 'rt') as f:
                    # Leggi tutto il contenuto
                    sequence_content = f.read()
                
                # Parse dalla stringa
                with StringIO(sequence_content) as f_in:
                    records = SeqIO.parse(f_in, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Salva in maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} scaricato: {len(seq_str):,} bp")
                
                # Salva versione decompressa
                with open(unzipped_file_path, 'w') as f_out:
                    f_out.write(sequence_content)
                
                # Rimuovi file compresso per salvare spazio
                os.remove(file_path)
                
        except Exception as e:
            print(f"✗ Errore scaricamento {chrom}: {e}")
            if os.path.exists(file_path):
                os.remove(file_path) # Pulisci file corrotto
            continue
    
    print(f"\n✓ Genoma hg18 scaricato: {len(genome_dict)} cromosomi")
    return genome_dict

def load_hg18_genome(genome_dir="hg18_genome"):
    """
    Carica genoma hg18 già scaricato
    """
    print("Caricamento genoma hg18...")
    
    genome_dict = {}
    
    chromosomes = [f"chr{i}" for i in range(1, 23)] + ["chrX", "chrY"]
    
    for chrom in chromosomes:
        file_path = os.path.join(genome_dir, f"{chrom}.fa")
        
        if os.path.exists(file_path):
            try:
                with open(file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Assicurati sia maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} caricato: {len(seq_str):,} bp")
            except Exception as e:
                print(f"✗ Errore caricamento {file_path}: {e}")
        else:
            print(f"✗ {chrom} non trovato in {genome_dir}")
    
    if not genome_dict:
        print(f"✗ Nessun file genoma trovato in {genome_dir}. Prova a scaricarlo prima.")
    else:
        print(f"\n✓ Genoma caricato: {len(genome_dict)} cromosomi")
    return genome_dict

#################################################################################################
# STEP 2: GESTIONE DATI WIG (GSE36979)
#################################################################################################

def parse_wig_file(wig_file_path, max_regions=100000):
    """
    Parsifica file WIG con dati di occupancy dei nucleosomi
    """
    print(f"  Parsing file WIG: {wig_file_path}")
    
    regions = []
    current_chrom = None
    current_start = None
    current_step = None
    current_span = 1
    position = 0
    lines_processed = 0
    
    try:
        # Gestisci file compressi o no
        opener = gzip.open if wig_file_path.endswith('.gz') else open
        
        with opener(wig_file_path, 'rt') as f:
            for line in f:
                line = line.strip()
                
                if not line or line.startswith('track') or line.startswith('browser'):
                    continue
                
                lines_processed += 1
                
                # --- MODIFICA 1: Riduzione print ---
                if lines_processed % 50000000 == 0:
                    print(f"    ... Processate {lines_processed:,} linee, trovate {len(regions):,} regioni...")
                
                # Stop se abbiamo raggiunto il limite (per debug)
                if len(regions) >= max_regions:
                    print(f"    ⚠ Raggiunto limite di {max_regions:,} regioni, stop parsing")
                    break
                
                # fixedStep format
                if line.startswith('fixedStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('start='):
                            current_start = int(part.split('=')[1])
                            position = current_start
                        elif part.startswith('step='):
                            current_step = int(part.split('=')[1])
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                
                # variableStep format
                elif line.startswith('variableStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                    current_step = None
                
                # Data line
                else:
                    try:
                        if current_step:
                            # fixedStep: solo valore
                            value = float(line)
                            if value > 0: # Solo regioni con segnale
                                regions.append({
                                    'chr': current_chrom,
                                    'start': position,
                                    'end': position + current_span,
                                    'score': value
                                })
                            position += current_step
                        else:
                            # variableStep: posizione e valore
                            pos, value = line.split()
                            pos = int(pos)
                            value = float(value)
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': pos,
                                    'end': pos + current_span,
                                    'score': value
                                })
                    except Exception:
                        # Salta righe dati malformate
                        continue
            
        print(f"    ✓ Trovate {len(regions):,} regioni con segnale")
        return pd.DataFrame(regions)
    
    except Exception as e:
        print(f"    ✗ Errore parsing WIG: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

def extract_tar_wig_files(tar_path, extract_dir="wig_files", filter_147bp=True):
    """
    Estrae file WIG da archivio TAR
    Se filter_147bp=True, estrae solo file con frammenti da 147bp (nucleosomi canonici)
    """
    print(f"Estraendo file da {tar_path}...")
    
    if not os.path.exists(extract_dir):
        os.makedirs(extract_dir)
    
    extracted_files = []
    skipped_files = []
    
    try:
        with tarfile.open(tar_path, 'r') as tar:
            members = tar.getmembers()
            print(f"Trovati {len(members)} file nell'archivio")
            
            if filter_147bp:
                print("Filtro attivo: solo file con frammenti 147bp (nucleosomi canonici)")
            
            for member in members:
                if member.name.endswith('.wig') or member.name.endswith('.wig.gz'):

                    # --- AGGIUNGI QUESTA RIGA PER SALTARE IL FILE ANOMALO ---
                    if "GSM907789" in member.name:
                        print(f"File anomalo saltato: {member.name}")
                        continue
                    # -----------------------------------------------------
                    
                    # Filtra solo file 147bp se richiesto
                    if filter_147bp and '_147.wig' not in member.name:
                        skipped_files.append(member.name)
                        continue
                    
                    target_path = os.path.join(extract_dir, member.name)
                    if not os.path.exists(target_path):
                        print(f"Estraendo: {member.name}")
                        tar.extract(member, extract_dir)
                    extracted_files.append(target_path)
            
        print(f"\n✓ Estratti {len(extracted_files)} file WIG (nucleosomi 147bp)")
        if skipped_files:
            print(f"✓ Saltati {len(skipped_files)} file con altre dimensioni")
        
        return extracted_files
    
    except Exception as e:
        print(f"✗ Errore estrazione TAR: {e}")
        return []

#################################################################################################
# STEP 3: GESTIONE ANNOTAZIONI (PROMOTORI / ENHANCER)
#################################################################################################

def load_genomic_features(gtf_path,  
                          feature_types,  
                          relative_to_tss=False,  
                          window_upstream=0,  
                          window_downstream=0):
    """
    Estrae coordinate per specifici feature genomici (es. esoni, promotori)
    da un file di annotazione GTF.
    """
    print(f"Caricamento annotazioni da {gtf_path} per feature: {feature_types}...")
    features_list = []

    try:
        with open(gtf_path, 'r') as f:
            for line in f:
                if line.startswith('#'):
                    continue
                parts = line.strip().split('\t')
                if len(parts) < 9:
                    continue
                
                chrom, source, feature, start_str, end_str, score, strand, frame, attrs = parts

                if feature in feature_types:
                    try:
                        start = int(start_str)
                        end = int(end_str)
                        
                        reg_start, reg_end = 0, 0

                        if relative_to_tss:
                            # Calcola regione relativa al TSS (per promotori)
                            tss = start if strand == '+' else end
                            reg_start = max(0, tss - window_upstream)
                            reg_end = tss + window_downstream
                        else:
                            # Usa coordinate dirette del feature + finestre
                            if strand == '+':
                                reg_start = max(0, start - window_upstream)
                                reg_end = end + window_downstream
                            else:
                                reg_start = max(0, start - window_downstream)
                                reg_end = end + window_upstream

                        features_list.append((chrom, reg_start, reg_end))
                        
                    except ValueError:
                        continue

    except FileNotFoundError:
        print(f"Errore: File GTF non trovato a {gtf_path}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
    except Exception as e:
        print(f"Errore during GTF read: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

    if not features_list:
        print(f"⚠ Nessun feature di tipo {feature_types} trovato nel file.")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

    df_features = pd.DataFrame(features_list, columns=['chr', 'start', 'end'])
    df_features = df_features.drop_duplicates().reset_index(drop=True)
    
    print(f"✓ Trovati {len(df_features):,} feature unici per '{', '.join(feature_types)}'")
    return df_features

def load_bed_features(bed_path):
    """
    Carica un file BED semplice (chr, start, end).
    """
    print(f"Caricamento features da file BED: {os.path.basename(bed_path)}")
    
    if not os.path.exists(bed_path):
        print(f"✗ File BED non trovato: {bed_path}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
        
    try:
        df = pd.read_csv(
            bed_path,  
            sep='\t',  
            header=None,
            comment='#', # Ignora righe di commento
            usecols=[0, 1, 2],  
            names=['chr', 'start', 'end']
        )
        
        df = df.dropna().drop_duplicates().reset_index(drop=True)
        df['start'] = df['start'].astype(int)
        df['end'] = df['end'].astype(int)
        
        print(f"✓ Trovati {len(df):,} features unici")
        return df
        
    except Exception as e:
        print(f"✗ Errore caricamento file BED: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

def get_feature_dataframe(gtf_path, region_key, promoter_window=1000):
    """
    Funzione "manager" che carica le coordinate per una regione genomica
    specificata da una chiave (key).
    """
    print(f"\n--- Richiesta regione: '{region_key}' ---")
    
    feature_map = {
        "promoter": (["transcript", "gene"], True, promoter_window, promoter_window),
        "exon": (["exon"], False, 0, 0),
        "gene_body": (["gene"], False, 0, 0),
        "utr5": (["five_prime_utr", "5UTR"], False, 0, 0),
        "utr3": (["three_prime_utr", "3UTR"], False, 0, 0),
        "enhancer": (["enhancer"], False, 0, 0), # Per file GTF
    }
    
    if region_key not in feature_map:
        print(f"Errore: Chiave regione '{region_key}' non riconosciuta.")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
        
    params = feature_map[region_key]
    
    return load_genomic_features(gtf_path,  
                                 feature_types=params[0],  
                                 relative_to_tss=params[1],  
                                 window_upstream=params[2],  
                                 window_downstream=params[3])

def download_refgene_hg18(output_dir="../data", promoter_window=1000):
    """
    Scarica le annotazioni RefSeq (hg18) da UCSC e genera regioni promotrici ±promoter_window bp dal TSS.
    """
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg18/database/refGene.txt.gz"
    local_gz = os.path.join(output_dir, "refGene.txt.gz")
    local_csv = os.path.join(output_dir, f"promoters_hg18_{promoter_window}bp.csv")

    if os.path.exists(local_csv):
        print(f"✓ File promotori già presente, caricamento: {local_csv}")
        try:
            return pd.read_csv(local_csv)
        except Exception as e:
            print(f"✗ Errore nel caricamento di {local_csv}: {e}. Tento di rigenerarlo.")

    if not os.path.exists(local_gz):
        print(f"Scaricando RefSeq hg18 da UCSC...")
        try:
            response = requests.get(url, stream=True, timeout=180)
            response.raise_for_status()
            with open(local_gz, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"✓ File salvato: {local_gz}")
        except Exception as e:
            print(f"✗ Errore nel download del file RefSeq: {e}")
            return None
    else:
        print(f"✓ File RefSeq già presente: {local_gz}")

    cols = [
        'bin', 'name', 'chrom', 'strand', 'txStart', 'txEnd',
        'cdsStart', 'cdsEnd', 'exonCount', 'exonStarts', 'exonEnds',
        'score', 'name2', 'cdsStartStat', 'cdsEndStat', 'exonFrames'
    ]
    
    print("Caricamento e parsing refGene...")
    try:
        refgene = pd.read_csv(local_gz, sep='\t', header=None, names=cols, compression='gzip')
    except Exception as e:
        print(f"✗ Errore parsing refGene: {e}")
        return None
        
    print(f"✓ {len(refgene):,} geni caricati")

    print(f"Calcolo regioni promotrici ±{promoter_window} bp dal TSS...")
    promoter_records = []

    for _, row in refgene.iterrows():
        chrom = row['chrom']
        if not isinstance(chrom, str) or not chrom.startswith("chr"):
            continue
        
        try:
            txStart = int(row['txStart'])
            txEnd = int(row['txEnd'])
            
            if row['strand'] == '+':
                tss = txStart
            else:
                tss = txEnd
            
            start = max(0, tss - promoter_window)
            end = tss + promoter_window

            promoter_records.append({
                'chr': chrom,
                'start': int(start),
                'end': int(end),
                'strand': row['strand'],
                'gene': row['name2']
            })
        except ValueError:
            continue

    df_promoters = pd.DataFrame(promoter_records)
    df_promoters = df_promoters.drop_duplicates(subset=['chr', 'start', 'end'])
    print(f"✓ {len(df_promoters):,} regioni promotrici uniche generate")

    try:
        df_promoters.to_csv(local_csv, index=False)
        print(f"✓ Salvati promotori in: {local_csv}")
    except Exception as e:
        print(f"✗ Errore salvataggio CSV promotori: {e}")
    
    return df_promoters

#################################################################################################
# STEP 4: CREAZIONE DATASET (Logica Positivo vs Negativo)
#################################################################################################

def filter_regions_by_feature(df_regions, df_features):
    """
    Mantiene solo le regioni (es. nucleosomi) che si sovrappongono  
    con i feature forniti (es. promotori, esoni).
    Restituisce un nuovo DataFrame filtrato.
    """
    print("  Filtraggio regioni per sovrapposizione con feature...")

    if df_features is None or df_features.empty:
        print("  ⚠ Nessuna annotazione feature fornita, ritorno tutte le regioni (nessun filtro).")
        return df_regions

    # Crea dizionario cromosoma → array di (start, end) per efficienza
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("  Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return pd.DataFrame(columns=df_regions.columns)
        
    keep = np.zeros(len(df_regions), dtype=bool)

    for i, row in enumerate(df_regions.itertuples(index=False)):
        chrom = getattr(row, 'chr', None)
        if chrom not in df_features_by_chr:
            continue
            
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        
        if s is None or e is None:
            continue

        # Cerca sovrapposizione
        for f_start, f_end in df_features_by_chr[chrom]:
            # (e < f_start) -> la regione è completamente prima del feature
            # (s > f_end)   -> la regione è completamente dopo il feature
            if not (e < f_start or s > f_end):  
                keep[i] = True
                break # Trovata sovrapposizione, passa alla prossima regione

    filtered_df = df_regions[keep]
    print(f"  ✓ {len(filtered_df):,}/{len(df_regions):,} regioni trovate nei feature forniti.")
    return filtered_df.reset_index(drop=True)

# --- MODIFICA 2: Funzione POSITIVI (score > 1.0) ---
def identify_nucleosome_positions(df_wig, threshold_value=1.0, nucleosome_length=201):
    """
    Identifica posizioni nucleosomi (ALTA occupazione)
    usando una SOGLIA ASSOLUTA (score > threshold)
    """
    print("  Identificazione posizioni nucleosomi (ALTA occupazione)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    # Usa il valore assoluto passato
    print(f"    Soglia segnale (Positivi): > {threshold_value:.2f}")
    
    # MODIFICA CRITICA: Cerca i punteggi STRETTAMENTE MAGGIORI del background
    df_nucleosomes = df_wig[df_wig['score'] > threshold_value].copy()
    
    # Estendi regioni alla lunghezza desiderata
    df_nucleosomes['center'] = (df_nucleosomes['start'] + df_nucleosomes['end']) // 2
    df_nucleosomes['start'] = df_nucleosomes['center'] - nucleosome_length // 2
    df_nucleosomes['end'] = df_nucleosomes['start'] + nucleosome_length
    
    print(f"    ✓ Identificati {len(df_nucleosomes)} nucleosomi (Positivi)")
    
    return df_nucleosomes[['chr', 'start', 'end', 'score']]

# --- MODIFICA 3: Funzione NEGATIVI (score == 1.0) ---
def identify_non_nucleosome_positions(df_wig, threshold_value=1.0, nucleosome_length=201):
    """
    Identifica posizioni NON-nucleosomi (BASSA occupazione)
    usando una SOGLIA ASSOLUTA (score == threshold)
    """
    print("  Identificazione posizioni NON-nucleosomi (BASSA occupazione)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    # Usa il valore assoluto passato
    print(f"    Soglia segnale (Negativi): == {threshold_value:.2f}")
    
    # Filtra regioni con segnale BASSO (esattamente il background)
    df_nucleosomes = df_wig[df_wig['score'] == threshold_value].copy()
    
    # Estendi regioni alla lunghezza desiderata
    df_nucleosomes['center'] = (df_nucleosomes['start'] + df_nucleosomes['end']) // 2
    df_nucleosomes['start'] = df_nucleosomes['center'] - nucleosome_length // 2
    df_nucleosomes['end'] = df_nucleosomes['start'] + nucleosome_length
    
    print(f"    ✓ Identificati {len(df_nucleosomes)} NON-nucleosomi (Negativi)")
    
    return df_nucleosomes[['chr', 'start', 'end', 'score']]

def extract_sequences_from_genome(df_regions, genome_dict, max_sequences=2000):
    """
    Estrae sequenze DNA dalle regioni specificate
    """
    print(f"    Estrazione sequenze da {len(df_regions)} regioni...")
    
    sequences = []
    
    count = 0
    processed = 0
    for idx, row in df_regions.iterrows():
        processed += 1
        if count >= max_sequences:
            print(f"      Raggiunto limite di {max_sequences} sequenze.")
            break
        
        chrom = row['chr']
        start = int(row['start'])
        end = int(row['end'])
        
        if chrom not in genome_dict:
            continue
        
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        sequences.append(seq)
        count += 1
        
        if count % 500 == 0 and count > 0:
            print(f"      Estraendole... {count}/{max_sequences} sequenze valide trovate (processate {processed} regioni)")
    
    print(f"    ✓ Estratte {len(sequences)} sequenze")
    return sequences

# --- MODIFICA 4: Funzione DATASET (con nuova logica di bilanciamento) ---
# def create_nucleosome_dataset(wig_files, genome_dict,  
#                             max_sequences_per_file=1000,
#                             threshold_high=1.0,
#                             threshold_low=1.0,
#                             max_regions_per_wig=100000,
#                             df_features=None,
#                             sequence_length=201):
#     """
#     Crea dataset finale combinando dati WIG e genoma.
#     Positivi: score > threshold_high
#     Negativi: score == threshold_low
#     """
#     print("\n" + "="*60)
#     print("CREAZIONE DATASET NUCLEOSOMI (Positivi vs Negativi in-loco)")
#     print("="*60)
#     print(f"Lunghezza sequenza: {sequence_length} bp")
#     print(f"Soglia ALTA (positivi): > {threshold_high:.2f}")
#     print(f"Soglia BASSA (negativi): == {threshold_low:.2f}")
#     print(f"Filtro regioni: {df_features is not None}")
    
#     all_sequences = []
#     all_labels = []
    
#     # Processa TUTTI i file WIG
#     for idx, wig_file in enumerate(wig_files, 1):
#         print(f"\n[{idx}/{len(wig_files)}] Processando: {os.path.basename(wig_file)}")
        
#         # Passa max_regions_per_wig al parser
#         df_wig = parse_wig_file(wig_file, max_regions=max_regions_per_wig)
        
#         if df_wig is None or len(df_wig) == 0:
#             print("  ⚠ Nessun dato trovato, skip")
#             continue
            
#         # -----------------------------------------------------------------
#         # 1. STREAM POSITIVO (Label 1) - score > 1.0
#         # -----------------------------------------------------------------
#         print("\n  --- Stream Positivo (Label 1) ---")
#         df_nucleosomes_high = identify_nucleosome_positions(
#             df_wig,  
#             threshold_value=threshold_high, # Passa threshold_high (1.0)
#             nucleosome_length=sequence_length
#         )
        
#         df_positive_filtered = filter_regions_by_feature(df_nucleosomes_high, df_features)
#         if df_positive_filtered.empty:
#             print("  ⚠ Nessuna regione positiva (>1.0) nei feature forniti, skip file")
#             continue
            
#         sequences_pos = extract_sequences_from_genome(
#             df_positive_filtered,  
#             genome_dict,  
#             max_sequences=max_sequences_per_file 
#         )
        
#         # -----------------------------------------------------------------
#         # 2. STREAM NEGATIVO (Label 0) - score == 1.0
#         # -----------------------------------------------------------------
#         print("\n  --- Stream Negativo (Label 0) ---")

#         df_nucleosomes_low = identify_non_nucleosome_positions(
#             df_wig,  
#             threshold_value=threshold_low, # Passa threshold_low (1.0)
#             nucleosome_length=sequence_length
#         )
        
#         df_negative_filtered = filter_regions_by_feature(df_nucleosomes_low, df_features)
#         if df_negative_filtered.empty:
#             print("  ⚠ Nessuna regione negativa (==1.0) nei feature forniti, skip negativi")
#             continue
            
#         sequences_neg = extract_sequences_from_genome(
#             df_negative_filtered,  
#             genome_dict,  
#             max_sequences=max_sequences_per_file # Estrai fino al limite
#         )
        
#         # -----------------------------------------------------------------
#         # 3. BILANCIAMENTO e AGGIUNTA AL DATASET (NUOVA LOGICA)
#         # -----------------------------------------------------------------
        
#         n_pos = len(sequences_pos)
#         n_neg = len(sequences_neg)

#         if n_pos == 0 or n_neg == 0:
#             print(f"  ⚠ Trovati {n_pos} positivi e {n_neg} negativi validi. Skip file.")
#             continue
            
#         # Bilancia sottocampionando la classe più numerosa
#         n_to_keep = min(n_pos, n_neg)
        
#         if n_pos > n_to_keep:
#             print(f"  Bilanciamento: Sottocampiono {n_pos} positivi a {n_to_keep}")
#             sequences_pos = random.sample(sequences_pos, n_to_keep)
        
#         if n_neg > n_to_keep:
#             print(f"  Bilanciamento: Sottocampiono {n_neg} negativi a {n_to_keep}")
#             sequences_neg = random.sample(sequences_neg, n_to_keep)

#         # -----------------------------------------------------------------
        
#         all_sequences.extend(sequences_pos)
#         all_labels.extend([1] * len(sequences_pos))
        
#         all_sequences.extend(sequences_neg)
#         all_labels.extend([0] * len(sequences_neg))
        
#         print(f"\n  RISULTATO FILE: Aggiunti {len(sequences_pos)} positivi e {len(sequences_neg)} negativi (Bilanciati).")
#         print(f"  ✓ Totale sequenze accumulate: {len(all_sequences):,}")

    
#     if not all_sequences:
#         print("⚠⚠ NESSUNA SEQUENZA TROVATA. Controlla i percorsi e i filtri.")
#         return pd.DataFrame(columns=['sequenza', 'nucleosoma'])

#     # Crea DataFrame
#     df = pd.DataFrame({
#         'sequenza': all_sequences,
#         'nucleosoma': all_labels
#     })
    
#     df = df.drop_duplicates(subset=['sequenza']).reset_index(drop=True)
#     df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    
#     print("\n" + "="*60)
#     print("DATASET COMPLETATO (Bilanciato per regione)")
#     print("="*60)
#     print(f"Totale sequenze: {len(df):,}")
#     print(f"Nucleosomi (positivi): {df['nucleosoma'].sum():,}")
#     print(f"Non-nucleosomi (negativi): {(1 - df['nucleosoma']).sum():,}")
    
#     if len(df) > 0:
#         print(f"Bilanciamento: {df['nucleosoma'].sum() / len(df) * 100:.1f}% positivi")
    
#     return df

def create_nucleosome_dataset(wig_files, genome_dict,  
                            max_sequences_per_file=1000,
                            threshold_high=1.0,
                            threshold_low=1.0,
                            max_regions_per_wig=100000,
                            df_features=None,
                            sequence_length=201):
    """
    Crea dataset finale combinando dati WIG e genoma.
    Positivi: score > threshold_high
    Negativi: score == threshold_low
    """
    print("\n" + "="*60)
    print("CREAZIONE DATASET NUCLEOSOMI (Positivi vs Negativi in-loco)")
    print("="*60)
    print(f"Lunghezza sequenza: {sequence_length} bp")
    print(f"Soglia ALTA (positivi): > {threshold_high:.2f}")
    print(f"Soglia BASSA (negativi): == {threshold_low:.2f}")
    print(f"Filtro regioni: {df_features is not None}")
    
    all_sequences = []
    all_labels = []
    
    # Processa TUTTI i file WIG
    for idx, wig_file in enumerate(wig_files, 1):
        print(f"\n[{idx}/{len(wig_files)}] Processando: {os.path.basename(wig_file)}")
        
        # Passa max_regions_per_wig al parser
        df_wig = parse_wig_file(wig_file, max_regions=max_regions_per_wig)
        
        if df_wig is None or len(df_wig) == 0:
            print("  ⚠ Nessun dato trovato, skip")
            continue
            
        # -----------------------------------------------------------------
        # 1. STREAM POSITIVO (Label 1) - Veloce, quindi invariato
        # -----------------------------------------------------------------
        print("\n  --- Stream Positivo (Label 1) ---")
        df_nucleosomes_high = identify_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_high, # Passa threshold_high (1.0)
            nucleosome_length=sequence_length
        )
        
        df_positive_filtered = filter_regions_by_feature(df_nucleosomes_high, df_features)
        if df_positive_filtered.empty:
            print("  ⚠ Nessuna regione positiva (>1.0) nei feature forniti, skip file")
            continue
            
        sequences_pos = extract_sequences_from_genome(
            df_positive_filtered,  
            genome_dict,  
            max_sequences=max_sequences_per_file 
        )
        
        # -----------------------------------------------------------------
        # 2. STREAM NEGATIVO (Label 0) - Logica efficiente
        # -----------------------------------------------------------------
        print("\n  --- Stream Negativo (Label 0) ---")
        
        n_to_find = len(sequences_pos) # Numero di negativi da trovare
        if n_to_find == 0:
             print("  ⚠ Nessuna sequenza positiva trovata, skip negativi")
             continue
        
        df_nucleosomes_low = identify_non_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_low, # Passa threshold_low (1.0)
            nucleosome_length=sequence_length
        )
        
        if df_nucleosomes_low.empty:
            print("  ⚠ Nessuna regione negativa (==1.0) trovata, skip negativi")
            continue
        
        # --- CHIAMATA ALLA NUOVA FUNZIONE EFFICIENTE ---
        sequences_neg = extract_filtered_sequences_efficiently(
            df_nucleosomes_low,     # I milioni di candidati
            df_features,            # I promotori
            genome_dict,
            max_sequences=n_to_find # Il numero da trovare
        )
        
        if not sequences_neg:
             print("  ⚠ Nessuna sequenza negativa valida trovata nei promotori.")
             continue
        
        # -----------------------------------------------------------------
        # 3. BILANCIAMENTO e AGGIUNTA AL DATASET
        # -----------------------------------------------------------------
        
        n_pos = len(sequences_pos)
        n_neg = len(sequences_neg)

        # La nuova funzione 'extract_filtered_sequences_efficiently' 
        # Cerca di trovare n_to_find (che è = n_pos), ma potrebbe trovarne meno.
        # Dobbiamo comunque bilanciare per la classe più piccola.
        
        n_to_keep = min(n_pos, n_neg)
        
        if n_pos > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_pos} positivi a {n_to_keep}")
            sequences_pos = random.sample(sequences_pos, n_to_keep)
        
        if n_neg > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_neg} negativi a {n_to_keep}")
            sequences_neg = random.sample(sequences_neg, n_to_keep)

        # -----------------------------------------------------------------
        
        all_sequences.extend(sequences_pos)
        all_labels.extend([1] * len(sequences_pos))
        
        all_sequences.extend(sequences_neg)
        all_labels.extend([0] * len(sequences_neg))
        
        print(f"\n  RISULTATO FILE: Aggiunti {len(sequences_pos)} positivi e {len(sequences_neg)} negativi (Bilanciati).")
        print(f"  ✓ Totale sequenze accumulate: {len(all_sequences):,}")

    
    if not all_sequences:
        print("⚠⚠ NESSUNA SEQUENZA TROVATA. Controlla i percorsi e i filtri.")
        return pd.DataFrame(columns=['sequenza', 'nucleosoma'])

    # Crea DataFrame
    df = pd.DataFrame({
        'sequenza': all_sequences,
        'nucleosoma': all_labels
    })
    
    df = df.drop_duplicates(subset=['sequenza']).reset_index(drop=True)
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    
    print("\n" + "="*60)
    print("DATASET COMPLETATO (Bilanciato per regione)")
    print("="*60)
    print(f"Totale sequenze: {len(df):,}")
    print(f"Nucleosomi (positivi): {df['nucleosoma'].sum():,}")
    print(f"Non-nucleosomi (negativi): {(1 - df['nucleosoma']).sum():,}")
    
    if len(df) > 0:
        print(f"Bilanciamento: {df['nucleosoma'].sum() / len(df) * 100:.1f}% positivi")
    
    return df



#AUMENTO VELOCITà

def extract_filtered_sequences_efficiently(df_regions, df_features, genome_dict, max_sequences=2000):
    """
    Combina filtro e estrazione in un unico passaggio efficiente.
    Mescola i candidati e si ferma non appena 'max_sequences' sono state trovate.
    """
    print(f"  Avvio estrazione efficiente per {max_sequences} sequenze da {len(df_regions):,} candidati...")
    
    sequences = []
    
    # 1. Crea il dizionario di features per una ricerca rapida
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("    Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return []

    # 2. Mescola i candidati (df_regions)
    #    Usiamo sample(frac=1) che è un modo efficiente in pandas per mescolare
    df_regions_shuffled = df_regions.sample(frac=1)

    processed_count = 0
    
    # 3. Itera, filtra, ed estrai in un unico loop
    for row in df_regions_shuffled.itertuples(index=False):
        processed_count += 1
        
        # Logica di 'filter_regions_by_feature'
        chrom = getattr(row, 'chr', None)
        if chrom not in df_features_by_chr:
            continue
            
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        
        if s is None or e is None:
            continue

        # Cerca sovrapposizione
        has_overlap = False
        for f_start, f_end in df_features_by_chr[chrom]:
            if not (e < f_start or s > f_end):
                has_overlap = True
                break
        
        if not has_overlap:
            continue # Non è in un promotore, passa al prossimo candidato

        # --- Se siamo qui, la regione E' in un promotore ---
        
        # Logica di 'extract_sequences_from_genome'
        start = int(s)
        end = int(e)
        
        if chrom not in genome_dict: # Controllo aggiunto per sicurezza
            continue
            
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        # --- Se siamo qui, la sequenza è VALIDA ---
        sequences.append(seq)
        
        # Logica di 'stop'
        if len(sequences) >= max_sequences:
            print(f"    ✓ Raggiunto limite di {max_sequences} sequenze valide (processati {processed_count:,} candidati).")
            break
            
    if len(sequences) < max_sequences:
        print(f"    ✓ Estrazione completata. Trovate {len(sequences)} sequenze valide (processati tutti i {len(df_regions):,} candidati).")

    return sequences






#################################################################################################
# ESECUZIONE PRINCIPALE
#################################################################################################

def main():
    print("="*60)
    print("ELABORAZIONE DATI NUCLEOSOMI GSE36979")
    print("="*60)

    # Imposta directory dati
    base_dir = os.getcwd()
    data_dir = os.path.abspath(os.path.join(base_dir, "..", "data"))
    
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        print(f"Creata directory dati: {data_dir}")

    # STEP 1: Estrai file WIG dal TAR
    tar_path = os.path.join(data_dir, "GSE36979_RAW.tar")
    wig_extract_dir = os.path.join(data_dir, "wig_files")

    if not os.path.exists(tar_path):
        print(f"✗ File non trovato: {tar_path}")
        print("Scaricare GSE36979_RAW.tar da GEO e posizionarlo in ../data/")
        return

    wig_files = extract_tar_wig_files(tar_path, extract_dir=wig_extract_dir)

    if not wig_files:
        print("✗ Nessun file WIG 147bp trovato nell'archivio. Controlla il file TAR.")
        return

    # STEP 2: Scarica/Carica genoma hg18
    print("\n" + "="*60)
    genome_dir = os.path.join(data_dir, "hg18_genome")

    if os.path.exists(genome_dir) and len(os.listdir(genome_dir)) > 0:
        print("Genoma hg18 già presente, caricamento...")
        genome_dict = load_hg18_genome(genome_dir)
    else:
        print("Scaricamento genoma hg18...")
        print("ATTENZIONE: Il download completo richiede ~1.5GB di spazio e tempo")
        genome_dict = download_hg18_genome(genome_dir)

    if not genome_dict:
        print("✗ Impossibile caricare genoma")
        return

    # STEP 2.5: Carica o scarica annotazioni RefGene per i promotori
    print("\n" + "="*60)
    print("Caricamento/Generazione annotazioni promotori (RefGene hg18)...")
    
    try:
        df_promoters = download_refgene_hg18(output_dir=data_dir, promoter_window=1000)
    except Exception as e:
        print(f"✗ Errore fatale durante il caricamento dei promotori: {e}")
        df_promoters = None

    if df_promoters is None or df_promoters.empty:
        print("✗ Impossibile caricare le annotazioni dei promotori. Uscita.")
        return
    else:
        print(f"✓ Annotazioni promotori caricate: {len(df_promoters)} regioni")
        
    # --- ALTERNATIVA: USA ENHANCER (esempio se avessi un file BED) ---
    # enhancer_bed_path = os.path.join(data_dir, "hg18_enhancers.bed")
    # df_features_to_use = load_bed_features(enhancer_bed_path)
    # ------------------------------------------------------------------
    df_features_to_use = df_promoters # Scegliamo di usare i promotori


    # --- MODIFICA 5: Chiamata principale con i nuovi parametri ---
    # STEP 3: Crea dataset
    df_final = create_nucleosome_dataset(
        wig_files,  
        genome_dict,
        max_sequences_per_file=100000,
        threshold_high=2.0,           # Positivi > 1.0
        threshold_low=1.0,            # Negativi == 1.0
        max_regions_per_wig=5000000,  # Leggi fino a 5 milioni di regioni per file
        df_features=df_features_to_use,
        sequence_length=201
    )

    if df_final is None or df_final.empty:
        print("✗ Creazione dataset fallita o dataset vuoto.")
        return

    # STEP 4: Salva risultati
    output_file = os.path.join(data_dir, "nucleosome_dataset_GSE36979_promoters_201pb_balanced.csv")
    try:
        df_final.to_csv(output_file, index=False)
        print(f"\n✓ Dataset salvato: {output_file}")
    except Exception as e:
        print(f"✗ Errore salvataggio dataset: {e}")
        return

    # STEP 5: Statistiche finali
    print("\n" + "="*60)
    print("STATISTICHE FINALI")
    print("="*60)
    print(df_final.head(10))
    
    print(f"\nDistribuzione lunghezze:")
    print(df_final['sequenza'].str.len().describe())

    # GC content
    df_final['gc_content'] = df_final['sequenza'].apply(calc_gc)
    print(f"\nGC content per classe:")
    print(df_final.groupby('nucleosoma')['gc_content'].describe())

    # Plot
    print("\nGenerazione plot GC content...")
    try:
        plt.figure(figsize=(10, 6))
        sns.kdeplot(data=df_final, x='gc_content', hue='nucleosoma', fill=True, common_norm=False, palette="Set1")
        plt.title('Distribuzione GC Content tra Classi (0=Bassa, 1=Alta Occupazione)')
        plot_file = os.path.join(data_dir, "gc_content_distribution.png")
        plt.savefig(plot_file)
        print(f"✓ Plot salvato: {plot_file}")
        # plt.show() # Decommenta se sei in un notebook
    except Exception as e:
        print(f"✗ Errore generazione plot: {e}")

    print("\n✓ ELABORAZIONE COMPLETATA!")

if __name__ == "__main__":  
    main()
```

```python
#SOTTO IL CODICE PER GARANTIRE VERI NEGATIVI LONTANI DAI POSITIVI


import pandas as pd
import numpy as np
import requests
import gzip
import os
from Bio import SeqIO
from io import StringIO
import tarfile
import random
import matplotlib.pyplot as plt
import seaborn as sns
from intervaltree import IntervalTree # Importazione necessaria

#################################################################################################
# FUNZIONI HELPER
#################################################################################################

def calc_gc(seq):
    """Calcola il contenuto GC di una sequenza."""
    if not isinstance(seq, str) or len(seq) == 0:
        return 0
    gc = seq.count('G') + seq.count('C')
    return (gc / len(seq)) * 100

#################################################################################################
# STEP 1: GESTIONE GENOMA HG18
#################################################################################################

def download_hg18_genome(output_dir="hg18_genome"):
    """
    Scarica il genoma umano hg18 da UCSC
    """
    print("Scaricando genoma hg18 da UCSC...")
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # URL base UCSC per hg18
    base_url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg18/chromosomes/"
    
    # Scarica cromosomi principali (1-22, X, Y)
    chromosomes = [f"chr{i}" for i in range(1, 23)] + ["chrX", "chrY"]
    
    genome_dict = {}
    
    for chrom in chromosomes:
        file_name = f"{chrom}.fa.gz"
        file_path = os.path.join(output_dir, file_name)
        
        # Percorso del file decompresso
        unzipped_file_path = file_path.replace('.gz', '')

        if os.path.exists(unzipped_file_path):
            print(f"✓ {chrom} già scaricato")
            # Carica sequenza già presente
            try:
                with open(unzipped_file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        genome_dict[chrom] = str(record.seq).upper() # Salva in maiuscolo
                continue
            except Exception as e:
                print(f"✗ Errore nel caricamento di {unzipped_file_path}: {e}. Tento di riscaricarlo.")

        
        url = base_url + file_name
        print(f"Scaricando {chrom}...")
        
        try:
            response = requests.get(url, stream=True, timeout=300)
            if response.status_code == 200:
                # Salva file compresso
                with open(file_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                
                # Decomprimi e leggi
                sequence_content = ""
                with gzip.open(file_path, 'rt') as f:
                    # Leggi tutto il contenuto
                    sequence_content = f.read()
                
                # Parse dalla stringa
                with StringIO(sequence_content) as f_in:
                    records = SeqIO.parse(f_in, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Salva in maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} scaricato: {len(seq_str):,} bp")
                
                # Salva versione decompressa
                with open(unzipped_file_path, 'w') as f_out:
                    f_out.write(sequence_content)
                
                # Rimuovi file compresso per salvare spazio
                os.remove(file_path)
                
        except Exception as e:
            print(f"✗ Errore scaricamento {chrom}: {e}")
            if os.path.exists(file_path):
                os.remove(file_path) # Pulisci file corrotto
            continue
    
    print(f"\n✓ Genoma hg18 scaricato: {len(genome_dict)} cromosomi")
    return genome_dict

def load_hg18_genome(genome_dir="hg18_genome"):
    """
    Carica genoma hg18 già scaricato
    """
    print("Caricamento genoma hg18...")
    
    genome_dict = {}
    
    chromosomes = [f"chr{i}" for i in range(1, 23)] + ["chrX", "chrY"]
    
    for chrom in chromosomes:
        file_path = os.path.join(genome_dir, f"{chrom}.fa")
        
        if os.path.exists(file_path):
            try:
                with open(file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Assicurati sia maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} caricato: {len(seq_str):,} bp")
            except Exception as e:
                print(f"✗ Errore caricamento {file_path}: {e}")
        else:
            print(f"✗ {chrom} non trovato in {genome_dir}")
    
    if not genome_dict:
        print(f"✗ Nessun file genoma trovato in {genome_dir}. Prova a scaricarlo prima.")
    else:
        print(f"\n✓ Genoma caricato: {len(genome_dict)} cromosomi")
    return genome_dict

#################################################################################################
# STEP 2: GESTIONE DATI WIG (GSE36979)
#################################################################################################

def parse_wig_file(wig_file_path, max_regions=100000):
    """
    Parsifica file WIG con dati di occupancy dei nucleosomi
    """
    print(f"  Parsing file WIG: {wig_file_path}")
    
    regions = []
    current_chrom = None
    current_start = None
    current_step = None
    current_span = 1
    position = 0
    lines_processed = 0
    
    try:
        # Gestisci file compressi o no
        opener = gzip.open if wig_file_path.endswith('.gz') else open
        
        with opener(wig_file_path, 'rt') as f:
            for line in f:
                line = line.strip()
                
                if not line or line.startswith('track') or line.startswith('browser'):
                    continue
                
                lines_processed += 1
                
                # --- MODIFICA 1: Riduzione print ---
                if lines_processed % 50000000 == 0:
                    print(f"    ... Processate {lines_processed:,} linee, trovate {len(regions):,} regioni...")
                
                # Stop se abbiamo raggiunto il limite (per debug)
                if len(regions) >= max_regions:
                    print(f"    ⚠ Raggiunto limite di {max_regions:,} regioni, stop parsing")
                    break
                
                # fixedStep format
                if line.startswith('fixedStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('start='):
                            current_start = int(part.split('=')[1])
                            position = current_start
                        elif part.startswith('step='):
                            current_step = int(part.split('=')[1])
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                
                # variableStep format
                elif line.startswith('variableStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                    current_step = None
                
                # Data line
                else:
                    try:
                        if current_step:
                            # fixedStep: solo valore
                            value = float(line)
                            if value > 0: # Solo regioni con segnale
                                regions.append({
                                    'chr': current_chrom,
                                    'start': position,
                                    'end': position + current_span,
                                    'score': value
                                })
                            position += current_step
                        else:
                            # variableStep: posizione e valore
                            pos, value = line.split()
                            pos = int(pos)
                            value = float(value)
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': pos,
                                    'end': pos + current_span,
                                    'score': value
                                })
                    except Exception:
                        # Salta righe dati malformate
                        continue
            
        print(f"    ✓ Trovate {len(regions):,} regioni con segnale")
        return pd.DataFrame(regions)
    
    except Exception as e:
        print(f"    ✗ Errore parsing WIG: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

def extract_tar_wig_files(tar_path, extract_dir="wig_files", filter_147bp=True):
    """
    Estrae file WIG da archivio TAR
    Se filter_147bp=True, estrae solo file con frammenti da 147bp (nucleosomi canonici)
    """
    print(f"Estraendo file da {tar_path}...")
    
    if not os.path.exists(extract_dir):
        os.makedirs(extract_dir)
    
    extracted_files = []
    skipped_files = []
    
    try:
        with tarfile.open(tar_path, 'r') as tar:
            members = tar.getmembers()
            print(f"Trovati {len(members)} file nell'archivio")
            
            if filter_147bp:
                print("Filtro attivo: solo file con frammenti 147bp (nucleosomi canonici)")
            
            for member in members:
                if member.name.endswith('.wig') or member.name.endswith('.wig.gz'):
                    
                    # ---PROVA A USARE SOLO UN FILE
                    if 'GSM907790' not in member.name:
                        continue
                    # ----FINE PROVA UN FILE 
                    
                    # --- MODIFICA 2: Esclusione file anomalo ---
                    if "GSM907789" in member.name:
                        print(f"File anomalo saltato: {member.name}")
                        continue
                    # ------------------------------------------
                    
                    # Filtra solo file 147bp se richiesto
                    if filter_147bp and '_147.wig' not in member.name:
                        skipped_files.append(member.name)
                        continue
                    
                    # Logica di estrazione (controlla se esiste già)
                    target_path = os.path.join(extract_dir, member.name)
                    if not os.path.exists(target_path):
                        print(f"Estraendo: {member.name}")
                        tar.extract(member, extract_dir)
                    
                    extracted_files.append(target_path) 
            
        print(f"\n✓ Estratti {len(extracted_files)} file WIG (nucleosomi 147bp)")
        if skipped_files:
            print(f"✓ Saltati {len(skipped_files)} file con altre dimensioni")
        
        return extracted_files
    
    except Exception as e:
        print(f"✗ Errore estrazione TAR: {e}")
        return []

#################################################################################################
# STEP 3: GESTIONE ANNOTAZIONI (PROMOTORI / ENHANCER)
#################################################################################################

def load_genomic_features(gtf_path,  
                          feature_types,  
                          relative_to_tss=False,  
                          window_upstream=0,  
                          window_downstream=0):
    """
    Estrae coordinate per specifici feature genomici (es. esoni, promotori)
    da un file di annotazione GTF.
    """
    print(f"Caricamento annotazioni da {gtf_path} per feature: {feature_types}...")
    features_list = []

    try:
        with open(gtf_path, 'r') as f:
            for line in f:
                if line.startswith('#'):
                    continue
                parts = line.strip().split('\t')
                if len(parts) < 9:
                    continue
                
                chrom, source, feature, start_str, end_str, score, strand, frame, attrs = parts

                if feature in feature_types:
                    try:
                        start = int(start_str)
                        end = int(end_str)
                        
                        reg_start, reg_end = 0, 0

                        if relative_to_tss:
                            # Calcola regione relativa al TSS (per promotori)
                            tss = start if strand == '+' else end
                            reg_start = max(0, tss - window_upstream)
                            reg_end = tss + window_downstream
                        else:
                            # Usa coordinate dirette del feature + finestre
                            if strand == '+':
                                reg_start = max(0, start - window_upstream)
                                reg_end = end + window_downstream
                            else:
                                reg_start = max(0, start - window_downstream)
                                reg_end = end + window_upstream

                        features_list.append((chrom, reg_start, reg_end))
                        
                    except ValueError:
                        continue

    except FileNotFoundError:
        print(f"Errore: File GTF non trovato a {gtf_path}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
    except Exception as e:
        print(f"Errore during GTF read: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

    if not features_list:
        print(f"⚠ Nessun feature di tipo {feature_types} trovato nel file.")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

    df_features = pd.DataFrame(features_list, columns=['chr', 'start', 'end'])
    df_features = df_features.drop_duplicates().reset_index(drop=True)
    
    print(f"✓ Trovati {len(df_features):,} feature unici per '{', '.join(feature_types)}'")
    return df_features

def load_bed_features(bed_path):
    """
    Carica un file BED semplice (chr, start, end).
    """
    print(f"Caricamento features da file BED: {os.path.basename(bed_path)}")
    
    if not os.path.exists(bed_path):
        print(f"✗ File BED non trovato: {bed_path}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
        
    try:
        df = pd.read_csv(
            bed_path,  
            sep='\t',  
            header=None,
            comment='#', # Ignora righe di commento
            usecols=[0, 1, 2],  
            names=['chr', 'start', 'end']
        )
        
        df = df.dropna().drop_duplicates().reset_index(drop=True)
        df['start'] = df['start'].astype(int)
        df['end'] = df['end'].astype(int)
        
        print(f"✓ Trovati {len(df):,} features unici")
        return df
        
    except Exception as e:
        print(f"✗ Errore caricamento file BED: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

def get_feature_dataframe(gtf_path, region_key, promoter_window=1000):
    """
    Funzione "manager" che carica le coordinate per una regione genomica
    specificata da una chiave (key).
    """
    print(f"\n--- Richiesta regione: '{region_key}' ---")
    
    feature_map = {
        "promoter": (["transcript", "gene"], True, promoter_window, promoter_window),
        "exon": (["exon"], False, 0, 0),
        "gene_body": (["gene"], False, 0, 0),
        "utr5": (["five_prime_utr", "5UTR"], False, 0, 0),
        "utr3": (["three_prime_utr", "3UTR"], False, 0, 0),
        "enhancer": (["enhancer"], False, 0, 0), # Per file GTF
    }
    
    if region_key not in feature_map:
        print(f"Errore: Chiave regione '{region_key}' non riconosciuta.")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
        
    params = feature_map[region_key]
    
    return load_genomic_features(gtf_path,  
                                 feature_types=params[0],  
                                 relative_to_tss=params[1],  
                                 window_upstream=params[2],  
                                 window_downstream=params[3])

def download_refgene_hg18(output_dir="../data", promoter_window=1000):
    """
    Scarica le annotazioni RefSeq (hg18) da UCSC e genera regioni promotrici ±promoter_window bp dal TSS.
    """
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg18/database/refGene.txt.gz"
    local_gz = os.path.join(output_dir, "refGene.txt.gz")
    local_csv = os.path.join(output_dir, f"promoters_hg18_{promoter_window}bp.csv")

    if os.path.exists(local_csv):
        print(f"✓ File promotori già presente, caricamento: {local_csv}")
        try:
            return pd.read_csv(local_csv)
        except Exception as e:
            print(f"✗ Errore nel caricamento di {local_csv}: {e}. Tento di rigenerarlo.")

    if not os.path.exists(local_gz):
        print(f"Scaricando RefSeq hg18 da UCSC...")
        try:
            response = requests.get(url, stream=True, timeout=180)
            response.raise_for_status()
            with open(local_gz, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"✓ File salvato: {local_gz}")
        except Exception as e:
            print(f"✗ Errore nel download del file RefSeq: {e}")
            return None
    else:
        print(f"✓ File RefSeq già presente: {local_gz}")

    cols = [
        'bin', 'name', 'chrom', 'strand', 'txStart', 'txEnd',
        'cdsStart', 'cdsEnd', 'exonCount', 'exonStarts', 'exonEnds',
        'score', 'name2', 'cdsStartStat', 'cdsEndStat', 'exonFrames'
    ]
    
    print("Caricamento e parsing refGene...")
    try:
        refgene = pd.read_csv(local_gz, sep='\t', header=None, names=cols, compression='gzip')
    except Exception as e:
        print(f"✗ Errore parsing refGene: {e}")
        return None
        
    print(f"✓ {len(refgene):,} geni caricati")

    print(f"Calcolo regioni promotrici ±{promoter_window} bp dal TSS...")
    promoter_records = []

    for _, row in refgene.iterrows():
        chrom = row['chrom']
        if not isinstance(chrom, str) or not chrom.startswith("chr"):
            continue
        
        try:
            txStart = int(row['txStart'])
            txEnd = int(row['txEnd'])
            
            if row['strand'] == '+':
                tss = txStart
            else:
                tss = txEnd
            
            start = max(0, tss - promoter_window)
            end = tss + promoter_window

            promoter_records.append({
                'chr': chrom,
                'start': int(start),
                'end': int(end),
                'strand': row['strand'],
                'gene': row['name2']
            })
        except ValueError:
            continue

    df_promoters = pd.DataFrame(promoter_records)
    df_promoters = df_promoters.drop_duplicates(subset=['chr', 'start', 'end'])
    print(f"✓ {len(df_promoters):,} regioni promotrici uniche generate")

    try:
        df_promoters.to_csv(local_csv, index=False)
        print(f"✓ Salvati promotori in: {local_csv}")
    except Exception as e:
        print(f"✗ Errore salvataggio CSV promotori: {e}")
    
    return df_promoters

#################################################################################################
# STEP 4: CREAZIONE DATASET (Logica Positivo vs Negativo)
#################################################################################################

def filter_regions_by_feature(df_regions, df_features):
    """
    Mantiene solo le regioni (es. nucleosomi) che si sovrappongono  
    con i feature forniti (es. promotori, esoni).
    Restituisce un nuovo DataFrame filtrato.
    """
    print("  Filtraggio regioni per sovrapposizione con feature...")

    if df_features is None or df_features.empty:
        print("  ⚠ Nessuna annotazione feature fornita, ritorno tutte le regioni (nessun filtro).")
        return df_regions

    # Crea dizionario cromosoma → array di (start, end) per efficienza
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("  Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return pd.DataFrame(columns=df_regions.columns)
        
    keep = np.zeros(len(df_regions), dtype=bool)

    for i, row in enumerate(df_regions.itertuples(index=False)):
        chrom = getattr(row, 'chr', None)
        if chrom not in df_features_by_chr:
            continue
            
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        
        if s is None or e is None:
            continue

        # Cerca sovrapposizione
        for f_start, f_end in df_features_by_chr[chrom]:
            # (e < f_start) -> la regione è completamente prima del feature
            # (s > f_end)   -> la regione è completamente dopo il feature
            if not (e < f_start or s > f_end):  
                keep[i] = True
                break # Trovata sovrapposizione, passa alla prossima regione

    filtered_df = df_regions[keep]
    print(f"  ✓ {len(filtered_df):,}/{len(df_regions):,} regioni trovate nei feature forniti.")
    return filtered_df.reset_index(drop=True)

# --- MODIFICA 3: Funzione POSITIVI (score > 1.0) ---
def identify_nucleosome_positions(df_wig, threshold_value=1.0, nucleosome_length=201):
    """
    Identifica posizioni nucleosomi (ALTA occupazione)
    usando una SOGLIA ASSOLUTA (score > threshold)
    """
    print("  Identificazione posizioni nucleosomi (ALTA occupazione)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    # Usa il valore assoluto passato
    print(f"    Soglia segnale (Positivi): > {threshold_value:.2f}")
    
    # MODIFICA CRITICA: Cerca i punteggi STRETTAMENTE MAGGIORI del background
    df_nucleosomes = df_wig[df_wig['score'] > threshold_value].copy()
    
    # Estendi regioni alla lunghezza desiderata
    df_nucleosomes['center'] = (df_nucleosomes['start'] + df_nucleosomes['end']) // 2
    df_nucleosomes['start'] = df_nucleosomes['center'] - nucleosome_length // 2
    df_nucleosomes['end'] = df_nucleosomes['start'] + nucleosome_length
    
    print(f"    ✓ Identificati {len(df_nucleosomes)} nucleosomi (Positivi)")
    
    return df_nucleosomes[['chr', 'start', 'end', 'score']]

# --- MODIFICA 4: Funzione NEGATIVI (score == 1.0) ---
def identify_non_nucleosome_positions(df_wig, threshold_value=1.0, nucleosome_length=201):
    """
    Identifica posizioni NON-nucleosomi (BASSA occupazione)
    usando una SOGLIA ASSOLUTA (score == threshold)
    """
    print("  Identificazione posizioni NON-nucleosomi (BASSA occupazione)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    # Usa il valore assoluto passato
    print(f"    Soglia segnale (Negativi): == {threshold_value:.2f}")
    
    # Filtra regioni con segnale BASSO (esattamente il background)
    df_nucleosomes = df_wig[df_wig['score'] == threshold_value].copy()
    
    # Estendi regioni alla lunghezza desiderata
    df_nucleosomes['center'] = (df_nucleosomes['start'] + df_nucleosomes['end']) // 2
    df_nucleosomes['start'] = df_nucleosomes['center'] - nucleosome_length // 2
    df_nucleosomes['end'] = df_nucleosomes['start'] + nucleosome_length
    
    print(f"    ✓ Identificati {len(df_nucleosomes)} NON-nucleosomi (Negativi)")
    
    return df_nucleosomes[['chr', 'start', 'end', 'score']]

def extract_sequences_from_genome(df_regions, genome_dict, max_sequences=2000):
    """
    Estrae sequenze DNA dalle regioni specificate
    """
    print(f"    Estrazione sequenze da {len(df_regions)} regioni...")
    
    sequences = []
    
    count = 0
    processed = 0
    for idx, row in df_regions.iterrows():
        processed += 1
        if count >= max_sequences:
            print(f"      Raggiunto limite di {max_sequences} sequenze.")
            break
        
        chrom = row['chr']
        start = int(row['start'])
        end = int(row['end'])
        
        if chrom not in genome_dict:
            continue
        
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        sequences.append(seq)
        count += 1
        
        if count % 500 == 0 and count > 0:
            print(f"      Estraendole... {count}/{max_sequences} sequenze valide trovate (processate {processed} regioni)")
    
    print(f"    ✓ Estratte {len(sequences)} sequenze")
    return sequences

# --- MODIFICA 5: NUOVA Funzione di Estrazione Efficiente (con Buffer Zone) ---
def extract_filtered_sequences_efficiently(df_regions, df_features, genome_dict, positive_regions_tree, max_sequences=2000, buffer_zone=201):
    """
    Combina filtro e estrazione in un unico passaggio efficiente.
    Mescola i candidati e si ferma non appena 'max_sequences' sono state trovate.
    ESCLUDE i negativi che si sovrappongono ai positivi (buffer_zone).
    """
    print(f"  Avvio estrazione efficiente per {max_sequences} sequenze da {len(df_regions):,} candidati...")
    
    sequences = []
    
    # 1. Crea il dizionario di features per una ricerca rapida
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("    Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return []

    # 2. Mescola i candidati (df_regions)
    #    Usiamo sample(frac=1) che è un modo efficiente in pandas per mescolare
    df_regions_shuffled = df_regions.sample(frac=1)

    processed_count = 0
    excluded_by_buffer = 0
    
    # 3. Itera, filtra, ed estrai in un unico loop
    for row in df_regions_shuffled.itertuples(index=False):
        processed_count += 1
        
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        chrom = getattr(row, 'chr', None)
        
        if s is None or e is None or chrom is None:
            continue
            
        # --- NUOVO CONTROLLO: ZONA CUSCINETTO ---
        # Controlla se questa regione negativa (s, e) è troppo vicina a un positivo.
        # Creiamo un intervallo buffer attorno ad essa.
        buffer_start = s - buffer_zone
        buffer_end = e + buffer_zone
        if chrom in positive_regions_tree and positive_regions_tree[chrom].overlaps(buffer_start, buffer_end):
            excluded_by_buffer += 1
            continue # Questo negativo è troppo vicino a un positivo, scartalo.
        # ----------------------------------------

        # Controllo 2: Logica di 'filter_regions_by_feature' (promotori)
        if chrom not in df_features_by_chr:
            continue
            
        has_overlap = False
        for f_start, f_end in df_features_by_chr[chrom]:
            if not (e < f_start or s > f_end):
                has_overlap = True
                break
        
        if not has_overlap:
            continue # Non è in un promotore, passa al prossimo candidato

        # --- Se siamo qui, la regione E' in un promotore E NON è vicina a un positivo ---
        
        # Logica di 'extract_sequences_from_genome'
        start = int(s)
        end = int(e)
        
        if chrom not in genome_dict:
            continue
            
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        # --- Se siamo qui, la sequenza è VALIDA ---
        sequences.append(seq)
        
        # Logica di 'stop'
        if len(sequences) >= max_sequences:
            print(f"    ✓ Raggiunto limite di {max_sequences} sequenze valide (processati {processed_count:,} candidati).")
            print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")
            break
            
    if len(sequences) < max_sequences:
        print(f"    ✓ Estrazione completata. Trovate {len(sequences)} sequenze valide (processati tutti i {len(df_regions):,} candidati).")
        print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")

    return sequences


# --- MODIFICA 6: Funzione DATASET (usa la funzione efficiente e l'albero) ---
def create_nucleosome_dataset(wig_files, genome_dict,  
                            max_sequences_per_file=1000,
                            threshold_high=1.0,
                            threshold_low=1.0,
                            max_regions_per_wig=100000,
                            df_features=None,
                            sequence_length=201):
    """
    Crea dataset finale combinando dati WIG e genoma.
    Positivi: score > threshold_high
    Negativi: score == threshold_low
    """
    print("\n" + "="*60)
    print("CREAZIONE DATASET NUCLEOSOMI (Positivi vs Negativi in-loco)")
    print("="*60)
    print(f"Lunghezza sequenza: {sequence_length} bp")
    print(f"Soglia ALTA (positivi): > {threshold_high:.2f}")
    print(f"Soglia BASSA (negativi): == {threshold_low:.2f}")
    print(f"Filtro regioni: {df_features is not None}")
    
    all_sequences = []
    all_labels = []
    
    # Processa TUTTI i file WIG
    for idx, wig_file in enumerate(wig_files, 1):
        print(f"\n[{idx}/{len(wig_files)}] Processando: {os.path.basename(wig_file)}")
        
        # Passa max_regions_per_wig al parser
        df_wig = parse_wig_file(wig_file, max_regions=max_regions_per_wig)
        
        if df_wig is None or len(df_wig) == 0:
            print("  ⚠ Nessun dato trovato, skip")
            continue
            
        # -----------------------------------------------------------------
        # 1. STREAM POSITIVO (Label 1)
        # -----------------------------------------------------------------
        print("\n  --- Stream Positivo (Label 1) ---")
        # 1a. Trova TUTTI i positivi (>1.0) nell'INTERO GENOMA
        df_nucleosomes_high = identify_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_high, # Passa threshold_high (1.0)
            nucleosome_length=sequence_length
        )
        
        # 1b. Filtra i positivi SOLO per i promotori (per il set di training)
        df_positive_filtered = filter_regions_by_feature(df_nucleosomes_high, df_features)
        if df_positive_filtered.empty:
            print("  ⚠ Nessuna regione positiva (>1.0) nei feature forniti, skip file")
            continue
            
        # 1c. Estrai sequenze
        sequences_pos = extract_sequences_from_genome(
            df_positive_filtered,  
            genome_dict,  
            max_sequences=max_sequences_per_file 
        )
        
        # --- NUOVO PASSAGGIO: Crea l'albero dei positivi per la zona cuscinetto ---
        # Usa la lista TOTALE (df_nucleosomes_high) per l'albero
        print("    Creazione albero intervalli positivi per zona cuscinetto...")
        positive_tree = {}
        for chrom in df_nucleosomes_high['chr'].unique():
            positive_tree[chrom] = IntervalTree()
            for row in df_nucleosomes_high[df_nucleosomes_high['chr'] == chrom].itertuples():
                positive_tree[chrom].addi(row.start, row.end)
        # ---------------------------------------------------------------------

        # -----------------------------------------------------------------
        # 2. STREAM NEGATIVO (Label 0) - Logica efficiente
        # -----------------------------------------------------------------
        print("\n  --- Stream Negativo (Label 0) ---")
        
        n_to_find = len(sequences_pos) # Numero di negativi da trovare
        if n_to_find == 0:
             print("  ⚠ Nessuna sequenza positiva trovata, skip negativi")
             continue
        
        # 2a. Trova TUTTI i candidati negativi (==1.0)
        df_nucleosomes_low = identify_non_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_low, # Passa threshold_low (1.0)
            nucleosome_length=sequence_length
        )
        
        if df_nucleosomes_low.empty:
            print("  ⚠ Nessuna regione negativa (==1.0) trovata, skip negativi")
            continue
        
        # 2b. --- CHIAMATA ALLA NUOVA FUNZIONE EFFICIENTE (CON ALBERO) ---
        sequences_neg = extract_filtered_sequences_efficiently(
            df_nucleosomes_low,     # I milioni di candidati
            df_features,            # I promotori
            genome_dict,
            positive_tree,          # L'albero dei positivi da evitare
            max_sequences=n_to_find # Il numero da trovare
        )
        
        if not sequences_neg:
             print("  ⚠ Nessuna sequenza negativa valida trovata (o tutte troppo vicine ai positivi).")
             continue
        
        # -----------------------------------------------------------------
        # 3. BILANCIAMENTO e AGGIUNTA AL DATASET
        # -----------------------------------------------------------------
        
        n_pos = len(sequences_pos)
        n_neg = len(sequences_neg)

        n_to_keep = min(n_pos, n_neg)
        
        if n_pos > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_pos} positivi a {n_to_keep}")
            sequences_pos = random.sample(sequences_pos, n_to_keep)
        
        if n_neg > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_neg} negativi a {n_to_keep}")
            sequences_neg = random.sample(sequences_neg, n_to_keep)

        # -----------------------------------------------------------------
        
        all_sequences.extend(sequences_pos)
        all_labels.extend([1] * len(sequences_pos))
        
        all_sequences.extend(sequences_neg)
        all_labels.extend([0] * len(sequences_neg))
        
        print(f"\n  RISULTATO FILE: Aggiunti {len(sequences_pos)} positivi e {len(sequences_neg)} negativi (Bilanciati).")
        print(f"  ✓ Totale sequenze accumulate: {len(all_sequences):,}")

    
    if not all_sequences:
        print("⚠⚠ NESSUNA SEQUENZA TROVATA. Controlla i percorsi e i filtri.")
        return pd.DataFrame(columns=['sequenza', 'nucleosoma'])

    # Crea DataFrame
    df = pd.DataFrame({
        'sequenza': all_sequences,
        'nucleosoma': all_labels
    })
    
    df = df.drop_duplicates(subset=['sequenza']).reset_index(drop=True)
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    
    print("\n" + "="*60)
    print("DATASET COMPLETATO (Bilanciato per regione)")
    print("="*60)
    print(f"Totale sequenze: {len(df):,}")
    print(f"Nucleosomi (positivi): {df['nucleosoma'].sum():,}")
    print(f"Non-nucleosomi (negativi): {(1 - df['nucleosoma']).sum():,}")
    
    if len(df) > 0:
        print(f"Bilanciamento: {df['nucleosoma'].sum() / len(df) * 100:.1f}% positivi")
    
    return df

#################################################################################################
# ESECUZIONE PRINCIPALE
#################################################################################################

def main():
    print("="*60)
    print("ELABORAZIONE DATI NUCLEOSOMI GSE36979")
    print("="*60)

    # Imposta directory dati
    base_dir = os.getcwd()
    data_dir = os.path.abspath(os.path.join(base_dir, "..", "data"))
    
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        print(f"Creata directory dati: {data_dir}")

    # STEP 1: Estrai file WIG dal TAR
    tar_path = os.path.join(data_dir, "GSE36979_RAW.tar")
    wig_extract_dir = os.path.join(data_dir, "wig_files")

    if not os.path.exists(tar_path):
        print(f"✗ File non trovato: {tar_path}")
        print("Scaricare GSE36979_RAW.tar da GEO e posizionarlo in ../data/")
        return

    wig_files = extract_tar_wig_files(tar_path, extract_dir=wig_extract_dir)

    if not wig_files:
        print("✗ Nessun file WIG 147bp trovato nell'archivio. Controlla il file TAR.")
        return

    # STEP 2: Scarica/Carica genoma hg18
    print("\n" + "="*60)
    genome_dir = os.path.join(data_dir, "hg18_genome")

    if os.path.exists(genome_dir) and len(os.listdir(genome_dir)) > 0:
        print("Genoma hg18 già presente, caricamento...")
        genome_dict = load_hg18_genome(genome_dir)
    else:
        print("Scaricamento genoma hg18...")
        print("ATTENZIONE: Il download completo richiede ~1.5GB di spazio e tempo")
        genome_dict = download_hg18_genome(genome_dir)

    if not genome_dict:
        print("✗ Impossibile caricare genoma")
        return

    # STEP 2.5: Carica o scarica annotazioni RefGene per i promotori
    print("\n" + "="*60)
    print("Caricamento/Generazione annotazioni promotori (RefGene hg18)...")
    
    try:
        df_promoters = download_refgene_hg18(output_dir=data_dir, promoter_window=1000)
    except Exception as e:
        print(f"✗ Errore fatale durante il caricamento dei promotori: {e}")
        df_promoters = None

    if df_promoters is None or df_promoters.empty:
        print("✗ Impossibile caricare le annotazioni dei promotori. Uscita.")
        return
    else:
        print(f"✓ Annotazioni promotori caricate: {len(df_promoters)} regioni")
        
    # --- ALTERNATIVA: USA ENHANCER (esempio se avessi un file BED) ---
    # enhancer_bed_path = os.path.join(data_dir, "hg18_enhancers.bed")
    # df_features_to_use = load_bed_features(enhancer_bed_path)
    # ------------------------------------------------------------------
    df_features_to_use = df_promoters # Scegliamo di usare i promotori


    # --- MODIFICA 7: Chiamata principale con i parametri finali ---
    # STEP 3: Crea dataset
    df_final = create_nucleosome_dataset(
        wig_files,  
        genome_dict,
        max_sequences_per_file=100000,
        threshold_high=2.0,           # Positivi > 1.0
        threshold_low=1.0,            # Negativi == 1.0
        max_regions_per_wig=5000000,  # Leggi fino a 5 milioni di regioni per file
        df_features=df_features_to_use,
        sequence_length=201
    )

    if df_final is None or df_final.empty:
        print("✗ Creazione dataset fallita o dataset vuoto.")
        return

    # STEP 4: Salva risultati
    output_file = os.path.join(data_dir, "nucleosome_dataset_GSE36979_promoters_201pb_balanced_intervalTree.csv")
    try:
        df_final.to_csv(output_file, index=False)
        print(f"\n✓ Dataset salvato: {output_file}")
    except Exception as e:
        print(f"✗ Errore salvataggio dataset: {e}")
        return

    # STEP 5: Statistiche finali
    print("\n" + "="*60)
    print("STATISTICHE FINALI")
    print("="*60)
    print(df_final.head(10))
    
    print(f"\nDistribuzione lunghezze:")
    print(df_final['sequenza'].str.len().describe())

    # GC content
    df_final['gc_content'] = df_final['sequenza'].apply(calc_gc)
    print(f"\nGC content per classe:")
    print(df_final.groupby('nucleosoma')['gc_content'].describe())

    # Plot
    print("\nGenerazione plot GC content...")
    try:
        plt.figure(figsize=(10, 6))
        sns.kdeplot(data=df_final, x='gc_content', hue='nucleosoma', fill=True, common_norm=False, palette="Set1")
        plt.title('Distribuzione GC Content tra Classi (0=Bassa, 1=Alta Occupazione)')
        plot_file = os.path.join(data_dir, "gc_content_distribution.png")
        plt.savefig(plot_file)
        print(f"✓ Plot salvato: {plot_file}")
        # plt.show() # Decommenta se sei in un notebook
    except Exception as e:
        print(f"✗ Errore generazione plot: {e}")

    print("\n✓ ELABORAZIONE COMPLETATA!")

if __name__ == "__main__":  
    main()
```

```python
##CODICE FINALE SOTTO 

import pandas as pd
import numpy as np
import requests
import gzip
import os
from Bio import SeqIO
from io import StringIO
import tarfile
import random
import matplotlib.pyplot as plt
import seaborn as sns
from intervaltree import IntervalTree # Importazione necessaria

#################################################################################################
# FUNZIONI HELPER
#################################################################################################

def calc_gc(seq):
    """Calcola il contenuto GC di una sequenza."""
    if not isinstance(seq, str) or len(seq) == 0:
        return 0
    gc = seq.count('G') + seq.count('C')
    return (gc / len(seq)) * 100

#################################################################################################
# STEP 1: GESTIONE GENOMA HG18
#################################################################################################

def download_hg18_genome(output_dir="hg18_genome"):
    """
    Scarica il genoma umano hg18 da UCSC
    """
    print("Scaricando genoma hg18 da UCSC...")
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # URL base UCSC per hg18
    base_url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg18/chromosomes/"
    
    # Scarica cromosomi principali (1-22, X, Y)
    chromosomes = [f"chr{i}" for i in range(1, 23)] + ["chrX", "chrY"]
    
    genome_dict = {}
    
    for chrom in chromosomes:
        file_name = f"{chrom}.fa.gz"
        file_path = os.path.join(output_dir, file_name)
        
        # Percorso del file decompresso
        unzipped_file_path = file_path.replace('.gz', '')

        if os.path.exists(unzipped_file_path):
            print(f"✓ {chrom} già scaricato")
            # Carica sequenza già presente
            try:
                with open(unzipped_file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        genome_dict[chrom] = str(record.seq).upper() # Salva in maiuscolo
                continue
            except Exception as e:
                print(f"✗ Errore nel caricamento di {unzipped_file_path}: {e}. Tento di riscaricarlo.")

        
        url = base_url + file_name
        print(f"Scaricando {chrom}...")
        
        try:
            response = requests.get(url, stream=True, timeout=300)
            if response.status_code == 200:
                # Salva file compresso
                with open(file_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                
                # Decomprimi e leggi
                sequence_content = ""
                with gzip.open(file_path, 'rt') as f:
                    # Leggi tutto il contenuto
                    sequence_content = f.read()
                
                # Parse dalla stringa
                with StringIO(sequence_content) as f_in:
                    records = SeqIO.parse(f_in, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Salva in maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} scaricato: {len(seq_str):,} bp")
                
                # Salva versione decompressa
                with open(unzipped_file_path, 'w') as f_out:
                    f_out.write(sequence_content)
                
                # Rimuovi file compresso per salvare spazio
                os.remove(file_path)
                
        except Exception as e:
            print(f"✗ Errore scaricamento {chrom}: {e}")
            if os.path.exists(file_path):
                os.remove(file_path) # Pulisci file corrotto
            continue
    
    print(f"\n✓ Genoma hg18 scaricato: {len(genome_dict)} cromosomi")
    return genome_dict

def load_hg18_genome(genome_dir="hg18_genome"):
    """
    Carica genoma hg18 già scaricato
    """
    print("Caricamento genoma hg18...")
    
    genome_dict = {}
    
    chromosomes = [f"chr{i}" for i in range(1, 23)] + ["chrX", "chrY"]
    
    for chrom in chromosomes:
        file_path = os.path.join(genome_dir, f"{chrom}.fa")
        
        if os.path.exists(file_path):
            try:
                with open(file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Assicurati sia maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} caricato: {len(seq_str):,} bp")
            except Exception as e:
                print(f"✗ Errore caricamento {file_path}: {e}")
        else:
            print(f"✗ {chrom} non trovato in {genome_dir}")
    
    if not genome_dict:
        print(f"✗ Nessun file genoma trovato in {genome_dir}. Prova a scaricarlo prima.")
    else:
        print(f"\n✓ Genoma caricato: {len(genome_dict)} cromosomi")
    return genome_dict

#################################################################################################
# STEP 2: GESTIONE DATI WIG (GSE36979)
#################################################################################################

def parse_wig_file(wig_file_path, max_regions=100000):
    """
    Parsifica file WIG con dati di occupancy dei nucleosomi
    """
    print(f"  Parsing file WIG: {wig_file_path}")
    
    regions = []
    current_chrom = None
    current_start = None
    current_step = None
    current_span = 1
    position = 0
    lines_processed = 0
    
    try:
        # Gestisci file compressi o no
        opener = gzip.open if wig_file_path.endswith('.gz') else open
        
        with opener(wig_file_path, 'rt') as f:
            for line in f:
                line = line.strip()
                
                if not line or line.startswith('track') or line.startswith('browser'):
                    continue
                
                lines_processed += 1
                
                # --- MODIFICA 1: Riduzione print ---
                if lines_processed % 50000000 == 0:
                    print(f"    ... Processate {lines_processed:,} linee, trovate {len(regions):,} regioni...")
                
                # Stop se abbiamo raggiunto il limite (per debug)
                if len(regions) >= max_regions:
                    print(f"    ⚠ Raggiunto limite di {max_regions:,} regioni, stop parsing")
                    break
                
                # fixedStep format
                if line.startswith('fixedStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('start='):
                            current_start = int(part.split('=')[1])
                            position = current_start
                        elif part.startswith('step='):
                            current_step = int(part.split('=')[1])
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                
                # variableStep format
                elif line.startswith('variableStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                    current_step = None
                
                # Data line
                else:
                    try:
                        if current_step:
                            # fixedStep: solo valore
                            value = float(line)
                            if value > 0: # Solo regioni con segnale
                                regions.append({
                                    'chr': current_chrom,
                                    'start': position,
                                    'end': position + current_span,
                                    'score': value
                                })
                            position += current_step
                        else:
                            # variableStep: posizione e valore
                            pos, value = line.split()
                            pos = int(pos)
                            value = float(value)
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': pos,
                                    'end': pos + current_span,
                                    'score': value
                                })
                    except Exception:
                        # Salta righe dati malformate
                        continue
            
        print(f"    ✓ Trovate {len(regions):,} regioni con segnale")
        return pd.DataFrame(regions)
    
    except Exception as e:
        print(f"    ✗ Errore parsing WIG: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

def extract_tar_wig_files(tar_path, extract_dir="wig_files", filter_147bp=True):
    """
    Estrae file WIG da archivio TAR
    Se filter_147bp=True, estrae solo file con frammenti da 147bp (nucleosomi canonici)
    """
    print(f"Estraendo file da {tar_path}...")
    
    if not os.path.exists(extract_dir):
        os.makedirs(extract_dir)
    
    extracted_files = []
    skipped_files = []
    
    try: # Inizio del blocco 'try'
        with tarfile.open(tar_path, 'r') as tar:
            members = tar.getmembers()
            print(f"Trovati {len(members)} file nell'archivio")
            
            if filter_147bp:
                print("Filtro attivo: solo file con frammenti 147bp (nucleosomi canonici)")
            
            for member in members:
                if member.name.endswith('.wig') or member.name.endswith('.wig.gz'):
                    # ---PROVA A USARE SOLO UN FILE
                    if 'GSM907790' not in member.name:
                        continue
                    # ----FINE PROVA UN FILE 
                    
                    # --- MODIFICA 2: Esclusione file anomalo ---
                    if "GSM907789" in member.name:                            
                        print(f"File anomalo saltato: {member.name}")
                        continue
                    # ------------------------------------------
                    
                    # Filtra solo file 147bp se richiesto
                    if filter_147bp and '_147.wig' not in member.name:
                        skipped_files.append(member.name)
                        continue
                    
                    # Logica di estrazione (controlla se esiste già)
                    target_path = os.path.join(extract_dir, member.name)
                    if not os.path.exists(target_path):
                        print(f"Estraendo: {member.name}")
                        tar.extract(member, extract_dir)
                    
                    extracted_files.append(target_path) 
            
        # --- CORREZIONE: QUESTE RIGHE SONO ORA INDENTATE ---
        # Si trovano ancora DENTRO il blocco 'try'
        print(f"\n✓ Estratti {len(extracted_files)} file WIG (nucleosomi 147bp)")
        if skipped_files:
            print(f"✓ Saltati {len(skipped_files)} file con altre dimensioni")
        
        return extracted_files
    
    except Exception as e: # Questo 'except' ora corrisponde correttamente al 'try'
        print(f"✗ Errore estrazione TAR: {e}")
        return []

#################################################################################################
# STEP 3: GESTIONE ANNOTAZIONI (PROMOTORI / ENHANCER)
#################################################################################################

def load_genomic_features(gtf_path,  
                          feature_types,  
                          relative_to_tss=False,  
                          window_upstream=0,  
                          window_downstream=0):
    """
    Estrae coordinate per specifici feature genomici (es. esoni, promotori)
    da un file di annotazione GTF.
    """
    print(f"Caricamento annotazioni da {gtf_path} per feature: {feature_types}...")
    features_list = []

    try:
        with open(gtf_path, 'r') as f:
            for line in f:
                if line.startswith('#'):
                    continue
                parts = line.strip().split('\t')
                if len(parts) < 9:
                    continue
                
                chrom, source, feature, start_str, end_str, score, strand, frame, attrs = parts

                if feature in feature_types:
                    try:
                        start = int(start_str)
                        end = int(end_str)
                        
                        reg_start, reg_end = 0, 0

                        if relative_to_tss:
                            # Calcola regione relativa al TSS (per promotori)
                            tss = start if strand == '+' else end
                            reg_start = max(0, tss - window_upstream)
                            reg_end = tss + window_downstream
                        else:
                            # Usa coordinate dirette del feature + finestre
                            if strand == '+':
                                reg_start = max(0, start - window_upstream)
                                reg_end = end + window_downstream
                            else:
                                reg_start = max(0, start - window_downstream)
                                reg_end = end + window_upstream

                        features_list.append((chrom, reg_start, reg_end))
                        
                    except ValueError:
                        continue

    except FileNotFoundError:
        print(f"Errore: File GTF non trovato a {gtf_path}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
    except Exception as e:
        print(f"Errore during GTF read: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

    if not features_list:
        print(f"⚠ Nessun feature di tipo {feature_types} trovato nel file.")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

    df_features = pd.DataFrame(features_list, columns=['chr', 'start', 'end'])
    df_features = df_features.drop_duplicates().reset_index(drop=True)
    
    print(f"✓ Trovati {len(df_features):,} feature unici per '{', '.join(feature_types)}'")
    return df_features

def load_bed_features(bed_path):
    """
    Carica un file BED semplice (chr, start, end).
    """
    print(f"Caricamento features da file BED: {os.path.basename(bed_path)}")
    
    if not os.path.exists(bed_path):
        print(f"✗ File BED non trovato: {bed_path}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
        
    try:
        df = pd.read_csv(
            bed_path,  
            sep='\t',  
            header=None,
            comment='#', # Ignora righe di commento
            usecols=[0, 1, 2],  
            names=['chr', 'start', 'end']
        )
        
        df = df.dropna().drop_duplicates().reset_index(drop=True)
        df['start'] = df['start'].astype(int)
        df['end'] = df['end'].astype(int)
        
        print(f"✓ Trovati {len(df):,} features unici")
        return df
        
    except Exception as e:
        print(f"✗ Errore caricamento file BED: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

def get_feature_dataframe(gtf_path, region_key, promoter_window=1000):
    """
    Funzione "manager" che carica le coordinate per una regione genomica
    specificata da una chiave (key).
    """
    print(f"\n--- Richiesta regione: '{region_key}' ---")
    
    feature_map = {
        "promoter": (["transcript", "gene"], True, promoter_window, promoter_window),
        "exon": (["exon"], False, 0, 0),
        "gene_body": (["gene"], False, 0, 0),
        "utr5": (["five_prime_utr", "5UTR"], False, 0, 0),
        "utr3": (["three_prime_utr", "3UTR"], False, 0, 0),
        "enhancer": (["enhancer"], False, 0, 0), # Per file GTF
    }
    
    if region_key not in feature_map:
        print(f"Errore: Chiave regione '{region_key}' non riconosciuta.")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
        
    params = feature_map[region_key]
    
    return load_genomic_features(gtf_path,  
                                 feature_types=params[0],  
                                 relative_to_tss=params[1],  
                                 window_upstream=params[2],  
                                 window_downstream=params[3])

def download_refgene_hg18(output_dir="../data", promoter_window=1000):
    """
    Scarica le annotazioni RefSeq (hg18) da UCSC e genera regioni promotrici ±promoter_window bp dal TSS.
    """
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg18/database/refGene.txt.gz"
    local_gz = os.path.join(output_dir, "refGene.txt.gz")
    local_csv = os.path.join(output_dir, f"promoters_hg18_{promoter_window}bp.csv")

    if os.path.exists(local_csv):
        print(f"✓ File promotori già presente, caricamento: {local_csv}")
        try:
            return pd.read_csv(local_csv)
        except Exception as e:
            print(f"✗ Errore nel caricamento di {local_csv}: {e}. Tento di rigenerarlo.")

    if not os.path.exists(local_gz):
        print(f"Scaricando RefSeq hg18 da UCSC...")
        try:
            response = requests.get(url, stream=True, timeout=180)
            response.raise_for_status()
            with open(local_gz, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"✓ File salvato: {local_gz}")
        except Exception as e:
            print(f"✗ Errore nel download del file RefSeq: {e}")
            return None
    else:
        print(f"✓ File RefSeq già presente: {local_gz}")

    cols = [
        'bin', 'name', 'chrom', 'strand', 'txStart', 'txEnd',
        'cdsStart', 'cdsEnd', 'exonCount', 'exonStarts', 'exonEnds',
        'score', 'name2', 'cdsStartStat', 'cdsEndStat', 'exonFrames'
    ]
    
    print("Caricamento e parsing refGene...")
    try:
        refgene = pd.read_csv(local_gz, sep='\t', header=None, names=cols, compression='gzip')
    except Exception as e:
        print(f"✗ Errore parsing refGene: {e}")
        return None
        
    print(f"✓ {len(refgene):,} geni caricati")

    print(f"Calcolo regioni promotrici ±{promoter_window} bp dal TSS...")
    promoter_records = []

    for _, row in refgene.iterrows():
        chrom = row['chrom']
        if not isinstance(chrom, str) or not chrom.startswith("chr"):
            continue
        
        try:
            txStart = int(row['txStart'])
            txEnd = int(row['txEnd'])
            
            if row['strand'] == '+':
                tss = txStart
            else:
                tss = txEnd
            
            start = max(0, tss - promoter_window)
            end = tss + promoter_window

            promoter_records.append({
                'chr': chrom,
                'start': int(start),
                'end': int(end),
                'strand': row['strand'],
                'gene': row['name2']
            })
        except ValueError:
            continue

    df_promoters = pd.DataFrame(promoter_records)
    df_promoters = df_promoters.drop_duplicates(subset=['chr', 'start', 'end'])
    print(f"✓ {len(df_promoters):,} regioni promotrici uniche generate")

    try:
        df_promoters.to_csv(local_csv, index=False)
        print(f"✓ Salvati promotori in: {local_csv}")
    except Exception as e:
        print(f"✗ Errore salvataggio CSV promotori: {e}")
    
    return df_promoters

#################################################################################################
# STEP 4: CREAZIONE DATASET (Logica Positivo vs Negativo)
#################################################################################################

def filter_regions_by_feature(df_regions, df_features):
    """
    Mantiene solo le regioni (es. nucleosomi) che si sovrappongono  
    con i feature forniti (es. promotori, esoni).
    Restituisce un nuovo DataFrame filtrato.
    """
    print("  Filtraggio regioni per sovrapposizione con feature...")

    if df_features is None or df_features.empty:
        print("  ⚠ Nessuna annotazione feature fornita, ritorno tutte le regioni (nessun filtro).")
        return df_regions

    # Crea dizionario cromosoma → array di (start, end) per efficienza
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("  Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return pd.DataFrame(columns=df_regions.columns)
        
    keep = np.zeros(len(df_regions), dtype=bool)

    for i, row in enumerate(df_regions.itertuples(index=False)):
        chrom = getattr(row, 'chr', None)
        if chrom not in df_features_by_chr:
            continue
            
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        
        if s is None or e is None:
            continue

        # Cerca sovrapposizione
        for f_start, f_end in df_features_by_chr[chrom]:
            # (e < f_start) -> la regione è completamente prima del feature
            # (s > f_end)   -> la regione è completamente dopo il feature
            if not (e < f_start or s > f_end):  
                keep[i] = True
                break # Trovata sovrapposizione, passa alla prossima regione

    filtered_df = df_regions[keep]
    print(f"  ✓ {len(filtered_df):,}/{len(df_regions):,} regioni trovate nei feature forniti.")
    return filtered_df.reset_index(drop=True)

# --- MODIFICA 3 (NUOVA): Funzione POSITIVI con Peak Calling ---
def identify_nucleosome_positions(df_wig, threshold_value=1.0, sequence_length=201):
    """
    Identifica le posizioni delle diadi dei nucleosomi (picchi)
    usando la logica 'Non-Maximum Suppression'.
    """
    print("  Identificazione posizioni nucleosomi (Peak Calling)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    # 1. Trova tutti i possibili candidati positivi
    df_positive_candidates = df_wig[df_wig['score'] > threshold_value].copy()
    print(f"    Trovati {len(df_positive_candidates):,} candidati positivi (> {threshold_value:.2f})")

    if df_positive_candidates.empty:
        print("    ⚠ Nessun candidato positivo trovato.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    # 2. Ordina i candidati dal più forte (score più alto) al più debole
    df_sorted = df_positive_candidates.sort_values('score', ascending=False)
    
    true_dyad_rows = []
    claimed_tree = {} # Un albero di intervalli per le regioni già "bloccate"
    
    # La nostra separazione minima è la lunghezza della sequenza
    min_separation = sequence_length 

    print(f"    Avvio Non-Maximum Suppression (min_sep = {min_separation} bp)...")
    
    # Inizializza gli alberi per i cromosomi
    for chrom in df_sorted['chr'].unique():
        claimed_tree[chrom] = IntervalTree()

    # 3. Itera dal picco più forte in giù
    for row in df_sorted.itertuples(index=False):
        
        # 4. Controlla se questo picco è già stato "bloccato" da un picco più forte
        if claimed_tree[row.chr].overlaps(row.start, row.end):
            continue # Sì, ignora questo picco
            
        # 5. Se non è bloccato, è un VERO picco!
        true_dyad_rows.append(row)
        
        # 6. Blocca la regione attorno ad esso
        buffer_start = row.start - (min_separation // 2)
        buffer_end = row.start + (min_separation // 2) + 1 # +1 perché gli intervalli sono half-open
        claimed_tree[row.chr].addi(buffer_start, buffer_end)

    if not true_dyad_rows:
        print("    ⚠ Nessun picco positivo trovato dopo il peak calling.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    print(f"    ✓ Identificati {len(true_dyad_rows):,} picchi unici (diadi)")

    # 7. Crea il DataFrame finale dei positivi
    df_true_dyads = pd.DataFrame(true_dyad_rows, columns=df_sorted.columns)

    # 8. Espandi i centri (diadi) alla lunghezza finale della sequenza (201 bp)
    df_true_dyads['center'] = (df_true_dyads['start'] + df_true_dyads['end']) // 2
    df_true_dyads['start'] = df_true_dyads['center'] - (sequence_length // 2)
    df_true_dyads['end'] = df_true_dyads['start'] + sequence_length
    
    return df_true_dyads[['chr', 'start', 'end', 'score']]

# --- MODIFICA 4: Funzione NEGATIVI (score == 1.0) ---
def identify_non_nucleosome_positions(df_wig, threshold_value=1.0, nucleosome_length=201):
    """
    Identifica posizioni NON-nucleosomi (BASSA occupazione)
    usando una SOGLIA ASSOLUTA (score == threshold)
    """
    print("  Identificazione posizioni NON-nucleosomi (BASSA occupazione)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    # Usa il valore assoluto passato
    print(f"    Soglia segnale (Negativi): == {threshold_value:.2f}")
    
    # Filtra regioni con segnale BASSO (esattamente il background)
    df_nucleosomes = df_wig[df_wig['score'] == threshold_value].copy()
    
    # Estendi regioni alla lunghezza desiderata
    df_nucleosomes['center'] = (df_nucleosomes['start'] + df_nucleosomes['end']) // 2
    df_nucleosomes['start'] = df_nucleosomes['center'] - nucleosome_length // 2
    df_nucleosomes['end'] = df_nucleosomes['start'] + nucleosome_length
    
    print(f"    ✓ Identificati {len(df_nucleosomes)} NON-nucleosomi (Negativi)")
    
    return df_nucleosomes[['chr', 'start', 'end', 'score']]

def extract_sequences_from_genome(df_regions, genome_dict, max_sequences=2000):
    """
    Estrae sequenze DNA dalle regioni specificate
    """
    print(f"    Estrazione sequenze da {len(df_regions)} regioni...")
    
    sequences = []
    
    count = 0
    processed = 0
    for idx, row in df_regions.iterrows():
        processed += 1
        if count >= max_sequences:
            print(f"      Raggiunto limite di {max_sequences} sequenze.")
            break
        
        chrom = row['chr']
        start = int(row['start'])
        end = int(row['end'])
        
        if chrom not in genome_dict:
            continue
        
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        sequences.append(seq)
        count += 1
        
        if count % 500 == 0 and count > 0:
            print(f"      Estraendole... {count}/{max_sequences} sequenze valide trovate (processate {processed} regioni)")
    
    print(f"    ✓ Estratte {len(sequences)} sequenze")
    return sequences

# --- MODIFICA 5: NUOVA Funzione di Estrazione Efficiente (con Buffer Zone) ---
def extract_filtered_sequences_efficiently(df_regions, df_features, genome_dict, positive_regions_tree, max_sequences=2000, buffer_zone=201):
    """
    Combina filtro e estrazione in un unico passaggio efficiente.
    Mescola i candidati e si ferma non appena 'max_sequences' sono state trovate.
    ESCLUDE i negativi che si sovrappongono ai positivi (buffer_zone).
    """
    print(f"  Avvio estrazione efficiente per {max_sequences} sequenze da {len(df_regions):,} candidati...")
    
    sequences = []
    
    # 1. Crea il dizionario di features per una ricerca rapida
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("    Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return []

    # 2. Mescola i candidati (df_regions)
    #    Usiamo sample(frac=1) che è un modo efficiente in pandas per mescolare
    df_regions_shuffled = df_regions.sample(frac=1)

    processed_count = 0
    excluded_by_buffer = 0
    
    # 3. Itera, filtra, ed estrai in un unico loop
    for row in df_regions_shuffled.itertuples(index=False):
        processed_count += 1
        
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        chrom = getattr(row, 'chr', None)
        
        if s is None or e is None or chrom is None:
            continue
            
        # --- NUOVO CONTROLLO: ZONA CUSCINETTO ---
        # Controlla se questa regione negativa (s, e) è troppo vicina a un positivo.
        # Creiamo un intervallo buffer attorno ad essa.
        buffer_start = s - buffer_zone
        buffer_end = e + buffer_zone
        if chrom in positive_regions_tree and positive_regions_tree[chrom].overlaps(buffer_start, buffer_end):
            excluded_by_buffer += 1
            continue # Questo negativo è troppo vicino a un positivo, scartalo.
        # ----------------------------------------

        # Controllo 2: Logica di 'filter_regions_by_feature' (promotori)
        if chrom not in df_features_by_chr:
            continue
            
        has_overlap = False
        for f_start, f_end in df_features_by_chr[chrom]:
            if not (e < f_start or s > f_end):
                has_overlap = True
                break
        
        if not has_overlap:
            continue # Non è in un promotore, passa al prossimo candidato

        # --- Se siamo qui, la regione E' in un promotore E NON è vicina a un positivo ---
        
        # Logica di 'extract_sequences_from_genome'
        start = int(s)
        end = int(e)
        
        if chrom not in genome_dict:
            continue
            
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        # --- Se siamo qui, la sequenza è VALIDA ---
        sequences.append(seq)
        
        # Logica di 'stop'
        if len(sequences) >= max_sequences:
            print(f"    ✓ Raggiunto limite di {max_sequences} sequenze valide (processati {processed_count:,} candidati).")
            print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")
            break
            
    if len(sequences) < max_sequences:
        print(f"    ✓ Estrazione completata. Trovate {len(sequences)} sequenze valide (processati tutti i {len(df_regions):,} candidati).")
        print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")

    return sequences


# --- MODIFICA 6: Funzione DATASET (usa la funzione efficiente e l'albero) ---
def create_nucleosome_dataset(wig_files, genome_dict,  
                            max_sequences_per_file=1000,
                            threshold_high=1.0,
                            threshold_low=1.0,
                            max_regions_per_wig=100000,
                            df_features=None,
                            sequence_length=201):
    """
    Crea dataset finale combinando dati WIG e genoma.
    Positivi: score > threshold_high
    Negativi: score == threshold_low
    """
    print("\n" + "="*60)
    print("CREAZIONE DATASET NUCLEOSOMI (Positivi vs Negativi in-loco)")
    print("="*60)
    print(f"Lunghezza sequenza: {sequence_length} bp")
    print(f"Soglia ALTA (positivi): > {threshold_high:.2f}")
    print(f"Soglia BASSA (negativi): == {threshold_low:.2f}")
    print(f"Filtro regioni: {df_features is not None}")
    
    all_sequences = []
    all_labels = []
    
    # Processa TUTTI i file WIG
    for idx, wig_file in enumerate(wig_files, 1):
        print(f"\n[{idx}/{len(wig_files)}] Processando: {os.path.basename(wig_file)}")
        
        # Passa max_regions_per_wig al parser
        df_wig = parse_wig_file(wig_file, max_regions=max_regions_per_wig)
        
        if df_wig is None or len(df_wig) == 0:
            print("  ⚠ Nessun dato trovato, skip")
            continue
            
        # -----------------------------------------------------------------
        # 1. STREAM POSITIVO (Label 1)
        # -----------------------------------------------------------------
        print("\n  --- Stream Positivo (Label 1) ---")
        # 1a. Trova TUTTI i positivi (>1.0) nell'INTERO GENOMA (con Peak Calling)
        df_nucleosomes_high = identify_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_high, # Passa threshold_high (1.0)
            sequence_length=sequence_length
        )
        
        # 1b. Filtra i positivi SOLO per i promotori (per il set di training)
        df_positive_filtered = filter_regions_by_feature(df_nucleosomes_high, df_features)

        if df_positive_filtered.empty:
            print("  ⚠ Nessuna regione positiva (>1.0) nei feature forniti, skip file")
            continue
            
        # 1c. Estrai sequenze
        sequences_pos = extract_sequences_from_genome(
            df_positive_filtered,  
            genome_dict,  
            max_sequences=max_sequences_per_file 
        )
        
        # --- NUOVO PASSAGGIO: Crea l'albero dei positivi per la zona cuscinetto ---
        # Usa la lista TOTALE (df_nucleosomes_high) per l'albero
        print("    Creazione albero intervalli positivi per zona cuscinetto...")
        positive_tree = {}
        for chrom in df_nucleosomes_high['chr'].unique():
            positive_tree[chrom] = IntervalTree()
            for row in df_nucleosomes_high[df_nucleosomes_high['chr'] == chrom].itertuples():
                positive_tree[chrom].addi(row.start, row.end)
        # ---------------------------------------------------------------------

        # -----------------------------------------------------------------
        # 2. STREAM NEGATIVO (Label 0) - Logica efficiente
        # -----------------------------------------------------------------
        print("\n  --- Stream Negativo (Label 0) ---")
        
        n_to_find = len(sequences_pos) # Numero di negativi da trovare
        if n_to_find == 0:
             print("  ⚠ Nessuna sequenza positiva trovata, skip negativi")
             continue
        
        # 2a. Trova TUTTI i candidati negativi (==1.0)
        df_nucleosomes_low = identify_non_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_low, # Passa threshold_low (1.0)
            nucleosome_length=sequence_length
        )
        
        if df_nucleosomes_low.empty:
            print("  ⚠ Nessuna regione negativa (==1.0) trovata, skip negativi")
            continue
        
        # 2b. --- CHIAMATA ALLA NUOVA FUNZIONE EFFICIENTE (CON ALBERO) ---
        sequences_neg = extract_filtered_sequences_efficiently(
            df_nucleosomes_low,     # I milioni di candidati
            df_features,            # I promotori
            genome_dict,
            positive_tree,          # L'albero dei positivi da evitare
            max_sequences=n_to_find, # Il numero da trovare
            buffer_zone=sequence_length # Buffer = 201 bp
        )
        
        if not sequences_neg:
             print("  ⚠ Nessuna sequenza negativa valida trovata (o tutte troppo vicine ai positivi).")
             continue
        
        # -----------------------------------------------------------------
        # 3. BILANCIAMENTO e AGGIUNTA AL DATASET
        # -----------------------------------------------------------------
        
        n_pos = len(sequences_pos)
        n_neg = len(sequences_neg)

        n_to_keep = min(n_pos, n_neg)
        
        if n_pos > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_pos} positivi a {n_to_keep}")
            sequences_pos = random.sample(sequences_pos, n_to_keep)
        
        if n_neg > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_neg} negativi a {n_to_keep}")
            sequences_neg = random.sample(sequences_neg, n_to_keep)

        # -----------------------------------------------------------------
        
        all_sequences.extend(sequences_pos)
        all_labels.extend([1] * len(sequences_pos))
        
        all_sequences.extend(sequences_neg)
        all_labels.extend([0] * len(sequences_neg))
        
        print(f"\n  RISULTATO FILE: Aggiunti {len(sequences_pos)} positivi e {len(sequences_neg)} negativi (Bilanciati).")
        print(f"  ✓ Totale sequenze accumulate: {len(all_sequences):,}")

    
    if not all_sequences:
        print("⚠⚠ NESSUNA SEQUENZA TROVATA. Controlla i percorsi e i filtri.")
        return pd.DataFrame(columns=['sequenza', 'nucleosoma'])

    # Crea DataFrame
    df = pd.DataFrame({
        'sequenza': all_sequences,
        'nucleosoma': all_labels
    })
    
    df = df.drop_duplicates(subset=['sequenza']).reset_index(drop=True)
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    
    print("\n" + "="*60)
    print("DATASET COMPLETATO (Bilanciato per regione)")
    print("="*60)
    print(f"Totale sequenze: {len(df):,}")
    print(f"Nucleosomi (positivi): {df['nucleosoma'].sum():,}")
    print(f"Non-nucleosomi (negativi): {(1 - df['nucleosoma']).sum():,}")
    
    if len(df) > 0:
        print(f"Bilanciamento: {df['nucleosoma'].sum() / len(df) * 100:.1f}% positivi")
    
    return df

#################################################################################################
# ESECUZIONE PRINCIPALE
#################################################################################################

def main():
    print("="*60)
    print("ELABORAZIONE DATI NUCLEOSOMI GSE36979")
    print("="*60)

    # Imposta directory dati
    base_dir = os.getcwd()
    data_dir = os.path.abspath(os.path.join(base_dir, "..", "data"))
    
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        print(f"Creata directory dati: {data_dir}")

    # STEP 1: Estrai file WIG dal TAR
    tar_path = os.path.join(data_dir, "GSE36979_RAW.tar")
    wig_extract_dir = os.path.join(data_dir, "wig_files")

    if not os.path.exists(tar_path):
        print(f"✗ File non trovato: {tar_path}")
        print("Scaricare GSE36979_RAW.tar da GEO e posizionarlo in ../data/")
        return

    wig_files = extract_tar_wig_files(tar_path, extract_dir=wig_extract_dir)

    if not wig_files:
        print("✗ Nessun file WIG 147bp trovato nell'archivio. Controlla il file TAR.")
        return

    # STEP 2: Scarica/Carica genoma hg18
    print("\n" + "="*60)
    genome_dir = os.path.join(data_dir, "hg18_genome")

    if os.path.exists(genome_dir) and len(os.listdir(genome_dir)) > 0:
        print("Genoma hg18 già presente, caricamento...")
        genome_dict = load_hg18_genome(genome_dir)
    else:
        print("Scaricamento genoma hg18...")
        print("ATTENZIONE: Il download completo richiede ~1.5GB di spazio e tempo")
        genome_dict = download_hg18_genome(genome_dir)

    if not genome_dict:
        print("✗ Impossibile caricare genoma")
        return

    # STEP 2.5: Carica o scarica annotazioni RefGene per i promotori
    print("\n" + "="*60)
    print("Caricamento/Generazione annotazioni promotori (RefGene hg18)...")
    
    try:
        df_promoters = download_refgene_hg18(output_dir=data_dir, promoter_window=1000)
    except Exception as e:
        print(f"✗ Errore fatale durante il caricamento dei promotori: {e}")
        df_promoters = None

    if df_promoters is None or df_promoters.empty:
        print("✗ Impossibile caricare le annotazioni dei promotori. Uscita.")
        return
    else:
        print(f"✓ Annotazioni promotori caricate: {len(df_promoters)} regioni")
        
    # --- ALTERNATIVA: USA ENHANCER (esempio se avessi un file BED) ---
    # enhancer_bed_path = os.path.join(data_dir, "hg18_enhancers.bed")
    # df_features_to_use = load_bed_features(enhancer_bed_path)
    # ------------------------------------------------------------------
    df_features_to_use = df_promoters # Scegliamo di usare i promotori


    # --- MODIFICA 7: Chiamata principale con i parametri finali ---
    # STEP 3: Crea dataset
    df_final = create_nucleosome_dataset(
        wig_files,  
        genome_dict,
        max_sequences_per_file=100000,
        threshold_high=2.0,           # Positivi > 1.0
        threshold_low=1.0,            # Negativi == 1.0
        max_regions_per_wig=5000000,  # Leggi fino a 5 milioni di regioni per file
        df_features=df_features_to_use,
        sequence_length=201
    )

    if df_final is None or df_final.empty:
        print("✗ Creazione dataset fallita o dataset vuoto.")
        return

    # STEP 4: Salva risultati
    output_file = os.path.join(data_dir, "nucleosome_dataset_GSE36979_promoters_201pb_balanced.csv")
    try:
        df_final.to_csv(output_file, index=False)
        print(f"\n✓ Dataset salvato: {output_file}")
    except Exception as e:
        print(f"✗ Errore salvataggio dataset: {e}")
        return

    # STEP 5: Statistiche finali
    print("\n" + "="*60)
    print("STATISTICHE FINALI")
    print("="*60)
    print(df_final.head(10))
    
    print(f"\nDistribuzione lunghezze:")
    print(df_final['sequenza'].str.len().describe())

    # GC content
    df_final['gc_content'] = df_final['sequenza'].apply(calc_gc)
    print(f"\nGC content per classe:")
    print(df_final.groupby('nucleosoma')['gc_content'].describe())

    # Plot
    print("\nGenerazione plot GC content...")
    try:
        plt.figure(figsize=(10, 6))
        sns.kdeplot(data=df_final, x='gc_content', hue='nucleosoma', fill=True, common_norm=False, palette="Set1")
        plt.title('Distribuzione GC Content tra Classi (0=Bassa, 1=Alta Occupazione)')
        plot_file = os.path.join(data_dir, "gc_content_distribution.png")
        plt.savefig(plot_file)
        print(f"✓ Plot salvato: {plot_file}")
        # plt.show() # Decommenta se sei in un notebook
    except Exception as e:
        print(f"✗ Errore generazione plot: {e}")

    print("\n✓ ELABORAZIONE COMPLETATA!")

if __name__ == "__main__":  
    main()
```

```python
import pandas as pd
import numpy as np
import requests
import gzip
import os
from Bio import SeqIO
from io import StringIO
import tarfile
import random
import matplotlib.pyplot as plt
import seaborn as sns
from intervaltree import IntervalTree # Importazione necessaria

#################################################################################################
# FUNZIONI HELPER
#################################################################################################

def calc_gc(seq):
    """Calcola il contenuto GC di una sequenza."""
    if not isinstance(seq, str) or len(seq) == 0:
        return 0
    gc = seq.count('G') + seq.count('C')
    return (gc / len(seq)) * 100

#################################################################################################
# STEP 1: GESTIONE GENOMA HG18
#################################################################################################

def download_hg18_genome(output_dir="hg18_genome"):
    """
    Scarica il genoma umano hg18 da UCSC (SOLO CHR20)
    """
    print("Scaricando genoma hg18 da UCSC...")
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # URL base UCSC per hg18
    base_url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg18/chromosomes/"
    
    ### MODIFICA: Filtra solo per chr20 ###
    chromosomes = ["chr20"]
    
    genome_dict = {}
    
    for chrom in chromosomes:
        file_name = f"{chrom}.fa.gz"
        file_path = os.path.join(output_dir, file_name)
        
        # Percorso del file decompresso
        unzipped_file_path = file_path.replace('.gz', '')

        if os.path.exists(unzipped_file_path):
            print(f"✓ {chrom} già scaricato")
            # Carica sequenza già presente
            try:
                with open(unzipped_file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        genome_dict[chrom] = str(record.seq).upper() # Salva in maiuscolo
                continue
            except Exception as e:
                print(f"✗ Errore nel caricamento di {unzipped_file_path}: {e}. Tento di riscaricarlo.")

        
        url = base_url + file_name
        print(f"Scaricando {chrom}...")
        
        try:
            response = requests.get(url, stream=True, timeout=300)
            if response.status_code == 200:
                # Salva file compresso
                with open(file_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                
                # Decomprimi e leggi
                sequence_content = ""
                with gzip.open(file_path, 'rt') as f:
                    # Leggi tutto il contenuto
                    sequence_content = f.read()
                
                # Parse dalla stringa
                with StringIO(sequence_content) as f_in:
                    records = SeqIO.parse(f_in, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Salva in maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} scaricato: {len(seq_str):,} bp")
                
                # Salva versione decompressa
                with open(unzipped_file_path, 'w') as f_out:
                    f_out.write(sequence_content)
                
                # Rimuovi file compresso per salvare spazio
                os.remove(file_path)
                
        except Exception as e:
            print(f"✗ Errore scaricamento {chrom}: {e}")
            if os.path.exists(file_path):
                os.remove(file_path) # Pulisci file corrotto
            continue
    
    print(f"\n✓ Genoma hg18 scaricato: {len(genome_dict)} cromosomi")
    return genome_dict

def load_hg18_genome(genome_dir="hg18_genome"):
    """
    Carica genoma hg18 già scaricato (SOLO CHR20)
    """
    print("Caricamento genoma hg18...")
    
    genome_dict = {}
    
    ### MODIFICA: Filtra solo per chr20 ###
    chromosomes = ["chr20"]
    
    for chrom in chromosomes:
        file_path = os.path.join(genome_dir, f"{chrom}.fa")
        
        if os.path.exists(file_path):
            try:
                with open(file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Assicurati sia maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} caricato: {len(seq_str):,} bp")
            except Exception as e:
                print(f"✗ Errore caricamento {file_path}: {e}")
        else:
            print(f"✗ {chrom} non trovato in {genome_dir}")
    
    if not genome_dict:
        print(f"✗ Nessun file genoma trovato in {genome_dir}. Prova a scaricarlo prima.")
    else:
        print(f"\n✓ Genoma caricato: {len(genome_dict)} cromosomi")
    return genome_dict

#################################################################################################
# STEP 2: GESTIONE DATI WIG (GSE36979)
#################################################################################################

def parse_wig_file(wig_file_path, max_regions=100000):
    """
    Parsifica file WIG con dati di occupancy dei nucleosomi (SOLO CHR20)
    """
    print(f"  Parsing file WIG: {wig_file_path}")
    
    regions = []
    current_chrom = None
    current_start = None
    current_step = None
    current_span = 1
    position = 0
    lines_processed = 0
    
    try: # Inizio del blocco 'try'
        # Gestisci file compressi o no
        opener = gzip.open if wig_file_path.endswith('.gz') else open
        
        with opener(wig_file_path, 'rt') as f:
            for line in f:
                line = line.strip()
                
                if not line or line.startswith('track') or line.startswith('browser'):
                    continue
                
                lines_processed += 1
                
                # --- MODIFICA 1: Riduzione print ---
                if lines_processed % 50000000 == 0:
                    print(f"    ... Processate {lines_processed:,} linee, trovate {len(regions):,} regioni...")
                
                # Stop se abbiamo raggiunto il limite (per debug)
                if len(regions) >= max_regions:
                    print(f"    ⚠ Raggiunto limite di {max_regions:,} regioni, stop parsing")
                    break
                
                # fixedStep format
                if line.startswith('fixedStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('start='):
                            current_start = int(part.split('=')[1])
                            position = current_start
                        elif part.startswith('step='):
                            current_step = int(part.split('=')[1])
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                
                # variableStep format
                elif line.startswith('variableStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                    current_step = None
                
                # Data line
                else:
                    ### MODIFICA: Filtra solo per chr20 ###
                    if current_chrom != "chr20":
                        continue
                    ### FINE MODIFICA ###
                        
                    try:
                        if current_step:
                            # fixedStep: solo valore
                            value = float(line)
                            if value > 0: # Solo regioni con segnale
                                regions.append({
                                    'chr': current_chrom,
                                    'start': position,
                                    'end': position + current_span,
                                    'score': value
                                })
                            position += current_step
                        else:
                            # variableStep: posizione e valore
                            pos, value = line.split()
                            pos = int(pos)
                            value = float(value)
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': pos,
                                    'end': pos + current_span,
                                    'score': value
                                })
                    except Exception:
                        # Salta righe dati malformate
                        continue
            
        # --- CORREZIONE: QUESTE RIGHE SONO ORA INDENTATE ---
        # Si trovano ancora DENTRO il blocco 'try'
        print(f"    ✓ Trovate {len(regions):,} regioni con segnale (solo chr20)")
        return pd.DataFrame(regions)
    
    except Exception as e: # Questo 'except' ora corrisponde correttamente al 'try'
        print(f"    ✗ Errore parsing WIG: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

import os
import tarfile

def extract_tar_wig_files(tar_path, extract_dir="wig_files", filter_147bp=True):
    """
    Estrae file WIG (.wig o .wig.gz) da un archivio TAR.

    Se filter_147bp=True, estrae solo i file che terminano 
    con '_147.wig' o '_147.wig.gz' (nucleosomi canonici).
    """
    print(f"Inizio estrazione da: {tar_path}...")

    if not os.path.exists(extract_dir):
        print(f"Creo directory: {extract_dir}")
        os.makedirs(extract_dir)

    extracted_files = []
    skipped_files = []

    try:
        with tarfile.open(tar_path, 'r') as tar:
            members = tar.getmembers()
            print(f"Trovati {len(members)} elementi nell'archivio.")

            if filter_147bp:
                print("Filtro 147bp (nucleosomi canonici) ATTIVO.")

            for member in members:
                # Salta le directory e altri elementi non-file
                if not member.isfile():
                    continue

                member_name = member.name
                
                # Controlla se è un file WIG (compresso o meno)
                if member_name.endswith('.wig') or member_name.endswith('.wig.gz'):
                    
                    # --- CORREZIONE: Logica di filtro migliorata ---
                    is_147bp_file = False
                    # Gestisce entrambi i casi: _147.wig e _147.wig.gz
                    if member_name.endswith('_147.wig') or member_name.endswith('_147.wig.gz'):
                        is_147bp_file = True
                    # --- FINE CORREZIONE ---

                    # --- Esclusione file anomalo ---
                    if "GSM907789" in member.name:
                        print(f"File anomalo saltato: {member.name}")
                        continue
                    # ------------------------------------------
                    

                    # Applica il filtro se richiesto
                    if filter_147bp and not is_147bp_file:
                        skipped_files.append(member_name)
                        continue
                    
                    # --- CODICE DI TEST/ESCLUSIONI RIMOSSO ---
                    # (Rimossi i blocchi per 'GSM907790' e 'GSM907789')
                    # ----------------------------------------


                    

                    # Logica di estrazione (Mantenuta come l'originale)
                    # NOTA: questo ricrea la struttura delle directory del TAR dentro 'extract_dir'
                    target_path = os.path.join(extract_dir, member_name)

                    if not os.path.exists(target_path):
                        print(f"Estraendo: {member_name}")
                        tar.extract(member, extract_dir)
                    # else:
                    #    print(f"File già esistente, salto estrazione: {member_name}")

                    extracted_files.append(target_path)

        print("-" * 30)
        print(f"✓ Estrazione completata.")
        print(f"✓ File estratti/trovati: {len(extracted_files)}")
        if skipped_files:
            print(f"✓ File saltati (filtro 147bp): {len(skipped_files)}")
            
        return extracted_files

    except tarfile.ReadError as e:
        print(f"✗ Errore lettura file TAR (potrebbe essere corrotto): {e}")
        return []
    except Exception as e:
        print(f"✗ Errore generico durante l'estrazione: {e}")
        return []

#################################################################################################
# STEP 3: GESTIONE ANNOTAZIONI (PROMOTORI / ENHANCER)
#################################################################################################

def load_genomic_features(gtf_path,  
                          feature_types,  
                          relative_to_tss=False,  
                          window_upstream=0,  
                          window_downstream=0):
    """
    Estrae coordinate per specifici feature genomici (es. esoni, promotori)
    da un file di annotazione GTF.
    """
    print(f"Caricamento annotazioni da {gtf_path} per feature: {feature_types}...")
    features_list = []

    try:
        with open(gtf_path, 'r') as f:
            for line in f:
                if line.startswith('#'):
                    continue
                parts = line.strip().split('\t')
                if len(parts) < 9:
                    continue
                
                chrom, source, feature, start_str, end_str, score, strand, frame, attrs = parts

                if feature in feature_types:
                    try:
                        start = int(start_str)
                        end = int(end_str)
                        
                        reg_start, reg_end = 0, 0

                        if relative_to_tss:
                            # Calcola regione relativa al TSS (per promotori)
                            tss = start if strand == '+' else end
                            reg_start = max(0, tss - window_upstream)
                            reg_end = tss + window_downstream
                        else:
                            # Usa coordinate dirette del feature + finestre
                            if strand == '+':
                                reg_start = max(0, start - window_upstream)
                                reg_end = end + window_downstream
                            else:
                                reg_start = max(0, start - window_downstream)
                                reg_end = end + window_upstream

                        features_list.append((chrom, reg_start, reg_end))
                        
                    except ValueError:
                        continue

    except FileNotFoundError:
        print(f"Errore: File GTF non trovato a {gtf_path}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
    except Exception as e:
        print(f"Errore during GTF read: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

    if not features_list:
        print(f"⚠ Nessun feature di tipo {feature_types} trovato nel file.")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

    df_features = pd.DataFrame(features_list, columns=['chr', 'start', 'end'])
    df_features = df_features.drop_duplicates().reset_index(drop=True)
    
    print(f"✓ Trovati {len(df_features):,} feature unici per '{', '.join(feature_types)}'")
    return df_features

def load_bed_features(bed_path):
    """
    Carica un file BED semplice (chr, start, end).
    """
    print(f"Caricamento features da file BED: {os.path.basename(bed_path)}")
    
    if not os.path.exists(bed_path):
        print(f"✗ File BED non trovato: {bed_path}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
        
    try:
        df = pd.read_csv(
            bed_path,  
            sep='\t',  
            header=None,
            comment='#', # Ignora righe di commento
            usecols=[0, 1, 2],  
            names=['chr', 'start', 'end']
        )
        
        df = df.dropna().drop_duplicates().reset_index(drop=True)
        df['start'] = df['start'].astype(int)
        df['end'] = df['end'].astype(int)
        
        print(f"✓ Trovati {len(df):,} features unici")
        return df
        
    except Exception as e:
        print(f"✗ Errore caricamento file BED: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end'])

def get_feature_dataframe(gtf_path, region_key, promoter_window=1000):
    """
    Funzione "manager" che carica le coordinate per una regione genomica
    specificata da una chiave (key).
    """
    print(f"\n--- Richiesta regione: '{region_key}' ---")
    
    feature_map = {
        "promoter": (["transcript", "gene"], True, promoter_window, promoter_window),
        "exon": (["exon"], False, 0, 0),
        "gene_body": (["gene"], False, 0, 0),
        "utr5": (["five_prime_utr", "5UTR"], False, 0, 0),
        "utr3": (["three_prime_utr", "3UTR"], False, 0, 0),
        "enhancer": (["enhancer"], False, 0, 0), # Per file GTF
    }
    
    if region_key not in feature_map:
        print(f"Errore: Chiave regione '{region_key}' non riconosciuta.")
        return pd.DataFrame(columns=['chr', 'start', 'end'])
        
    params = feature_map[region_key]
    
    return load_genomic_features(gtf_path,  
                                 feature_types=params[0],  
                                 relative_to_tss=params[1],  
                                 window_upstream=params[2],  
                                 window_downstream=params[3])

def download_refgene_hg18(output_dir="../data", promoter_window=1000):
    """
    Scarica le annotazioni RefSeq (hg18) da UCSC e genera regioni promotrici ±promoter_window bp dal TSS.
    (FILTRATO SOLO PER CHR20)
    """
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg18/database/refGene.txt.gz"
    local_gz = os.path.join(output_dir, "refGene.txt.gz")
    
    ### MODIFICA: Cambia nome file CSV per riflettere chr20 ###
    local_csv = os.path.join(output_dir, f"promoters_hg18_{promoter_window}bp_chr20.csv")

    if os.path.exists(local_csv):
        print(f"✓ File promotori (chr20) già presente, caricamento: {local_csv}")
        try:
            return pd.read_csv(local_csv)
        except Exception as e:
            print(f"✗ Errore nel caricamento di {local_csv}: {e}. Tento di rigenerarlo.")

    if not os.path.exists(local_gz):
        print(f"Scaricando RefSeq hg18 da UCSC...")
        try:
            response = requests.get(url, stream=True, timeout=180)
            response.raise_for_status()
            with open(local_gz, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"✓ File salvato: {local_gz}")
        except Exception as e:
            print(f"✗ Errore nel download del file RefSeq: {e}")
            return None
    else:
        print(f"✓ File RefSeq già presente: {local_gz}")

    cols = [
        'bin', 'name', 'chrom', 'strand', 'txStart', 'txEnd',
        'cdsStart', 'cdsEnd', 'exonCount', 'exonStarts', 'exonEnds',
        'score', 'name2', 'cdsStartStat', 'cdsEndStat', 'exonFrames'
    ]
    
    print("Caricamento e parsing refGene...")
    try:
        refgene = pd.read_csv(local_gz, sep='\t', header=None, names=cols, compression='gzip')
    except Exception as e:
        print(f"✗ Errore parsing refGene: {e}")
        return None
        
    print(f"✓ {len(refgene):,} geni totali caricati.")

    ### MODIFICA: Filtra solo per chr20 ###
    refgene = refgene[refgene['chrom'] == "chr20"].copy()
    print(f"✓ {len(refgene):,} geni trovati su chr20.")
    ### FINE MODIFICA ###

    print(f"Calcolo regioni promotrici ±{promoter_window} bp dal TSS...")
    promoter_records = []

    for _, row in refgene.iterrows():
        chrom = row['chrom']
        if not isinstance(chrom, str) or not chrom.startswith("chr"):
            continue
        
        # (Controllo ridondante ma sicuro)
        if chrom != "chr20":
            continue
            
        try:
            txStart = int(row['txStart'])
            txEnd = int(row['txEnd'])
            
            if row['strand'] == '+':
                tss = txStart
            else:
                tss = txEnd
            
            start = max(0, tss - promoter_window)
            end = tss + promoter_window

            promoter_records.append({
                'chr': chrom,
                'start': int(start),
                'end': int(end),
                'strand': row['strand'],
                'gene': row['name2']
            })
        except ValueError:
            continue

    df_promoters = pd.DataFrame(promoter_records)
    df_promoters = df_promoters.drop_duplicates(subset=['chr', 'start', 'end'])
    print(f"✓ {len(df_promoters):,} regioni promotrici uniche generate (su chr20)")

    try:
        df_promoters.to_csv(local_csv, index=False)
        print(f"✓ Salvati promotori (chr20) in: {local_csv}")
    except Exception as e:
        print(f"✗ Errore salvataggio CSV promotori: {e}")
    
    return df_promoters

#################################################################################################
# STEP 4: CREAZIONE DATASET (Logica Positivo vs Negativo)
#################################################################################################

def filter_regions_by_feature(df_regions, df_features):
    """
    Mantiene solo le regioni (es. nucleosomi) che si sovrappongono  
    con i feature forniti (es. promotori, esoni).
    Restituisce un nuovo DataFrame filtrato.
    """
    print("  Filtraggio regioni per sovrapposizione con feature...")

    if df_features is None or df_features.empty:
        print("  ⚠ Nessuna annotazione feature fornita, ritorno tutte le regioni (nessun filtro).")
        return df_regions

    # Crea dizionario cromosoma → array di (start, end) per efficienza
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("  Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return pd.DataFrame(columns=df_regions.columns)
        
    keep = np.zeros(len(df_regions), dtype=bool)

    for i, row in enumerate(df_regions.itertuples(index=False)):
        chrom = getattr(row, 'chr', None)
        if chrom not in df_features_by_chr:
            continue
            
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        
        if s is None or e is None:
            continue

        # Cerca sovrapposizione
        for f_start, f_end in df_features_by_chr[chrom]:
            # (e < f_start) -> la regione è completamente prima del feature
            # (s > f_end)   -> la regione è completamente dopo il feature
            if not (e < f_start or s > f_end):  
                keep[i] = True
                break # Trovata sovrapposizione, passa alla prossima regione

    filtered_df = df_regions[keep]
    print(f"  ✓ {len(filtered_df):,}/{len(df_regions):,} regioni trovate nei feature forniti.")
    return filtered_df.reset_index(drop=True)

# --- MODIFICA 3 (NUOVA): Funzione POSITIVI con Peak Calling ---
def identify_nucleosome_positions(df_wig, threshold_value=1.0, sequence_length=201):
    """
    Identifica le posizioni delle diadi dei nucleosomi (picchi)
    usando la logica 'Non-Maximum Suppression'.
    """
    print("  Identificazione posizioni nucleosomi (Peak Calling)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    # 1. Trova tutti i possibili candidati positivi
    df_positive_candidates = df_wig[df_wig['score'] > threshold_value].copy()
    print(f"    Trovati {len(df_positive_candidates):,} candidati positivi (> {threshold_value:.2f})")

    if df_positive_candidates.empty:
        print("    ⚠ Nessun candidato positivo trovato.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    # 2. Ordina i candidati dal più forte (score più alto) al più debole
    df_sorted = df_positive_candidates.sort_values('score', ascending=False)
    
    true_dyad_rows = []
    claimed_tree = {} # Un albero di intervalli per le regioni già "bloccate"
    
    # La nostra separazione minima è la lunghezza della sequenza
    min_separation = sequence_length 

    print(f"    Avvio Non-Maximum Suppression (min_sep = {min_separation} bp)...")
    
    # Inizializza gli alberi per i cromosomi
    for chrom in df_sorted['chr'].unique():
        claimed_tree[chrom] = IntervalTree()

    # 3. Itera dal picco più forte in giù
    for row in df_sorted.itertuples(index=False):
        
        # 4. Controlla se questo picco è già stato "bloccato" da un picco più forte
        if claimed_tree[row.chr].overlaps(row.start, row.end):
            continue # Sì, ignora questo picco
            
        # 5. Se non è bloccato, è un VERO picco!
        true_dyad_rows.append(row)
        
        # 6. Blocca la regione attorno ad esso
        buffer_start = row.start - (min_separation // 2)
        buffer_end = row.start + (min_separation // 2) + 1 # +1 perché gli intervalli sono half-open
        claimed_tree[row.chr].addi(buffer_start, buffer_end)

    if not true_dyad_rows:
        print("    ⚠ Nessun picco positivo trovato dopo il peak calling.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    print(f"    ✓ Identificati {len(true_dyad_rows):,} picchi unici (diadi)")

    # 7. Crea il DataFrame finale dei positivi
    df_true_dyads = pd.DataFrame(true_dyad_rows, columns=df_sorted.columns)

    # 8. Espandi i centri (diadi) alla lunghezza finale della sequenza (201 bp)
    df_true_dyads['center'] = (df_true_dyads['start'] + df_true_dyads['end']) // 2
    df_true_dyads['start'] = df_true_dyads['center'] - (sequence_length // 2)
    df_true_dyads['end'] = df_true_dyads['start'] + sequence_length
    
    return df_true_dyads[['chr', 'start', 'end', 'score']]

# --- MODIFICA 4: Funzione NEGATIVI (score == 1.0) ---
def identify_non_nucleosome_positions(df_wig, threshold_value=1.0, nucleosome_length=201):
    """
    Identifica posizioni NON-nucleosomi (BASSA occupazione)
    usando una SOGLIA ASSOLUTA (score == threshold)
    """
    print("  Identificazione posizioni NON-nucleosomi (BASSA occupazione)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    # Usa il valore assoluto passato
    print(f"    Soglia segnale (Negativi): == {threshold_value:.2f}")
    
    # Filtra regioni con segnale BASSO (esattamente il background)
    df_nucleosomes = df_wig[df_wig['score'] == threshold_value].copy()
    
    # Estendi regioni alla lunghezza desiderata
    df_nucleosomes['center'] = (df_nucleosomes['start'] + df_nucleosomes['end']) // 2
    df_nucleosomes['start'] = df_nucleosomes['center'] - nucleosome_length // 2
    df_nucleosomes['end'] = df_nucleosomes['start'] + nucleosome_length
    
    print(f"    ✓ Identificati {len(df_nucleosomes)} NON-nucleosomi (Negativi)")
    
    return df_nucleosomes[['chr', 'start', 'end', 'score']]

def extract_sequences_from_genome(df_regions, genome_dict, max_sequences=2000):
    """
    Estrae sequenze DNA dalle regioni specificate
    """
    print(f"    Estrazione sequenze da {len(df_regions)} regioni...")
    
    sequences = []
    
    count = 0
    processed = 0
    for idx, row in df_regions.iterrows():
        processed += 1
        if count >= max_sequences:
            print(f"      Raggiunto limite di {max_sequences} sequenze.")
            break
        
        chrom = row['chr']
        start = int(row['start'])
        end = int(row['end'])
        
        if chrom not in genome_dict:
            continue
        
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        sequences.append(seq)
        count += 1
        
        if count % 500 == 0 and count > 0:
            print(f"      Estraendole... {count}/{max_sequences} sequenze valide trovate (processate {processed} regioni)")
    
    print(f"    ✓ Estratte {len(sequences)} sequenze")
    return sequences

# --- MODIFICA 5: NUOVA Funzione di Estrazione Efficiente (con Buffer Zone) ---
def extract_filtered_sequences_efficiently(df_regions, df_features, genome_dict, positive_regions_tree, max_sequences=2000, buffer_zone=201):
    """
    Combina filtro e estrazione in un unico passaggio efficiente.
    Mescola i candidati e si ferma non appena 'max_sequences' sono state trovate.
    ESCLUDE i negativi che si sovrappongono ai positivi (buffer_zone).
    """
    print(f"  Avvio estrazione efficiente per {max_sequences} sequenze da {len(df_regions):,} candidati...")
    
    sequences = []
    
    # 1. Crea il dizionario di features per una ricerca rapida
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("    Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return []

    # 2. Mescola i candidati (df_regions)
    #    Usiamo sample(frac=1) che è un modo efficiente in pandas per mescolare
    df_regions_shuffled = df_regions.sample(frac=1)

    processed_count = 0
    excluded_by_buffer = 0
    
    # 3. Itera, filtra, ed estrai in un unico loop
    for row in df_regions_shuffled.itertuples(index=False):
        processed_count += 1
        
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        chrom = getattr(row, 'chr', None)
        
        if s is None or e is None or chrom is None:
            continue
            
        # --- NUOVO CONTROLLO: ZONA CUSCINETTO ---
        # Controlla se questa regione negativa (s, e) è troppo vicina a un positivo.
        # Creiamo un intervallo buffer attorno ad essa.
        buffer_start = s - buffer_zone
        buffer_end = e + buffer_zone
        if chrom in positive_regions_tree and positive_regions_tree[chrom].overlaps(buffer_start, buffer_end):
            excluded_by_buffer += 1
            continue # Questo negativo è troppo vicino a un positivo, scartalo.
        # ----------------------------------------

        # Controllo 2: Logica di 'filter_regions_by_feature' (promotori)
        if chrom not in df_features_by_chr:
            continue
            
        has_overlap = False
        for f_start, f_end in df_features_by_chr[chrom]:
            if not (e < f_start or s > f_end):
                has_overlap = True
                break
        
        if not has_overlap:
            continue # Non è in un promotore, passa al prossimo candidato

        # --- Se siamo qui, la regione E' in un promotore E NON è vicina a un positivo ---
        
        # Logica di 'extract_sequences_from_genome'
        start = int(s)
        end = int(e)
        
        if chrom not in genome_dict:
            continue
            
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        # --- Se siamo qui, la sequenza è VALIDA ---
        sequences.append(seq)
        
        # Logica di 'stop'
        if len(sequences) >= max_sequences:
            print(f"    ✓ Raggiunto limite di {max_sequences} sequenze valide (processati {processed_count:,} candidati).")
            print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")
            break
            
    if len(sequences) < max_sequences:
        print(f"    ✓ Estrazione completata. Trovate {len(sequences)} sequenze valide (processati tutti i {len(df_regions):,} candidati).")
        print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")

    return sequences


# --- MODIFICA 6: Funzione DATASET (usa la funzione efficiente e l'albero) ---
def create_nucleosome_dataset(wig_files, genome_dict,  
                              max_sequences_per_file=1000,
                              threshold_high=1.0,
                              threshold_low=1.0,
                              max_regions_per_wig=100000,
                              df_features=None,
                              sequence_length=201):
    """
    Crea dataset finale combinando dati WIG e genoma.
    Positivi: score > threshold_high
    Negativi: score == threshold_low
    """
    print("\n" + "="*60)
    print("CREAZIONE DATASET NUCLEOSOMI (Positivi vs Negativi in-loco)")
    print("="*60)
    print(f"Lunghezza sequenza: {sequence_length} bp")
    print(f"Soglia ALTA (positivi): > {threshold_high:.2f}")
    print(f"Soglia BASSA (negativi): == {threshold_low:.2f}")
    print(f"Filtro regioni: {df_features is not None}")
    
    all_sequences = []
    all_labels = []
    
    # Processa TUTTI i file WIG
    for idx, wig_file in enumerate(wig_files, 1):
        print(f"\n[{idx}/{len(wig_files)}] Processando: {os.path.basename(wig_file)}")
        
        # Passa max_regions_per_wig al parser
        df_wig = parse_wig_file(wig_file, max_regions=max_regions_per_wig)
        
        if df_wig is None or len(df_wig) == 0:
            print("  ⚠ Nessun dato trovato, skip")
            continue
            
        # -----------------------------------------------------------------
        # 1. STREAM POSITIVO (Label 1)
        # -----------------------------------------------------------------
        print("\n  --- Stream Positivo (Label 1) ---")
        # 1a. Trova TUTTI i positivi (>1.0) nell'INTERO GENOMA (con Peak Calling)
        df_nucleosomes_high = identify_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_high, # Passa threshold_high (1.0)
            sequence_length=sequence_length
        )
        
        # 1b. Filtra i positivi SOLO per i promotori (per il set di training)
        df_positive_filtered = filter_regions_by_feature(df_nucleosomes_high, df_features)

        if df_positive_filtered.empty:
            print("  ⚠ Nessuna regione positiva (>1.0) nei feature forniti, skip file")
            continue
            
        # 1c. Estrai sequenze
        sequences_pos = extract_sequences_from_genome(
            df_positive_filtered,  
            genome_dict,  
            max_sequences=max_sequences_per_file 
        )
        
        # --- NUOVO PASSAGGIO: Crea l'albero dei positivi per la zona cuscinetto ---
        # Usa la lista TOTALE (df_nucleosomes_high) per l'albero
        print("    Creazione albero intervalli positivi per zona cuscinetto...")
        positive_tree = {}
        for chrom in df_nucleosomes_high['chr'].unique():
            positive_tree[chrom] = IntervalTree()
            for row in df_nucleosomes_high[df_nucleosomes_high['chr'] == chrom].itertuples():
                positive_tree[chrom].addi(row.start, row.end)
        # ---------------------------------------------------------------------

        # -----------------------------------------------------------------
        # 2. STREAM NEGATIVO (Label 0) - Logica efficiente
        # -----------------------------------------------------------------
        print("\n  --- Stream Negativo (Label 0) ---")
        
        n_to_find = len(sequences_pos) # Numero di negativi da trovare
        if n_to_find == 0:
             print("  ⚠ Nessuna sequenza positiva trovata, skip negativi")
             continue
        
        # 2a. Trova TUTTI i candidati negativi (==1.0)
        df_nucleosomes_low = identify_non_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_low, # Passa threshold_low (1.0)
            nucleosome_length=sequence_length
        )
        
        if df_nucleosomes_low.empty:
            print("  ⚠ Nessuna regione negativa (==1.0) trovata, skip negativi")
            continue
        
        # 2b. --- CHIAMATA ALLA NUOVA FUNZIONE EFFICIENTE (CON ALBERO) ---
        sequences_neg = extract_filtered_sequences_efficiently(
            df_nucleosomes_low,     # I milioni di candidati
            df_features,            # I promotori
            genome_dict,
            positive_tree,          # L'albero dei positivi da evitare
            max_sequences=n_to_find, # Il numero da trovare
            buffer_zone=sequence_length # Buffer = 201 bp
        )
        
        if not sequences_neg:
             print("  ⚠ Nessuna sequenza negativa valida trovata (o tutte troppo vicine ai positivi).")
             continue
        
        # -----------------------------------------------------------------
        # 3. BILANCIAMENTO e AGGIUNTA AL DATASET
        # -----------------------------------------------------------------
        
        n_pos = len(sequences_pos)
        n_neg = len(sequences_neg)

        n_to_keep = min(n_pos, n_neg)
        
        if n_pos > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_pos} positivi a {n_to_keep}")
            sequences_pos = random.sample(sequences_pos, n_to_keep)
        
        if n_neg > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_neg} negativi a {n_to_keep}")
            sequences_neg = random.sample(sequences_neg, n_to_keep)

        # -----------------------------------------------------------------
        
        all_sequences.extend(sequences_pos)
        all_labels.extend([1] * len(sequences_pos))
        
        all_sequences.extend(sequences_neg)
        all_labels.extend([0] * len(sequences_neg))
        
        print(f"\n  RISULTATO FILE: Aggiunti {len(sequences_pos)} positivi e {len(sequences_neg)} negativi (Bilanciati).")
        print(f"  ✓ Totale sequenze accumulate: {len(all_sequences):,}")

    
    if not all_sequences:
        print("⚠⚠ NESSUNA SEQUENZA TROVATA. Controlla i percorsi e i filtri.")
        return pd.DataFrame(columns=['sequenza', 'nucleosoma'])

    # Crea DataFrame
    df = pd.DataFrame({
        'sequenza': all_sequences,
        'nucleosoma': all_labels
    })
    
    df = df.drop_duplicates(subset=['sequenza']).reset_index(drop=True)
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    
    print("\n" + "="*60)
    print("DATASET COMPLETATO (Bilanciato per regione)")
    print("="*60)
    print(f"Totale sequenze: {len(df):,}")
    print(f"Nucleosomi (positivi): {df['nucleosoma'].sum():,}")
    print(f"Non-nucleosomi (negativi): {(1 - df['nucleosoma']).sum():,}")
    
    if len(df) > 0:
        print(f"Bilanciamento: {df['nucleosoma'].sum() / len(df) * 100:.1f}% positivi")
    
    return df

#################################################################################################
# ESECUZIONE PRINCIPALE
#################################################################################################

def main():
    print("="*60)
    print("ELABORAZIONE DATI NUCLEOSOMI GSE36979 (SOLO CHR20)")
    print("="*60)

    # Imposta directory dati
    base_dir = os.getcwd()
    data_dir = os.path.abspath(os.path.join(base_dir, "..", "data"))
    
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        print(f"Creata directory dati: {data_dir}")

    # STEP 1: Estrai file WIG dal TAR
    tar_path = os.path.join(data_dir, "GSE36979_RAW.tar")
    wig_extract_dir = os.path.join(data_dir, "wig_files")

    if not os.path.exists(tar_path):
        print(f"✗ File non trovato: {tar_path}")
        print("Scaricare GSE36979_RAW.tar da GEO e posizionarlo in ../data/")
        return

    wig_files = extract_tar_wig_files(tar_path, extract_dir=wig_extract_dir)

    if not wig_files:
        print("✗ Nessun file WIG 147bp trovato nell'archivio. Controlla il file TAR.")
        return

    # STEP 2: Scarica/Carica genoma hg18
    print("\n" + "="*60)
    genome_dir = os.path.join(data_dir, "hg18_genome")

    if os.path.exists(genome_dir) and os.path.exists(os.path.join(genome_dir, "chr20.fa")):
        print("Genoma hg18 (chr20) già presente, caricamento...")
        genome_dict = load_hg18_genome(genome_dir)
    else:
        print("Scaricamento genoma hg18 (solo chr20)...")
        genome_dict = download_hg18_genome(genome_dir)

    if not genome_dict:
        print("✗ Impossibile caricare genoma")
        return

    # STEP 2.5: Carica o scarica annotazioni RefGene per i promotori
    print("\n" + "="*60)
    print("Caricamento/Generazione annotazioni promotori (RefGene hg18 - CHR20)...")
    
    try:
        df_promoters = download_refgene_hg18(output_dir=data_dir, promoter_window=1000)
    except Exception as e:
        print(f"✗ Errore fatale durante il caricamento dei promotori: {e}")
        df_promoters = None

    if df_promoters is None or df_promoters.empty:
        print("✗ Impossibile caricare le annotazioni dei promotori. Uscita.")
        return
    else:
        print(f"✓ Annotazioni promotori caricate: {len(df_promoters)} regioni (su chr20)")
        
    # --- ALTERNATIVA: USA ENHANCER (esempio se avessi un file BED) ---
    # enhancer_bed_path = os.path.join(data_dir, "hg18_enhancers.bed")
    # df_features_to_use = load_bed_features(enhancer_bed_path)
    # ------------------------------------------------------------------
    df_features_to_use = df_promoters # Scegliamo di usare i promotori


    # --- MODIFICA 7: Chiamata principale con i parametri finali ---
    # STEP 3: Crea dataset
    df_final = create_nucleosome_dataset(
        wig_files,  
        genome_dict,
        max_sequences_per_file=100000,
        threshold_high=1.0,           # Positivi > 2.0 (era 1.0, ma 2.0 è più selettivo)
        threshold_low=1.0,            # Negativi == 1.0
        max_regions_per_wig=10000000, # Aumentato per sicurezza, anche se filtriamo
        df_features=df_features_to_use,
        sequence_length=201
    )

    if df_final is None or df_final.empty:
        print("✗ Creazione dataset fallita o dataset vuoto.")
        return

    # STEP 4: Salva risultati
    ### MODIFICA: Cambiato nome file di output ###
    output_file = os.path.join(data_dir, "nucleosome_dataset_GSE36979_promoters_201pb_balanced_CHR20_ONLY.csv")
    try:
        df_final.to_csv(output_file, index=False)
        print(f"\n✓ Dataset (CHR20) salvato: {output_file}")
    except Exception as e:
        print(f"✗ Errore salvataggio dataset: {e}")
        return

    # STEP 5: Statistiche finali
    print("\n" + "="*60)
    print("STATISTICHE FINALI (CHR20)")
    print("="*60)
    print(df_final.head(10))
    
    print(f"\nDistribuzione lunghezze:")
    print(df_final['sequenza'].str.len().describe())

    # GC content
    df_final['gc_content'] = df_final['sequenza'].apply(calc_gc)
    print(f"\nGC content per classe:")
    print(df_final.groupby('nucleosoma')['gc_content'].describe())

    # Plot
    print("\nGenerazione plot GC content...")
    try:
        plt.figure(figsize=(10, 6))
        sns.kdeplot(data=df_final, x='gc_content', hue='nucleosoma', fill=True, common_norm=False, palette="Set1")
        plt.title('Distribuzione GC Content tra Classi (0=Bassa, 1=Alta Occupazione) - CHR20')
        plot_file = os.path.join(data_dir, "gc_content_distribution_CHR20_ONLY.png")
        plt.savefig(plot_file)
        print(f"✓ Plot salvato: {plot_file}")
        # plt.show() # Decommenta se sei in un notebook
    except Exception as e:
        print(f"✗ Errore generazione plot: {e}")

    print("\n✓ ELABORAZIONE (CHR20) COMPLETATA!")

if __name__ == "__main__":  
    main()
```

```python
import pandas as pd
import numpy as np
import requests
import gzip
import os
from Bio import SeqIO
from io import StringIO
import random
import matplotlib.pyplot as plt
import seaborn as sns
from intervaltree import IntervalTree 
import glob # Per cercare i file

#################################################################################################
# CONFIGURAZIONE GLOBALE
#################################################################################################

# Processa tutti i cromosomi standard
CHROMS_TO_PROCESS = [f"chr{i}" for i in range(1, 23)] + ["chrX", "chrY"]

#################################################################################################
# FUNZIONI HELPER
#################################################################################################

def calc_gc(seq):
    """Calcola il contenuto GC di una sequenza."""
    if not isinstance(seq, str) or len(seq) == 0:
        return 0
    gc = seq.count('G') + seq.count('C')
    return (gc / len(seq)) * 100

#################################################################################################
# STEP 1: GESTIONE GENOMA HG19 (TUTTI I CROMOSOMI)
#################################################################################################

def download_hg19_genome(output_dir="hg19_genome"):
    """
    Scarica il genoma umano hg19 da UCSC (TUTTI I CROMOSOMI STANDARD)
    """
    print(f"Scaricando genoma hg19 da UCSC per {len(CHROMS_TO_PROCESS)} cromosomi...")
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # URL base UCSC per hg19
    base_url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg19/chromosomes/"
    
    ### MODIFICA: Processa tutti i cromosomi ###
    chromosomes = CHROMS_TO_PROCESS
    
    genome_dict = {}
    
    for chrom in chromosomes:
        file_name = f"{chrom}.fa.gz"
        file_path = os.path.join(output_dir, file_name)
        
        # Percorso del file decompresso
        unzipped_file_path = file_path.replace('.gz', '')

        if os.path.exists(unzipped_file_path):
            print(f"✓ {chrom} (hg19) già scaricato")
            # Carica sequenza già presente
            try:
                with open(unzipped_file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        genome_dict[chrom] = str(record.seq).upper() # Salva in maiuscolo
                continue
            except Exception as e:
                print(f"✗ Errore nel caricamento di {unzipped_file_path}: {e}. Tento di riscaricarlo.")

        
        url = base_url + file_name
        print(f"Scaricando {chrom} (hg19)...")
        
        try:
            response = requests.get(url, stream=True, timeout=300)
            if response.status_code == 200:
                # Salva file compresso
                with open(file_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                
                # Decomprimi e leggi
                sequence_content = ""
                with gzip.open(file_path, 'rt') as f:
                    sequence_content = f.read()
                
                # Parse dalla stringa
                with StringIO(sequence_content) as f_in:
                    records = SeqIO.parse(f_in, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Salva in maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} (hg19) scaricato: {len(seq_str):,} bp")
                
                # Salva versione decompressa
                with open(unzipped_file_path, 'w') as f_out:
                    f_out.write(sequence_content)
                
                # Rimuovi file compresso per salvare spazio
                os.remove(file_path)
                
        except Exception as e:
            print(f"✗ Errore scaricamento {chrom} (hg19): {e}")
            if os.path.exists(file_path):
                os.remove(file_path) # Pulisci file corrotto
            continue
    
    print(f"\n✓ Genoma hg19 scaricato: {len(genome_dict)} cromosomi")
    return genome_dict

def load_hg19_genome(genome_dir="hg19_genome"):
    """
    Carica genoma hg19 già scaricato (TUTTI I CROMOSOMI STANDARD)
    """
    print("Caricamento genoma hg19...")
    
    genome_dict = {}
    
    ### MODIFICA: Processa tutti i cromosomi ###
    chromosomes = CHROMS_TO_PROCESS
    
    for chrom in chromosomes:
        file_path = os.path.join(genome_dir, f"{chrom}.fa")
        
        if os.path.exists(file_path):
            try:
                with open(file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Assicurati sia maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} (hg19) caricato: {len(seq_str):,} bp")
            except Exception as e:
                print(f"✗ Errore caricamento {file_path}: {e}")
        else:
            print(f"✗ {chrom} non trovato in {genome_dir}")
    
    if not genome_dict:
        print(f"✗ Nessun file genoma hg19 trovato in {genome_dir}. Prova a scaricarlo prima.")
    else:
        print(f"\n✓ Genoma hg19 caricato: {len(genome_dict)} cromosomi")
    return genome_dict

#################################################################################################
# STEP 2: GESTIONE DATI WIG (TUTTI I CROMOSOMI)
#################################################################################################
def parse_wig_file(wig_file_path, max_regions=500000000):
    """
    Parsifica file WIG con dati di occupancy dei nucleosomi (TUTTI I CROMOSOMI)
    VERSIONE v3 (ROBUSTA): Gestisce fixedStep, variableStep, e bedGraph
    """
    print(f"  Parsing file WIG: {os.path.basename(wig_file_path)}")
    
    regions = []
    current_chrom = None
    current_start = None
    current_step = None
    current_span = 1
    position = 0
    lines_processed = 0
    
    try: 
        opener = gzip.open if wig_file_path.endswith('.gz') else open
        
        with opener(wig_file_path, 'rt') as f:
            for line in f:
                line = line.strip()
                
                # Salta intestazioni e righe vuote
                if not line or line.startswith('track') or line.startswith('browser'):
                    continue
                
                lines_processed += 1
                
                if lines_processed % 50000000 == 0:
                    print(f"    ... Processate {lines_processed:,} linee, trovate {len(regions):,} regioni...")
                
                if len(regions) >= max_regions:
                    print(f"    ⚠ Raggiunto limite di {max_regions:,} regioni, stop parsing")
                    break
                
                # --- GESTIONE HEADER ---
                if line.startswith('fixedStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('start='):
                            current_start = int(part.split('=')[1])
                            position = current_start
                        elif part.startswith('step='):
                            current_step = int(part.split('=')[1])
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                
                elif line.startswith('variableStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                    current_step = None # Flag per variableStep
                
                # --- GESTIONE DATA LINE ---
                else:
                    try:
                        parts = line.split()
                        
                        # --- MODIFICA CHIAVE: Controlla il formato (bedGraph, variableStep, fixedStep) ---
                        
                        if len(parts) >= 4:
                            # Formato bedGraph (chr start end value ...)
                            chrom, start_str, end_str, value_str = parts[0], parts[1], parts[2], parts[3]
                            if chrom not in CHROMS_TO_PROCESS:
                                continue
                            value = float(value_str)
                            if value > 0: # Filtro originale (corretto, se i dati sono >= 1)
                                regions.append({
                                    'chr': chrom,
                                    'start': int(start_str),
                                    'end': int(end_str),
                                    'score': value
                                })
                        
                        elif len(parts) == 2 and current_step is None:
                            # Formato variableStep (pos value)
                            if current_chrom not in CHROMS_TO_PROCESS:
                                continue
                            pos, value_str = parts[0], parts[1]
                            pos = int(pos)
                            value = float(value_str)
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': pos,
                                    'end': pos + current_span,
                                    'score': value
                                })
                        
                        elif len(parts) == 1 and current_step is not None:
                            # Formato fixedStep (value)
                            if current_chrom not in CHROMS_TO_PROCESS:
                                continue
                            value = float(parts[0])
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': position,
                                    'end': position + current_span,
                                    'score': value
                                })
                            position += current_step
                        
                        # Altri formati (es. bedGraph senza header) vengono ignorati

                    except Exception:
                        # Salta righe dati malformate
                        continue
            
        print(f"    ✓ Trovate {len(regions):,} regioni con segnale (Tutti i cromosomi)")
        return pd.DataFrame(regions)
    
    except Exception as e:
        print(f"    ✗ Errore parsing WIG: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

#################################################################################################
# STEP 3: GESTIONE ANNOTAZIONI (Funzioni non più usate, ma definite)
#################################################################################################
# Queste funzioni (load_genomic_features, download_refgene_hg19, etc.) 
# non sono più chiamate da main() ma sono lasciate qui se servissero in futuro.

#################################################################################################
# STEP 4: CREAZIONE DATASET (Funzioni generiche, modificate per "No Filtro")
#################################################################################################

def filter_regions_by_feature(df_regions, df_features):
    """
    Mantiene solo le regioni che si sovrappongono con i feature forniti.
    Se df_features è None, restituisce TUTTE le regioni.
    """
    print("  Filtraggio regioni per sovrapposizione con feature...")

    if df_features is None or df_features.empty:
        print("  ⚠ Nessuna annotazione feature fornita, ritorno tutte le regioni (nessun filtro).")
        return df_regions # <-- Restituisce tutto se non c'è filtro

    # Crea dizionario cromosoma → array di (start, end) per efficienza
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("  Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return pd.DataFrame(columns=df_regions.columns)
        
    keep = np.zeros(len(df_regions), dtype=bool)

    for i, row in enumerate(df_regions.itertuples(index=False)):
        chrom = getattr(row, 'chr', None)
        if chrom not in df_features_by_chr:
            continue
            
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        
        if s is None or e is None:
            continue

        # Cerca sovrapposizione
        for f_start, f_end in df_features_by_chr[chrom]:
            if not (e < f_start or s > f_end):  
                keep[i] = True
                break # Trovata sovrapposizione, passa alla prossima regione

    filtered_df = df_regions[keep]
    print(f"  ✓ {len(filtered_df):,}/{len(df_regions):,} regioni trovate nei feature forniti.")
    return filtered_df.reset_index(drop=True)

def identify_nucleosome_positions(df_wig, threshold_value=1.0, sequence_length=201):
    """
    Identifica le posizioni delle diadi dei nucleosomi (picchi)
    usando la logica 'Non-Maximum Suppression'.
    """
    print("  Identificazione posizioni nucleosomi (Peak Calling)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    df_positive_candidates = df_wig[df_wig['score'] > threshold_value].copy()
    print(f"    Trovati {len(df_positive_candidates):,} candidati positivi (> {threshold_value:.2f})")

    if df_positive_candidates.empty:
        print("    ⚠ Nessun candidato positivo trovato.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    df_sorted = df_positive_candidates.sort_values('score', ascending=False)
    
    true_dyad_rows = []
    claimed_tree = {} # Un albero di intervalli per le regioni già "bloccate"
    
    min_separation = sequence_length  

    print(f"    Avvio Non-Maximum Suppression (min_sep = {min_separation} bp)...")
    
    # Inizializza gli alberi per i cromosomi
    for chrom in df_sorted['chr'].unique():
        if chrom in CHROMS_TO_PROCESS: # Processa solo i nostri cromosomi
            claimed_tree[chrom] = IntervalTree()

    # Itera dal picco più forte in giù
    for row in df_sorted.itertuples(index=False):
        
        # Ignora cromosomi non standard
        if row.chr not in claimed_tree:
            continue
            
        # Controlla se questo picco è già stato "bloccato" da un picco più forte
        if claimed_tree[row.chr].overlaps(row.start, row.end):
            continue # Sì, ignora questo picco
            
        # Se non è bloccato, è un VERO picco!
        true_dyad_rows.append(row)
        
        # Blocca la regione attorno ad esso
        buffer_start = row.start - (min_separation // 2)
        buffer_end = row.start + (min_separation // 2) + 1 
        claimed_tree[row.chr].addi(buffer_start, buffer_end)

    if not true_dyad_rows:
        print("    ⚠ Nessun picco positivo trovato dopo il peak calling.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    print(f"    ✓ Identificati {len(true_dyad_rows):,} picchi unici (diadi)")

    df_true_dyads = pd.DataFrame(true_dyad_rows, columns=df_sorted.columns)

    # Espandi i centri (diadi) alla lunghezza finale della sequenza (201 bp)
    df_true_dyads['center'] = (df_true_dyads['start'] + df_true_dyads['end']) // 2
    df_true_dyads['start'] = df_true_dyads['center'] - (sequence_length // 2)
    df_true_dyads['end'] = df_true_dyads['start'] + sequence_length
    
    return df_true_dyads[['chr', 'start', 'end', 'score']]

def identify_non_nucleosome_positions(df_wig, threshold_value=1.0, nucleosome_length=201):
    """
    Identifica posizioni NON-nucleosomi (BASSA occupazione)
    usando una SOGLIA ASSOLUTA (score == threshold)
    """
    print("  Identificazione posizioni NON-nucleosomi (BASSA occupazione)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    print(f"    Soglia segnale (Negativi): == {threshold_value:.2f}")
    
    # Filtra regioni con segnale BASSO (esattamente il background)
    df_nucleosomes = df_wig[df_wig['score'] == threshold_value].copy()
    
    # Estendi regioni alla lunghezza desiderata
    df_nucleosomes['center'] = (df_nucleosomes['start'] + df_nucleosomes['end']) // 2
    df_nucleosomes['start'] = df_nucleosomes['center'] - nucleosome_length // 2
    df_nucleosomes['end'] = df_nucleosomes['start'] + nucleosome_length
    
    print(f"    ✓ Identificati {len(df_nucleosomes)} NON-nucleosomi (Negativi)")
    
    return df_nucleosomes[['chr', 'start', 'end', 'score']]

def extract_sequences_from_genome(df_regions, genome_dict, max_sequences=2000):
    """
    Estrae sequenze DNA dalle regioni specificate
    """
    print(f"    Estrazione sequenze da {len(df_regions)} regioni...")
    
    sequences = []
    
    count = 0
    processed = 0
    # Mescola per ottenere un campione rappresentativo se max_sequences è basso
    df_shuffled = df_regions.sample(frac=1) 
    
    for idx, row in df_shuffled.iterrows():
        processed += 1
        if count >= max_sequences:
            print(f"      Raggiunto limite di {max_sequences} sequenze.")
            break
        
        chrom = row['chr']
        start = int(row['start'])
        end = int(row['end'])
        
        if chrom not in genome_dict:
            continue
        
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        sequences.append(seq)
        count += 1
        
        if count % 5000 == 0 and count > 0:
            print(f"      Estraendole... {count}/{max_sequences} sequenze valide trovate (processate {processed} regioni)")
    
    print(f"    ✓ Estratte {len(sequences)} sequenze")
    return sequences

def extract_filtered_sequences_efficiently(df_regions, df_features, genome_dict, positive_regions_tree, max_sequences=2000, buffer_zone=201):
    """
    Combina filtro e estrazione in un unico passaggio efficiente.
    Mescola i candidati e si ferma non appena 'max_sequences' sono state trovate.
    ESCLUDE i negativi che si sovrappongono ai positivi (buffer_zone).
    Se df_features è None, salta il filtro per feature.
    """
    print(f"  Avvio estrazione efficiente per {max_sequences} sequenze da {len(df_regions):,} candidati...")
    
    sequences = []
    
    # 1. Crea il dizionario di features per una ricerca rapida
    df_features_by_chr = None
    if df_features is not None and not df_features.empty:
        print("  (Filtro feature ATTIVO)")
        try:
            df_features_by_chr = {
                chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
                for chrom in df_features['chr'].unique()
            }
        except KeyError:
            print("    Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
            return []
    else:
        print("  (Filtro feature DISABILITATO)")

    # 2. Mescola i candidati (df_regions)
    df_regions_shuffled = df_regions.sample(frac=1)

    processed_count = 0
    excluded_by_buffer = 0
    
    # 3. Itera, filtra, ed estrai in un unico loop
    for row in df_regions_shuffled.itertuples(index=False):
        processed_count += 1
        
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        chrom = getattr(row, 'chr', None)
        
        if s is None or e is None or chrom is None or chrom not in CHROMS_TO_PROCESS:
            continue
            
        # --- CONTROLLO 1: ZONA CUSCINETTO ---
        buffer_start = s - buffer_zone
        buffer_end = e + buffer_zone
        if chrom in positive_regions_tree and positive_regions_tree[chrom].overlaps(buffer_start, buffer_end):
            excluded_by_buffer += 1
            continue # Questo negativo è troppo vicino a un positivo, scartalo.
        
        # --- CONTROLLO 2: FILTRO FEATURE (es. promotori) ---
        if df_features_by_chr is not None:
            if chrom not in df_features_by_chr:
                continue # Regione in un cromosoma senza feature
                
            has_overlap = False
            for f_start, f_end in df_features_by_chr[chrom]:
                if not (e < f_start or s > f_end):
                    has_overlap = True
                    break
            
            if not has_overlap:
                continue # Non è in un feature, passa al prossimo candidato

        # --- Se siamo qui, la regione E' valida (o non filtrata) E NON è vicina a un positivo ---
        
        # Logica di 'extract_sequences_from_genome'
        start = int(s)
        end = int(e)
        
        if chrom not in genome_dict:
            continue
            
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        # --- Se siamo qui, la sequenza è VALIDA ---
        sequences.append(seq)
        
        # Logica di 'stop'
        if len(sequences) >= max_sequences:
            print(f"    ✓ Raggiunto limite di {max_sequences} sequenze valide (processati {processed_count:,} candidati).")
            print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")
            break
            
    if len(sequences) < max_sequences:
        print(f"    ✓ Estrazione completata. Trovate {len(sequences)} sequenze valide (processati tutti i {len(df_regions):,} candidati).")
        print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")

    return sequences


def create_nucleosome_dataset(wig_files, genome_dict,  
                              max_sequences_per_file=1000,
                              threshold_high=1.0,
                              threshold_low=1.0,
                              max_regions_per_wig=100000,
                              df_features=None,
                              sequence_length=201):
    """
    Crea dataset finale combinando dati WIG e genoma.
    Positivi: score > threshold_high
    Negativi: score == threshold_low
    """
    print("\n" + "="*60)
    print("CREAZIONE DATASET NUCLEOSOMI (Positivi vs Negativi in-loco)")
    print("="*60)
    print(f"Lunghezza sequenza: {sequence_length} bp")
    print(f"Soglia ALTA (positivi): > {threshold_high:.2f}")
    print(f"Soglia BASSA (negativi): == {threshold_low:.2f}")
    if df_features is not None:
        print(f"Filtro regioni: ATTIVO (usando {len(df_features)} feature)")
    else:
        print("Filtro regioni: DISABILITATO (intero genoma)")
    
    all_sequences = []
    all_labels = []
    
    # Processa TUTTI i file WIG
    for idx, wig_file in enumerate(wig_files, 1):
        print(f"\n[{idx}/{len(wig_files)}] Processando: {os.path.basename(wig_file)}")
        
        df_wig = parse_wig_file(wig_file, max_regions=max_regions_per_wig)
        
        if df_wig is None or len(df_wig) == 0:
            print("  ⚠ Nessun dato trovato, skip")
            continue
            
        # -----------------------------------------------------------------
        # 1. STREAM POSITIVO (Label 1)
        # -----------------------------------------------------------------
        print("\n  --- Stream Positivo (Label 1) ---")
        # 1a. Trova TUTTI i positivi (>1.0) nell'INTERO GENOMA (con Peak Calling)
        df_nucleosomes_high = identify_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_high, 
            sequence_length=sequence_length
        )
        
        # 1b. Filtra i positivi SOLO per i feature (se forniti)
        df_positive_filtered = filter_regions_by_feature(df_nucleosomes_high, df_features)

        if df_positive_filtered.empty:
            print("  ⚠ Nessuna regione positiva (>1.0) trovata [dopo il filtro], skip file")
            continue
            
        # 1c. Estrai sequenze
        sequences_pos = extract_sequences_from_genome(
            df_positive_filtered,  
            genome_dict,  
            max_sequences=max_sequences_per_file 
        )
        
        # --- Crea l'albero dei positivi per la zona cuscinetto ---
        # Usa la lista TOTALE (df_nucleosomes_high) per l'albero
        print("    Creazione albero intervalli positivi per zona cuscinetto...")
        positive_tree = {}
        for chrom in df_nucleosomes_high['chr'].unique():
            if chrom in CHROMS_TO_PROCESS: # Assicurati
                positive_tree[chrom] = IntervalTree()
                for row in df_nucleosomes_high[df_nucleosomes_high['chr'] == chrom].itertuples():
                    positive_tree[chrom].addi(row.start, row.end)
        # ---------------------------------------------------------------------

        # -----------------------------------------------------------------
        # 2. STREAM NEGATIVO (Label 0) - Logica efficiente
        # -----------------------------------------------------------------
        print("\n  --- Stream Negativo (Label 0) ---")
        
        n_to_find = len(sequences_pos) # Numero di negativi da trovare
        if n_to_find == 0:
             print("  ⚠ Nessuna sequenza positiva trovata, skip negativi")
             continue
        
        # 2a. Trova TUTTI i candidati negativi (==1.0)
        df_nucleosomes_low = identify_non_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_low,
            nucleosome_length=sequence_length
        )
        
        if df_nucleosomes_low.empty:
            print("  ⚠ Nessuna regione negativa (==1.0) trovata, skip negativi")
            continue
        
        # 2b. --- CHIAMATA ALLA NUOVA FUNZIONE EFFICIENTE (CON ALBERO) ---
        sequences_neg = extract_filtered_sequences_efficiently(
            df_nucleosomes_low,       # I milioni di candidati
            df_features,              # I promotori (o None)
            genome_dict,
            positive_tree,            # L'albero dei positivi da evitare
            max_sequences=n_to_find,  # Il numero da trovare
            buffer_zone=sequence_length 
        )
        
        if not sequences_neg:
             print("  ⚠ Nessuna sequenza negativa valida trovata (o tutte troppo vicine ai positivi).")
             continue
        
        # -----------------------------------------------------------------
        # 3. BILANCIAMENTO e AGGIUNTA AL DATASET
        # -----------------------------------------------------------------
        
        n_pos = len(sequences_pos)
        n_neg = len(sequences_neg)

        n_to_keep = min(n_pos, n_neg)
        
        if n_pos > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_pos} positivi a {n_to_keep}")
            sequences_pos = random.sample(sequences_pos, n_to_keep)
        
        if n_neg > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_neg} negativi a {n_to_keep}")
            sequences_neg = random.sample(sequences_neg, n_to_keep)

        all_sequences.extend(sequences_pos)
        all_labels.extend([1] * len(sequences_pos))
        
        all_sequences.extend(sequences_neg)
        all_labels.extend([0] * len(sequences_neg))
        
        print(f"\n  RISULTATO FILE: Aggiunti {len(sequences_pos)} positivi e {len(sequences_neg)} negativi (Bilanciati).")
        print(f"  ✓ Totale sequenze accumulate: {len(all_sequences):,}")

    
    if not all_sequences:
        print("⚠⚠ NESSUNA SEQUENZA TROVATA. Controlla i percorsi e i filtri.")
        return pd.DataFrame(columns=['sequenza', 'nucleosoma'])

    # Crea DataFrame
    df = pd.DataFrame({
        'sequenza': all_sequences,
        'nucleosoma': all_labels
    })
    
    df = df.drop_duplicates(subset=['sequenza']).reset_index(drop=True)
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    
    print("\n" + "="*60)
    print("DATASET COMPLETATO (Bilanciato per regione)")
    print("="*60)
    print(f"Totale sequenze: {len(df):,}")
    print(f"Nucleosomi (positivi): {df['nucleosoma'].sum():,}")
    print(f"Non-nucleosomi (negativi): {(1 - df['nucleosoma']).sum():,}")
    
    if len(df) > 0:
        print(f"Bilanciamento: {df['nucleosoma'].sum() / len(df) * 100:.1f}% positivi")
    
    return df

#################################################################################################
# ESECUZIONE PRINCIPALE (Modificata per hg19 / INTERO GENOMA)
#################################################################################################

def main():
    print("="*60)
    print("ELABORAZIONE DATI NUCLEOSOMI GSE36979 (HG19 - INTERO GENOMA)")
    print("="*60)

    # Imposta directory dati
    base_dir = os.getcwd()
    
    ### MODIFICA: Torna indietro di un livello ("..") per trovare la cartella 'data'
    data_dir = os.path.abspath(os.path.join(base_dir, "..", "data"))
    
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        print(f"Creata directory dati: {data_dir}")

    # STEP 1: Trova i file WIG hg19
    print("\n" + "="*60)
    print(f"Ricerca file *_147.hg19.wig in {data_dir}")
    
    wig_search_path = os.path.join(data_dir, "*_147.hg19.wig")
    wig_files = glob.glob(wig_search_path)
    
    if not wig_files:
        print(f"✗ Nessun file WIG 147.hg19.wig trovato in {data_dir}.")
        print("  Assicurati che i file .wig convertiti siano in questa cartella.")
        return
    else:
        print(f"✓ Trovati {len(wig_files)} file WIG hg19:")
        for f in wig_files:
            print(f"  - {os.path.basename(f)}")

    # STEP 2: Scarica/Carica genoma hg19 (TUTTI I CROMOSOMI)
    print("\n" + "="*60)
    genome_dir = os.path.join(data_dir, "hg19_genome") 

    # Controlla solo il primo cromosoma per vedere se la cartella esiste ed è popolata
    check_fa_path = os.path.join(genome_dir, f"{CHROMS_TO_PROCESS[0]}.fa")
    
    if os.path.exists(genome_dir) and os.path.exists(check_fa_path):
        print(f"Genoma hg19 (cartella {genome_dir}) già presente, caricamento...")
        genome_dict = load_hg19_genome(genome_dir) 
    else:
        print("Scaricamento genoma hg19 (tutti i cromosomi standard)...")
        genome_dict = download_hg19_genome(genome_dir) 

    if not genome_dict:
        print("✗ Impossibile caricare genoma hg19")
        return

    # STEP 2.5: Filtro promotori DISABILITATO
    print("\n" + "="*60)
    print("Filtro per promotori DISABILITATO. Si processa l'intero genoma.")
    df_features_to_use = None # <-- CAMBIAMENTO CHIAVE
        
    # STEP 3: Crea dataset
    df_final = create_nucleosome_dataset(
        wig_files,  
        genome_dict,
        max_sequences_per_file=1000000, # Aumenta per file (ora copre l'intero genoma)
        threshold_high=5.0,           
        threshold_low=1.0,             
        max_regions_per_wig=5000000000,#500000000, # Max molto alto per leggere l'intero file
        df_features=df_features_to_use, # Passa None
        sequence_length=201
    )

    if df_final is None or df_final.empty:
        print("✗ Creazione dataset fallita o dataset vuoto.")
        return

    # STEP 4: Salva risultati
    ### MODIFICA: Cambiato nome file di output ###
    output_file = os.path.join(data_dir, "nucleosome_dataset_GSE36979_hg19_whole_genome_201pb_balanced.csv")
    try:
        df_final.to_csv(output_file, index=False)
        print(f"\n✓ Dataset (hg19, Intero Genoma) salvato: {output_file}")
    except Exception as e:
        print(f"✗ Errore salvataggio dataset: {e}")
        return

    # STEP 5: Statistiche finali
    print("\n" + "="*60)
    print("STATISTICHE FINALI (HG19 - INTERO GENOMA)")
    print("="*60)
    print(df_final.head(10))
    
    print(f"\nDistribuzione lunghezze:")
    print(df_final['sequenza'].str.len().describe())

    # GC content
    df_final['gc_content'] = df_final['sequenza'].apply(calc_gc)
    print(f"\nGC content per classe:")
    print(df_final.groupby('nucleosoma')['gc_content'].describe())

    # Plot
    print("\nGenerazione plot GC content...")
    try:
        plt.figure(figsize=(10, 6))
        sns.kdeplot(data=df_final, x='gc_content', hue='nucleosoma', fill=True, common_norm=False, palette="Set1")
        plt.title('Distribuzione GC Content (0=Bassa, 1=Alta Occupazione) - HG19 / Intero Genoma') 
        ### MODIFICA: Cambiato nome file plot ###
        plot_file = os.path.join(data_dir, "gc_content_distribution_hg19_whole_genome.png")
        plt.savefig(plot_file)
        print(f"✓ Plot salvato: {plot_file}")
    except Exception as e:
        print(f"✗ Errore generazione plot: {e}")

    print("\n✓ ELABORAZIONE (HG19 - INTERO GENOMA) COMPLETATA!") 

if __name__ == "__main__":  
    main()
```

```python
import pandas as pd
import numpy as np
import requests
import gzip
import os
from Bio import SeqIO
from io import StringIO
import random
import matplotlib.pyplot as plt
import seaborn as sns
from intervaltree import IntervalTree 
import glob # Per cercare i file

#################################################################################################
# CONFIGURAZIONE GLOBALE
#################################################################################################

# Processa tutti i cromosomi standard
CHROMS_TO_PROCESS = [f"chr{i}" for i in range(1, 23)] + ["chrX", "chrY"]

#################################################################################################
# FUNZIONI HELPER
#################################################################################################

def calc_gc(seq):
    """Calcola il contenuto GC di una sequenza."""
    if not isinstance(seq, str) or len(seq) == 0:
        return 0
    gc = seq.count('G') + seq.count('C')
    return (gc / len(seq)) * 100

#################################################################################################
# STEP 1: GESTIONE GENOMA HG19 (TUTTI I CROMOSOMI)
#################################################################################################

def download_hg19_genome(output_dir="hg19_genome"):
    """
    Scarica il genoma umano hg19 da UCSC (TUTTI I CROMOSOMI STANDARD)
    """
    print(f"Scaricando genoma hg19 da UCSC per {len(CHROMS_TO_PROCESS)} cromosomi...")
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # URL base UCSC per hg19
    base_url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg19/chromosomes/"
    
    ### MODIFICA: Processa tutti i cromosomi ###
    chromosomes = CHROMS_TO_PROCESS
    
    genome_dict = {}
    
    for chrom in chromosomes:
        file_name = f"{chrom}.fa.gz"
        file_path = os.path.join(output_dir, file_name)
        
        # Percorso del file decompresso
        unzipped_file_path = file_path.replace('.gz', '')

        if os.path.exists(unzipped_file_path):
            print(f"✓ {chrom} (hg19) già scaricato")
            # Carica sequenza già presente
            try:
                with open(unzipped_file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        genome_dict[chrom] = str(record.seq).upper() # Salva in maiuscolo
                continue
            except Exception as e:
                print(f"✗ Errore nel caricamento di {unzipped_file_path}: {e}. Tento di riscaricarlo.")

        
        url = base_url + file_name
        print(f"Scaricando {chrom} (hg19)...")
        
        try:
            response = requests.get(url, stream=True, timeout=300)
            if response.status_code == 200:
                # Salva file compresso
                with open(file_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                
                # Decomprimi e leggi
                sequence_content = ""
                with gzip.open(file_path, 'rt') as f:
                    sequence_content = f.read()
                
                # Parse dalla stringa
                with StringIO(sequence_content) as f_in:
                    records = SeqIO.parse(f_in, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Salva in maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} (hg19) scaricato: {len(seq_str):,} bp")
                
                # Salva versione decompressa
                with open(unzipped_file_path, 'w') as f_out:
                    f_out.write(sequence_content)
                
                # Rimuovi file compresso per salvare spazio
                os.remove(file_path)
                
        except Exception as e:
            print(f"✗ Errore scaricamento {chrom} (hg19): {e}")
            if os.path.exists(file_path):
                os.remove(file_path) # Pulisci file corrotto
            continue
    
    print(f"\n✓ Genoma hg19 scaricato: {len(genome_dict)} cromosomi")
    return genome_dict

def load_hg19_genome(genome_dir="hg19_genome"):
    """
    Carica genoma hg19 già scaricato (TUTTI I CROMOSOMI STANDARD)
    """
    print("Caricamento genoma hg19...")
    
    genome_dict = {}
    
    ### MODIFICA: Processa tutti i cromosomi ###
    chromosomes = CHROMS_TO_PROCESS
    
    for chrom in chromosomes:
        file_path = os.path.join(genome_dir, f"{chrom}.fa")
        
        if os.path.exists(file_path):
            try:
                with open(file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Assicurati sia maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} (hg19) caricato: {len(seq_str):,} bp")
            except Exception as e:
                print(f"✗ Errore caricamento {file_path}: {e}")
        else:
            print(f"✗ {chrom} non trovato in {genome_dir}")
    
    if not genome_dict:
        print(f"✗ Nessun file genoma hg19 trovato in {genome_dir}. Prova a scaricarlo prima.")
    else:
        print(f"\n✓ Genoma hg19 caricato: {len(genome_dict)} cromosomi")
    return genome_dict

#################################################################################################
# STEP 2: GESTIONE DATI WIG (TUTTI I CROMOSOMI)
#################################################################################################
def parse_wig_file(wig_file_path, max_regions=500000000):
    """
    Parsifica file WIG con dati di occupancy dei nucleosomi (TUTTI I CROMOSOMI)
    VERSIONE v3 (ROBUSTA): Gestisce fixedStep, variableStep, e bedGraph
    """
    print(f"  Parsing file WIG: {os.path.basename(wig_file_path)}")
    
    regions = []
    current_chrom = None
    current_start = None
    current_step = None
    current_span = 1
    position = 0
    lines_processed = 0
    
    try: 
        opener = gzip.open if wig_file_path.endswith('.gz') else open
        
        with opener(wig_file_path, 'rt') as f:
            for line in f:
                line = line.strip()
                
                # Salta intestazioni e righe vuote
                if not line or line.startswith('track') or line.startswith('browser'):
                    continue
                
                lines_processed += 1
                
                if lines_processed % 50000000 == 0:
                    print(f"    ... Processate {lines_processed:,} linee, trovate {len(regions):,} regioni...")
                
                if len(regions) >= max_regions:
                    print(f"    ⚠ Raggiunto limite di {max_regions:,} regioni, stop parsing")
                    break
                
                # --- GESTIONE HEADER ---
                if line.startswith('fixedStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('start='):
                            current_start = int(part.split('=')[1])
                            position = current_start
                        elif part.startswith('step='):
                            current_step = int(part.split('=')[1])
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                
                elif line.startswith('variableStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                    current_step = None # Flag per variableStep
                
                # --- GESTIONE DATA LINE ---
                else:
                    try:
                        parts = line.split()
                        
                        # --- MODIFICA CHIAVE: Controlla il formato (bedGraph, variableStep, fixedStep) ---
                        
                        if len(parts) >= 4:
                            # Formato bedGraph (chr start end value ...)
                            chrom, start_str, end_str, value_str = parts[0], parts[1], parts[2], parts[3]
                            if chrom not in CHROMS_TO_PROCESS:
                                continue
                            value = float(value_str)
                            if value > 0: # Filtro originale (corretto, se i dati sono >= 1)
                                regions.append({
                                    'chr': chrom,
                                    'start': int(start_str),
                                    'end': int(end_str),
                                    'score': value
                                })
                        
                        elif len(parts) == 2 and current_step is None:
                            # Formato variableStep (pos value)
                            if current_chrom not in CHROMS_TO_PROCESS:
                                continue
                            pos, value_str = parts[0], parts[1]
                            pos = int(pos)
                            value = float(value_str)
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': pos,
                                    'end': pos + current_span,
                                    'score': value
                                })
                        
                        elif len(parts) == 1 and current_step is not None:
                            # Formato fixedStep (value)
                            if current_chrom not in CHROMS_TO_PROCESS:
                                continue
                            value = float(parts[0])
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': position,
                                    'end': position + current_span,
                                    'score': value
                                })
                            position += current_step
                        
                        # Altri formati (es. bedGraph senza header) vengono ignorati

                    except Exception:
                        # Salta righe dati malformate
                        continue
            
        print(f"    ✓ Trovate {len(regions):,} regioni con segnale (Tutti i cromosomi)")
        return pd.DataFrame(regions)
    
    except Exception as e:
        print(f"    ✗ Errore parsing WIG: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

#################################################################################################
# STEP 3: GESTIONE ANNOTAZIONI (Funzioni non più usate, ma definite)
#################################################################################################
# Queste funzioni (load_genomic_features, download_refgene_hg19, etc.) 
# non sono più chiamate da main() ma sono lasciate qui se servissero in futuro.

#################################################################################################
# STEP 4: CREAZIONE DATASET (Funzioni generiche, modificate per "No Filtro")
#################################################################################################

def filter_regions_by_feature(df_regions, df_features):
    """
    Mantiene solo le regioni che si sovrappongono con i feature forniti.
    Se df_features è None, restituisce TUTTE le regioni.
    """
    print("  Filtraggio regioni per sovrapposizione con feature...")

    if df_features is None or df_features.empty:
        print("  ⚠ Nessuna annotazione feature fornita, ritorno tutte le regioni (nessun filtro).")
        return df_regions # <-- Restituisce tutto se non c'è filtro

    # Crea dizionario cromosoma → array di (start, end) per efficienza
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("  Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return pd.DataFrame(columns=df_regions.columns)
        
    keep = np.zeros(len(df_regions), dtype=bool)

    for i, row in enumerate(df_regions.itertuples(index=False)):
        chrom = getattr(row, 'chr', None)
        if chrom not in df_features_by_chr:
            continue
            
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        
        if s is None or e is None:
            continue

        # Cerca sovrapposizione
        for f_start, f_end in df_features_by_chr[chrom]:
            if not (e < f_start or s > f_end):  
                keep[i] = True
                break # Trovata sovrapposizione, passa alla prossima regione

    filtered_df = df_regions[keep]
    print(f"  ✓ {len(filtered_df):,}/{len(df_regions):,} regioni trovate nei feature forniti.")
    return filtered_df.reset_index(drop=True)

def identify_nucleosome_positions(df_wig, threshold_value=1.0, sequence_length=201):
    """
    Identifica le posizioni delle diadi dei nucleosomi (picchi)
    usando la logica 'Non-Maximum Suppression'.
    """
    print("  Identificazione posizioni nucleosomi (Peak Calling)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    df_positive_candidates = df_wig[df_wig['score'] > threshold_value].copy()
    print(f"    Trovati {len(df_positive_candidates):,} candidati positivi (> {threshold_value:.2f})")

    if df_positive_candidates.empty:
        print("    ⚠ Nessun candidato positivo trovato.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    df_sorted = df_positive_candidates.sort_values('score', ascending=False)
    
    true_dyad_rows = []
    claimed_tree = {} # Un albero di intervalli per le regioni già "bloccate"
    
    min_separation = sequence_length  

    print(f"    Avvio Non-Maximum Suppression (min_sep = {min_separation} bp)...")
    
    # Inizializza gli alberi per i cromosomi
    for chrom in df_sorted['chr'].unique():
        if chrom in CHROMS_TO_PROCESS: # Processa solo i nostri cromosomi
            claimed_tree[chrom] = IntervalTree()

    # Itera dal picco più forte in giù
    for row in df_sorted.itertuples(index=False):
        
        # Ignora cromosomi non standard
        if row.chr not in claimed_tree:
            continue
            
        # Controlla se questo picco è già stato "bloccato" da un picco più forte
        if claimed_tree[row.chr].overlaps(row.start, row.end):
            continue # Sì, ignora questo picco
            
        # Se non è bloccato, è un VERO picco!
        true_dyad_rows.append(row)
        
        # Blocca la regione attorno ad esso
        buffer_start = row.start - (min_separation // 2)
        buffer_end = row.start + (min_separation // 2) + 1 
        claimed_tree[row.chr].addi(buffer_start, buffer_end)

    if not true_dyad_rows:
        print("    ⚠ Nessun picco positivo trovato dopo il peak calling.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    print(f"    ✓ Identificati {len(true_dyad_rows):,} picchi unici (diadi)")

    df_true_dyads = pd.DataFrame(true_dyad_rows, columns=df_sorted.columns)

    # Espandi i centri (diadi) alla lunghezza finale della sequenza (201 bp)
    df_true_dyads['center'] = (df_true_dyads['start'] + df_true_dyads['end']) // 2
    df_true_dyads['start'] = df_true_dyads['center'] - (sequence_length // 2)
    df_true_dyads['end'] = df_true_dyads['start'] + sequence_length
    
    return df_true_dyads[['chr', 'start', 'end', 'score']]

def identify_non_nucleosome_positions(df_wig, threshold_value=1.0, nucleosome_length=201):
    """
    Identifica posizioni NON-nucleosomi (BASSA occupazione)
    usando una SOGLIA ASSOLUTA (score == threshold)
    """
    print("  Identificazione posizioni NON-nucleosomi (BASSA occupazione)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    print(f"    Soglia segnale (Negativi): == {threshold_value:.2f}")
    
    # Filtra regioni con segnale BASSO (esattamente il background)
    df_nucleosomes = df_wig[df_wig['score'] == threshold_value].copy()
    
    # Estendi regioni alla lunghezza desiderata
    df_nucleosomes['center'] = (df_nucleosomes['start'] + df_nucleosomes['end']) // 2
    df_nucleosomes['start'] = df_nucleosomes['center'] - nucleosome_length // 2
    df_nucleosomes['end'] = df_nucleosomes['start'] + nucleosome_length
    
    print(f"    ✓ Identificati {len(df_nucleosomes)} NON-nucleosomi (Negativi)")
    
    return df_nucleosomes[['chr', 'start', 'end', 'score']]

# ---------------------------------------------------------------------------------
# MODIFICA 1: La funzione ora restituisce (sequences, chromosomes)
# ---------------------------------------------------------------------------------
def extract_sequences_from_genome(df_regions, genome_dict, max_sequences=2000):
    """
    Estrae sequenze DNA dalle regioni specificate
    """
    print(f"    Estrazione sequenze da {len(df_regions)} regioni...")
    
    sequences = []
    chromosomes = [] # <-- AGGIUNTO
    
    count = 0
    processed = 0
    # Mescola per ottenere un campione rappresentativo se max_sequences è basso
    df_shuffled = df_regions.sample(frac=1)  
    
    for idx, row in df_shuffled.iterrows():
        processed += 1
        if count >= max_sequences:
            print(f"      Raggiunto limite di {max_sequences} sequenze.")
            break
        
        chrom = row['chr']
        start = int(row['start'])
        end = int(row['end'])
        
        if chrom not in genome_dict:
            continue
        
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        sequences.append(seq)
        chromosomes.append(chrom) # <-- AGGIUNTO
        count += 1
        
        if count % 5000 == 0 and count > 0:
            print(f"      Estraendole... {count}/{max_sequences} sequenze valide trovate (processate {processed} regioni)")
    
    print(f"    ✓ Estratte {len(sequences)} sequenze")
    return sequences, chromosomes # <-- MODIFICATO

# ---------------------------------------------------------------------------------
# MODIFICA 2: La funzione ora restituisce (sequences, chromosomes)
# ---------------------------------------------------------------------------------
def extract_filtered_sequences_efficiently(df_regions, df_features, genome_dict, positive_regions_tree, max_sequences=2000, buffer_zone=201):
    """
    Combina filtro e estrazione in un unico passaggio efficiente.
    Mescola i candidati e si ferma non appena 'max_sequences' sono state trovate.
    ESCLUDE i negativi che si sovrappongono ai positivi (buffer_zone).
    Se df_features è None, salta il filtro per feature.
    """
    print(f"  Avvio estrazione efficiente per {max_sequences} sequenze da {len(df_regions):,} candidati...")
    
    sequences = []
    chromosomes = [] # <-- AGGIUNTO
    
    # 1. Crea il dizionario di features per una ricerca rapida
    df_features_by_chr = None
    if df_features is not None and not df_features.empty:
        print("  (Filtro feature ATTIVO)")
        try:
            df_features_by_chr = {
                chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
                for chrom in df_features['chr'].unique()
            }
        except KeyError:
            print("    Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
            return [], [] # <-- MODIFICATO
    else:
        print("  (Filtro feature DISABILITATO)")

    # 2. Mescola i candidati (df_regions)
    df_regions_shuffled = df_regions.sample(frac=1)

    processed_count = 0
    excluded_by_buffer = 0
    
    # 3. Itera, filtra, ed estrai in un unico loop
    for row in df_regions_shuffled.itertuples(index=False):
        processed_count += 1
        
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        chrom = getattr(row, 'chr', None)
        
        if s is None or e is None or chrom is None or chrom not in CHROMS_TO_PROCESS:
            continue
            
        # --- CONTROLLO 1: ZONA CUSCINETTO ---
        buffer_start = s - buffer_zone
        buffer_end = e + buffer_zone
        if chrom in positive_regions_tree and positive_regions_tree[chrom].overlaps(buffer_start, buffer_end):
            excluded_by_buffer += 1
            continue # Questo negativo è troppo vicino a un positivo, scartalo.
        
        # --- CONTROLLO 2: FILTRO FEATURE (es. promotori) ---
        if df_features_by_chr is not None:
            if chrom not in df_features_by_chr:
                continue # Regione in un cromosoma senza feature
                
            has_overlap = False
            for f_start, f_end in df_features_by_chr[chrom]:
                if not (e < f_start or s > f_end):
                    has_overlap = True
                    break
            
            if not has_overlap:
                continue # Non è in un feature, passa al prossimo candidato

        # --- Se siamo qui, la regione E' valida (o non filtrata) E NON è vicina a un positivo ---
        
        # Logica di 'extract_sequences_from_genome'
        start = int(s)
        end = int(e)
        
        if chrom not in genome_dict:
            continue
            
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        # --- Se siamo qui, la sequenza è VALIDA ---
        sequences.append(seq)
        chromosomes.append(chrom) # <-- AGGIUNTO
        
        # Logica di 'stop'
        if len(sequences) >= max_sequences:
            print(f"    ✓ Raggiunto limite di {max_sequences} sequenze valide (processati {processed_count:,} candidati).")
            print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")
            break
            
    if len(sequences) < max_sequences:
        print(f"    ✓ Estrazione completata. Trovate {len(sequences)} sequenze valide (processati tutti i {len(df_regions):,} candidati).")
        print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")

    return sequences, chromosomes # <-- MODIFICATO


# ---------------------------------------------------------------------------------
# MODIFICA 3: La funzione ora raccoglie i 'chr' e li aggiunge al DataFrame
# ---------------------------------------------------------------------------------
def create_nucleosome_dataset(wig_files, genome_dict,  
                              max_sequences_per_file=1000,
                              threshold_high=1.0,
                              threshold_low=1.0,
                              max_regions_per_wig=100000,
                              df_features=None,
                              sequence_length=201):
    """
    Crea dataset finale combinando dati WIG e genoma.
    Positivi: score > threshold_high
    Negativi: score == threshold_low
    """
    print("\n" + "="*60)
    print("CREAZIONE DATASET NUCLEOSOMI (Positivi vs Negativi in-loco)")
    print("="*60)
    print(f"Lunghezza sequenza: {sequence_length} bp")
    print(f"Soglia ALTA (positivi): > {threshold_high:.2f}")
    print(f"Soglia BASSA (negativi): == {threshold_low:.2f}")
    if df_features is not None:
        print(f"Filtro regioni: ATTIVO (usando {len(df_features)} feature)")
    else:
        print("Filtro regioni: DISABILITATO (intero genoma)")
    
    all_sequences = []
    all_labels = []
    all_chromosomes = [] # <-- AGGIUNTO
    
    # Processa TUTTI i file WIG
    for idx, wig_file in enumerate(wig_files, 1):
        print(f"\n[{idx}/{len(wig_files)}] Processando: {os.path.basename(wig_file)}")
        
        df_wig = parse_wig_file(wig_file, max_regions=max_regions_per_wig)
        
        if df_wig is None or len(df_wig) == 0:
            print("  ⚠ Nessun dato trovato, skip")
            continue
            
        # -----------------------------------------------------------------
        # 1. STREAM POSITIVO (Label 1)
        # -----------------------------------------------------------------
        print("\n  --- Stream Positivo (Label 1) ---")
        # 1a. Trova TUTTI i positivi (>1.0) nell'INTERO GENOMA (con Peak Calling)
        df_nucleosomes_high = identify_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_high,  
            sequence_length=sequence_length
        )
        
        # 1b. Filtra i positivi SOLO per i feature (se forniti)
        df_positive_filtered = filter_regions_by_feature(df_nucleosomes_high, df_features)

        if df_positive_filtered.empty:
            print("  ⚠ Nessuna regione positiva (>1.0) trovata [dopo il filtro], skip file")
            continue
            
        # 1c. Estrai sequenze
        #     MODIFICA: Ora riceve anche chroms_pos
        sequences_pos, chroms_pos = extract_sequences_from_genome(
            df_positive_filtered,  
            genome_dict,  
            max_sequences=max_sequences_per_file  
        )
        
        # --- Crea l'albero dei positivi per la zona cuscinetto ---
        # Usa la lista TOTALE (df_nucleosomes_high) per l'albero
        print("    Creazione albero intervalli positivi per zona cuscinetto...")
        positive_tree = {}
        for chrom in df_nucleosomes_high['chr'].unique():
            if chrom in CHROMS_TO_PROCESS: # Assicurati
                positive_tree[chrom] = IntervalTree()
                for row in df_nucleosomes_high[df_nucleosomes_high['chr'] == chrom].itertuples():
                    positive_tree[chrom].addi(row.start, row.end)
        # ---------------------------------------------------------------------

        # -----------------------------------------------------------------
        # 2. STREAM NEGATIVO (Label 0) - Logica efficiente
        # -----------------------------------------------------------------
        print("\n  --- Stream Negativo (Label 0) ---")
        
        n_to_find = len(sequences_pos) # Numero di negativi da trovare
        if n_to_find == 0:
             print("  ⚠ Nessuna sequenza positiva trovata, skip negativi")
             continue
        
        # 2a. Trova TUTTI i candidati negativi (==1.0)
        df_nucleosomes_low = identify_non_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_low,
            nucleosome_length=sequence_length
        )
        
        if df_nucleosomes_low.empty:
            print("  ⚠ Nessuna regione negativa (==1.0) trovata, skip negativi")
            continue
        
        # 2b. --- CHIAMATA ALLA NUOVA FUNZIONE EFFICIENTE (CON ALBERO) ---
        #     MODIFICA: Ora riceve anche chroms_neg
        sequences_neg, chroms_neg = extract_filtered_sequences_efficiently(
            df_nucleosomes_low,     # I milioni di candidati
            df_features,            # I promotori (o None)
            genome_dict,
            positive_tree,          # L'albero dei positivi da evitare
            max_sequences=n_to_find,  # Il numero da trovare
            buffer_zone=sequence_length 
        )
        
        if not sequences_neg:
             print("  ⚠ Nessuna sequenza negativa valida trovata (o tutte troppo vicine ai positivi).")
             continue
        
        # -----------------------------------------------------------------
        # 3. BILANCIAMENTO e AGGIUNTA AL DATASET
        # -----------------------------------------------------------------
        
        n_pos = len(sequences_pos)
        n_neg = len(sequences_neg)

        n_to_keep = min(n_pos, n_neg)
        
        # MODIFICA: Zippa, campiona e unzippa per mantenere l'allineamento
        if n_pos > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_pos} positivi a {n_to_keep}")
            # Combina seq e chr, campiona, e separa di nuovo
            data_pos = list(zip(sequences_pos, chroms_pos))
            sampled_data_pos = random.sample(data_pos, n_to_keep)
            sequences_pos_sampled, chroms_pos_sampled = zip(*sampled_data_pos)
            # Riconverti in lista
            sequences_pos = list(sequences_pos_sampled)
            chroms_pos = list(chroms_pos_sampled)
        
        if n_neg > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_neg} negativi a {n_to_keep}")
            # Combina seq e chr, campiona, e separa di nuovo
            data_neg = list(zip(sequences_neg, chroms_neg))
            sampled_data_neg = random.sample(data_neg, n_to_keep)
            sequences_neg_sampled, chroms_neg_sampled = zip(*sampled_data_neg)
            # Riconverti in lista
            sequences_neg = list(sequences_neg_sampled)
            chroms_neg = list(chroms_neg_sampled)

        # Aggiungi i dati positivi
        all_sequences.extend(sequences_pos)
        all_labels.extend([1] * len(sequences_pos))
        all_chromosomes.extend(chroms_pos) # <-- AGGIUNTO
        
        # Aggiungi i dati negativi
        all_sequences.extend(sequences_neg)
        all_labels.extend([0] * len(sequences_neg))
        all_chromosomes.extend(chroms_neg) # <-- AGGIUNTO
        
        print(f"\n  RISULTATO FILE: Aggiunti {len(sequences_pos)} positivi e {len(sequences_neg)} negativi (Bilanciati).")
        print(f"  ✓ Totale sequenze accumulate: {len(all_sequences):,}")

    
    if not all_sequences:
        print("⚠⚠ NESSUNA SEQUENZA TROVATA. Controlla i percorsi e i filtri.")
        # MODIFICA: Aggiungi 'chr' alle colonne
        return pd.DataFrame(columns=['sequenza', 'nucleosoma', 'chr'])

    # Crea DataFrame
    # MODIFICA: Aggiungi 'chr' al dizionario
    df = pd.DataFrame({
        'sequenza': all_sequences,
        'nucleosoma': all_labels,
        'chr': all_chromosomes 
    })
    
    # Rimuovi duplicati basati sulla sequenza (mantiene chr e label del primo)
    df = df.drop_duplicates(subset=['sequenza']).reset_index(drop=True)
    # Mescola il dataset finale
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    
    print("\n" + "="*60)
    print("DATASET COMPLETATO (Bilanciato per regione)")
    print("="*60)
    print(f"Totale sequenze: {len(df):,}")
    print(f"Nucleosomi (positivi): {df['nucleosoma'].sum():,}")
    print(f"Non-nucleosomi (negativi): {(1 - df['nucleosoma']).sum():,}")
    
    if len(df) > 0:
        print(f"Bilanciamento: {df['nucleosoma'].sum() / len(df) * 100:.1f}% positivi")
    
    return df

#################################################################################################
# ESECUZIONE PRINCIPALE (Modificata per hg19 / INTERO GENOMA)
#################################################################################################

def main():
    print("="*60)
    print("ELABORAZIONE DATI NUCLEOSOMI GSE36979 (HG19 - INTERO GENOMA)")
    print("="*60)

    # Imposta directory dati
    base_dir = os.getcwd()
    
    ### MODIFICA: Torna indietro di un livello ("..") per trovare la cartella 'data'
    data_dir = os.path.abspath(os.path.join(base_dir, "..", "data", "cd4t_data"))
    
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        print(f"Creata directory dati: {data_dir}")

    # STEP 1: Trova i file WIG hg19
    print("\n" + "="*60)
    print(f"Ricerca file *_147.hg19.wig in {data_dir}")
    
    wig_search_path = os.path.join(data_dir, "ActivatedNucleosomes-chr*.hg19.wig")#"*_147.hg19.wig")
    wig_files = glob.glob(wig_search_path)
    
    if not wig_files:
        print(f"✗ Nessun file WIG 147.hg19.wig trovato in {data_dir}.")
        print("  Assicurati che i file .wig convertiti siano in questa cartella.")
        return
    else:
        print(f"✓ Trovati {len(wig_files)} file WIG hg19:")
        for f in wig_files:
            print(f"  - {os.path.basename(f)}")

    # STEP 2: Scarica/Carica genoma hg19 (TUTTI I CROMOSOMI)
    print("\n" + "="*60)
    genome_dir = os.path.join(data_dir, "hg19_genome") 

    # Controlla solo il primo cromosoma per vedere se la cartella esiste ed è popolata
    check_fa_path = os.path.join(genome_dir, f"{CHROMS_TO_PROCESS[0]}.fa")
    
    if os.path.exists(genome_dir) and os.path.exists(check_fa_path):
        print(f"Genoma hg19 (cartella {genome_dir}) già presente, caricamento...")
        genome_dict = load_hg19_genome(genome_dir) 
    else:
        print("Scaricamento genoma hg19 (tutti i cromosomi standard)...")
        genome_dict = download_hg19_genome(genome_dir) 

    if not genome_dict:
        print("✗ Impossibile caricare genoma hg19")
        return

    # STEP 2.5: Filtro promotori DISABILITATO
    print("\n" + "="*60)
    print("Filtro per promotori DISABILITATO. Si processa l'intero genoma.")
    df_features_to_use = None # <-- CAMBIAMENTO CHIAVE
        
    # STEP 3: Crea dataset
    df_final = create_nucleosome_dataset(
        wig_files,  
        genome_dict,
        max_sequences_per_file=10000, # Aumenta per file (ora copre l'intero genoma)
        threshold_high=30.0,  #5  se uso train      
        threshold_low=1.0,           
        max_regions_per_wig=5000000000,#500000000, # Max molto alto per leggere l'intero file
        df_features=df_features_to_use, # Passa None
        sequence_length=201
    )

    if df_final is None or df_final.empty:
        print("✗ Creazione dataset fallita o dataset vuoto.")
        return

    # STEP 4: Salva risultati
    ### MODIFICA: Cambiato nome file di output ###
    output_file = os.path.join(data_dir, "CD4T_h19_Act_tot.csv")#"nucleosome_dataset_GSE36979_hg19_whole_genome_201pb_balanced.csv")
    try:
        df_final.to_csv(output_file, index=False)
        print(f"\n✓ Dataset (hg19, Intero Genoma) salvato: {output_file}")
    except Exception as e:
        print(f"✗ Errore salvataggio dataset: {e}")
        return

    # STEP 5: Statistiche finali
    print("\n" + "="*60)
    print("STATISTICHE FINALI (HG19 - INTERO GENOMA)")
    print("="*60)
    # Stampa l'head includendo la nuova colonna 'chr'
    print(df_final.head(10))
    
    print(f"\nDistribuzione lunghezze:")
    print(df_final['sequenza'].str.len().describe())

    # GC content
    df_final['gc_content'] = df_final['sequenza'].apply(calc_gc)
    print(f"\nGC content per classe:")
    print(df_final.groupby('nucleosoma')['gc_content'].describe())

    # Plot
    print("\nGenerazione plot GC content...")
    try:
        plt.figure(figsize=(10, 6))
        sns.kdeplot(data=df_final, x='gc_content', hue='nucleosoma', fill=True, common_norm=False, palette="Set1")
        plt.title('Distribuzione GC Content (0=Bassa, 1=Alta Occupazione) - HG19 / Intero Genoma') 
        ### MODIFICA: Cambiato nome file plot ###
        plot_file = os.path.join(data_dir, "gc_content_distribution_hg19_whole_genome.png")
        plt.savefig(plot_file)
        print(f"✓ Plot salvato: {plot_file}")
    except Exception as e:
        print(f"✗ Errore generazione plot: {e}")

    print("\n✓ ELABORAZIONE (HG19 - INTERO GENOMA) COMPLETATA!") 

if __name__ == "__main__":  
    main()
```

```python
#PROVO A USARE SOLO QUESTA CELLA PER TUTTI I DATASET

#cella uguale a quella sopra ma uso percentile 99.5 invece di 30
import pandas as pd
import numpy as np
import requests
import gzip
import os
from Bio import SeqIO
from io import StringIO
import random
import matplotlib.pyplot as plt
import seaborn as sns
from intervaltree import IntervalTree 
import glob # Per cercare i file

#################################################################################################
# CONFIGURAZIONE GLOBALE
#################################################################################################

# Processa tutti i cromosomi standard
CHROMS_TO_PROCESS = [f"chr{i}" for i in range(1, 23)] + ["chrX", "chrY"]

#################################################################################################
# FUNZIONI HELPER
#################################################################################################

def calc_gc(seq):
    """Calcola il contenuto GC di una sequenza."""
    if not isinstance(seq, str) or len(seq) == 0:
        return 0
    gc = seq.count('G') + seq.count('C')
    return (gc / len(seq)) * 100

#################################################################################################
# STEP 1: GESTIONE GENOMA HG19 (TUTTI I CROMOSOMI)
#################################################################################################

def download_hg19_genome(output_dir="hg19_genome"):
    """
    Scarica il genoma umano hg19 da UCSC (TUTTI I CROMOSOMI STANDARD)
    """
    print(f"Scaricando genoma hg19 da UCSC per {len(CHROMS_TO_PROCESS)} cromosomi...")
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # URL base UCSC per hg19
    base_url = "http://hgdownload.cse.ucsc.edu/goldenPath/hg19/chromosomes/"
    
    ### MODIFICA: Processa tutti i cromosomi ###
    chromosomes = CHROMS_TO_PROCESS
    
    genome_dict = {}
    
    for chrom in chromosomes:
        file_name = f"{chrom}.fa.gz"
        file_path = os.path.join(output_dir, file_name)
        
        # Percorso del file decompresso
        unzipped_file_path = file_path.replace('.gz', '')

        if os.path.exists(unzipped_file_path):
            print(f"✓ {chrom} (hg19) già scaricato")
            # Carica sequenza già presente
            try:
                with open(unzipped_file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        genome_dict[chrom] = str(record.seq).upper() # Salva in maiuscolo
                continue
            except Exception as e:
                print(f"✗ Errore nel caricamento di {unzipped_file_path}: {e}. Tento di riscaricarlo.")

        
        url = base_url + file_name
        print(f"Scaricando {chrom} (hg19)...")
        
        try:
            response = requests.get(url, stream=True, timeout=300)
            if response.status_code == 200:
                # Salva file compresso
                with open(file_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                
                # Decomprimi e leggi
                sequence_content = ""
                with gzip.open(file_path, 'rt') as f:
                    sequence_content = f.read()
                
                # Parse dalla stringa
                with StringIO(sequence_content) as f_in:
                    records = SeqIO.parse(f_in, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Salva in maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} (hg19) scaricato: {len(seq_str):,} bp")
                
                # Salva versione decompressa
                with open(unzipped_file_path, 'w') as f_out:
                    f_out.write(sequence_content)
                
                # Rimuovi file compresso per salvare spazio
                os.remove(file_path)
                
        except Exception as e:
            print(f"✗ Errore scaricamento {chrom} (hg19): {e}")
            if os.path.exists(file_path):
                os.remove(file_path) # Pulisci file corrotto
            continue
    
    print(f"\n✓ Genoma hg19 scaricato: {len(genome_dict)} cromosomi")
    return genome_dict

def load_hg19_genome(genome_dir="hg19_genome"):
    """
    Carica genoma hg19 già scaricato (TUTTI I CROMOSOMI STANDARD)
    """
    print("Caricamento genoma hg19...")
    
    genome_dict = {}
    
    ### MODIFICA: Processa tutti i cromosomi ###
    chromosomes = CHROMS_TO_PROCESS
    
    for chrom in chromosomes:
        file_path = os.path.join(genome_dir, f"{chrom}.fa")
        
        if os.path.exists(file_path):
            try:
                with open(file_path, 'r') as f:
                    records = SeqIO.parse(f, "fasta")
                    for record in records:
                        seq_str = str(record.seq).upper() # Assicurati sia maiuscolo
                        genome_dict[chrom] = seq_str
                        print(f"✓ {chrom} (hg19) caricato: {len(seq_str):,} bp")
            except Exception as e:
                print(f"✗ Errore caricamento {file_path}: {e}")
        else:
            print(f"✗ {chrom} non trovato in {genome_dir}")
    
    if not genome_dict:
        print(f"✗ Nessun file genoma hg19 trovato in {genome_dir}. Prova a scaricarlo prima.")
    else:
        print(f"\n✓ Genoma hg19 caricato: {len(genome_dict)} cromosomi")
    return genome_dict

#################################################################################################
# STEP 2: GESTIONE DATI WIG (TUTTI I CROMOSOMI)
#################################################################################################
def parse_wig_file(wig_file_path, max_regions=500000000):
    """
    Parsifica file WIG con dati di occupancy dei nucleosomi (TUTTI I CROMOSOMI)
    VERSIONE v3 (ROBUSTA): Gestisce fixedStep, variableStep, e bedGraph
    """
    print(f"  Parsing file WIG: {os.path.basename(wig_file_path)}")
    
    regions = []
    current_chrom = None
    current_start = None
    current_step = None
    current_span = 1
    position = 0
    lines_processed = 0
    
    try: 
        opener = gzip.open if wig_file_path.endswith('.gz') else open
        
        with opener(wig_file_path, 'rt') as f:
            for line in f:
                line = line.strip()
                
                # Salta intestazioni e righe vuote
                if not line or line.startswith('track') or line.startswith('browser'):
                    continue
                
                lines_processed += 1
                
                if lines_processed % 50000000 == 0:
                    print(f"    ... Processate {lines_processed:,} linee, trovate {len(regions):,} regioni...")
                
                if len(regions) >= max_regions:
                    print(f"    ⚠ Raggiunto limite di {max_regions:,} regioni, stop parsing")
                    break
                
                # --- GESTIONE HEADER ---
                if line.startswith('fixedStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('start='):
                            current_start = int(part.split('=')[1])
                            position = current_start
                        elif part.startswith('step='):
                            current_step = int(part.split('=')[1])
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                
                elif line.startswith('variableStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                    current_step = None # Flag per variableStep
                
                # --- GESTIONE DATA LINE ---
                else:
                    try:
                        parts = line.split()
                        
                        # --- MODIFICA CHIAVE: Controlla il formato (bedGraph, variableStep, fixedStep) ---
                        
                        if len(parts) >= 4:
                            # Formato bedGraph (chr start end value ...)
                            chrom, start_str, end_str, value_str = parts[0], parts[1], parts[2], parts[3]
                            if chrom not in CHROMS_TO_PROCESS:
                                continue
                            value = float(value_str)
                            if value > 0: # Filtro originale (corretto, se i dati sono >= 1)
                                regions.append({
                                    'chr': chrom,
                                    'start': int(start_str),
                                    'end': int(end_str),
                                    'score': value
                                })
                        
                        elif len(parts) == 2 and current_step is None:
                            # Formato variableStep (pos value)
                            if current_chrom not in CHROMS_TO_PROCESS:
                                continue
                            pos, value_str = parts[0], parts[1]
                            pos = int(pos)
                            value = float(value_str)
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': pos,
                                    'end': pos + current_span,
                                    'score': value
                                })
                        
                        elif len(parts) == 1 and current_step is not None:
                            # Formato fixedStep (value)
                            if current_chrom not in CHROMS_TO_PROCESS:
                                continue
                            value = float(parts[0])
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': position,
                                    'end': position + current_span,
                                    'score': value
                                })
                            position += current_step
                        
                        # Altri formati (es. bedGraph senza header) vengono ignorati

                    except Exception:
                        # Salta righe dati malformate
                        continue
            
        print(f"    ✓ Trovate {len(regions):,} regioni con segnale (Tutti i cromosomi)")
        return pd.DataFrame(regions)
    
    except Exception as e:
        print(f"    ✗ Errore parsing WIG: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

#################################################################################################
# STEP 3: GESTIONE ANNOTAZIONI (Funzioni non più usate, ma definite)
#################################################################################################

def filter_regions_by_feature(df_regions, df_features):
    """
    Mantiene solo le regioni che si sovrappongono con i feature forniti.
    Se df_features è None, restituisce TUTTE le regioni.
    """
    print("  Filtraggio regioni per sovrapposizione con feature...")

    if df_features is None or df_features.empty:
        print("  ⚠ Nessuna annotazione feature fornita, ritorno tutte le regioni (nessun filtro).")
        return df_regions # <-- Restituisce tutto se non c'è filtro

    # Crea dizionario cromosoma → array di (start, end) per efficienza
    try:
        df_features_by_chr = {
            chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
            for chrom in df_features['chr'].unique()
        }
    except KeyError:
        print("  Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
        return pd.DataFrame(columns=df_regions.columns)
        
    keep = np.zeros(len(df_regions), dtype=bool)

    for i, row in enumerate(df_regions.itertuples(index=False)):
        chrom = getattr(row, 'chr', None)
        if chrom not in df_features_by_chr:
            continue
            
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        
        if s is None or e is None:
            continue

        # Cerca sovrapposizione
        for f_start, f_end in df_features_by_chr[chrom]:
            if not (e < f_start or s > f_end):  
                keep[i] = True
                break # Trovata sovrapposizione, passa alla prossima regione

    filtered_df = df_regions[keep]
    print(f"  ✓ {len(filtered_df):,}/{len(df_regions):,} regioni trovate nei feature forniti.")
    return filtered_df.reset_index(drop=True)

def identify_nucleosome_positions(df_wig, threshold_value=1.0, sequence_length=201):
    """
    Identifica le posizioni delle diadi dei nucleosomi (picchi)
    usando la logica 'Non-Maximum Suppression'.
    """
    print("  Identificazione posizioni nucleosomi (Peak Calling)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    df_positive_candidates = df_wig[df_wig['score'] > threshold_value].copy()
    print(f"    Trovati {len(df_positive_candidates):,} candidati positivi (> {threshold_value:.2f})")

    if df_positive_candidates.empty:
        print("    ⚠ Nessun candidato positivo trovato.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    df_sorted = df_positive_candidates.sort_values('score', ascending=False)
    
    true_dyad_rows = []
    claimed_tree = {} # Un albero di intervalli per le regioni già "bloccate"
    
    min_separation = sequence_length  

    print(f"    Avvio Non-Maximum Suppression (min_sep = {min_separation} bp)...")
    
    # Inizializza gli alberi per i cromosomi
    for chrom in df_sorted['chr'].unique():
        if chrom in CHROMS_TO_PROCESS: # Processa solo i nostri cromosomi
            claimed_tree[chrom] = IntervalTree()

    # Itera dal picco più forte in giù
    for row in df_sorted.itertuples(index=False):
        
        # Ignora cromosomi non standard
        if row.chr not in claimed_tree:
            continue
            
        # Controlla se questo picco è già stato "bloccato" da un picco più forte
        if claimed_tree[row.chr].overlaps(row.start, row.end):
            continue # Sì, ignora questo picco
            
        # Se non è bloccato, è un VERO picco!
        true_dyad_rows.append(row)
        
        # Blocca la regione attorno ad esso
        buffer_start = row.start - (min_separation // 2)
        buffer_end = row.start + (min_separation // 2) + 1 
        claimed_tree[row.chr].addi(buffer_start, buffer_end)

    if not true_dyad_rows:
        print("    ⚠ Nessun picco positivo trovato dopo il peak calling.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
        
    print(f"    ✓ Identificati {len(true_dyad_rows):,} picchi unici (diadi)")

    df_true_dyads = pd.DataFrame(true_dyad_rows, columns=df_sorted.columns)

    # Espandi i centri (diadi) alla lunghezza finale della sequenza (201 bp)
    df_true_dyads['center'] = (df_true_dyads['start'] + df_true_dyads['end']) // 2
    df_true_dyads['start'] = df_true_dyads['center'] - (sequence_length // 2)
    df_true_dyads['end'] = df_true_dyads['start'] + sequence_length
    
    return df_true_dyads[['chr', 'start', 'end', 'score']]

def identify_non_nucleosome_positions(df_wig, threshold_value=1.0, nucleosome_length=201):
    """
    Identifica posizioni NON-nucleosomi (BASSA occupazione)
    usando una SOGLIA ASSOLUTA (score == threshold)
    """
    print("  Identificazione posizioni NON-nucleosomi (BASSA occupazione)...")
    
    if df_wig.empty or 'score' not in df_wig.columns:
        print("    ⚠ DataFrame WIG vuoto o 'score' mancante.")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

    print(f"    Soglia segnale (Negativi): == {threshold_value:.2f}")
    
    # Filtra regioni con segnale BASSO (esattamente il background)
    df_nucleosomes = df_wig[df_wig['score'] == threshold_value].copy()
    
    # Estendi regioni alla lunghezza desiderata
    df_nucleosomes['center'] = (df_nucleosomes['start'] + df_nucleosomes['end']) // 2
    df_nucleosomes['start'] = df_nucleosomes['center'] - nucleosome_length // 2
    df_nucleosomes['end'] = df_nucleosomes['start'] + nucleosome_length
    
    print(f"    ✓ Identificati {len(df_nucleosomes)} NON-nucleosomi (Negativi)")
    
    return df_nucleosomes[['chr', 'start', 'end', 'score']]

def extract_sequences_from_genome(df_regions, genome_dict, max_sequences=2000):
    """
    Estrae sequenze DNA dalle regioni specificate
    """
    print(f"    Estrazione sequenze da {len(df_regions)} regioni...")
    
    sequences = []
    chromosomes = [] 
    
    count = 0
    processed = 0
    # Mescola per ottenere un campione rappresentativo se max_sequences è basso
    df_shuffled = df_regions.sample(frac=1)  
    
    for idx, row in df_shuffled.iterrows():
        processed += 1
        if count >= max_sequences:
            print(f"      Raggiunto limite di {max_sequences} sequenze.")
            break
        
        chrom = row['chr']
        start = int(row['start'])
        end = int(row['end'])
        
        if chrom not in genome_dict:
            continue
        
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        sequences.append(seq)
        chromosomes.append(chrom) 
        count += 1
        
        if count % 5000 == 0 and count > 0:
            print(f"      Estraendole... {count}/{max_sequences} sequenze valide trovate (processate {processed} regioni)")
    
    print(f"    ✓ Estratte {len(sequences)} sequenze")
    return sequences, chromosomes 

def extract_filtered_sequences_efficiently(df_regions, df_features, genome_dict, positive_regions_tree, max_sequences=2000, buffer_zone=201):
    """
    Combina filtro e estrazione in un unico passaggio efficiente.
    Mescola i candidati e si ferma non appena 'max_sequences' sono state trovate.
    ESCLUDE i negativi che si sovrappongono ai positivi (buffer_zone).
    """
    print(f"  Avvio estrazione efficiente per {max_sequences} sequenze da {len(df_regions):,} candidati...")
    
    sequences = []
    chromosomes = [] 
    
    # 1. Crea il dizionario di features per una ricerca rapida
    df_features_by_chr = None
    if df_features is not None and not df_features.empty:
        print("  (Filtro feature ATTIVO)")
        try:
            df_features_by_chr = {
                chrom: df_features[df_features['chr'] == chrom][['start', 'end']].values
                for chrom in df_features['chr'].unique()
            }
        except KeyError:
            print("    Errore: df_features deve contenere le colonne 'chr', 'start', 'end'.")
            return [], []
    else:
        print("  (Filtro feature DISABILITATO)")

    # 2. Mescola i candidati (df_regions)
    df_regions_shuffled = df_regions.sample(frac=1)

    processed_count = 0
    excluded_by_buffer = 0
    
    # 3. Itera, filtra, ed estrai in un unico loop
    for row in df_regions_shuffled.itertuples(index=False):
        processed_count += 1
        
        s = getattr(row, 'start', None)
        e = getattr(row, 'end', None)
        chrom = getattr(row, 'chr', None)
        
        if s is None or e is None or chrom is None or chrom not in CHROMS_TO_PROCESS:
            continue
            
        # --- CONTROLLO 1: ZONA CUSCINETTO ---
        buffer_start = s - buffer_zone
        buffer_end = e + buffer_zone
        if chrom in positive_regions_tree and positive_regions_tree[chrom].overlaps(buffer_start, buffer_end):
            excluded_by_buffer += 1
            continue # Questo negativo è troppo vicino a un positivo, scartalo.
        
        # --- CONTROLLO 2: FILTRO FEATURE (es. promotori) ---
        if df_features_by_chr is not None:
            if chrom not in df_features_by_chr:
                continue # Regione in un cromosoma senza feature
                
            has_overlap = False
            for f_start, f_end in df_features_by_chr[chrom]:
                if not (e < f_start or s > f_end):
                    has_overlap = True
                    break
            
            if not has_overlap:
                continue # Non è in un feature, passa al prossimo candidato

        # --- Se siamo qui, la regione E' valida (o non filtrata) E NON è vicina a un positivo ---
        
        # Logica di 'extract_sequences_from_genome'
        start = int(s)
        end = int(e)
        
        if chrom not in genome_dict:
            continue
            
        if start < 0 or end > len(genome_dict[chrom]):
            continue
        
        seq = genome_dict[chrom][start:end]
        
        if len(seq) != (end - start):
            continue
        
        if 'N' in seq:
            continue

        # --- Se siamo qui, la sequenza è VALIDA ---
        sequences.append(seq)
        chromosomes.append(chrom) 
        
        # Logica di 'stop'
        if len(sequences) >= max_sequences:
            print(f"    ✓ Raggiunto limite di {max_sequences} sequenze valide (processati {processed_count:,} candidati).")
            print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")
            break
            
    if len(sequences) < max_sequences:
        print(f"    ✓ Estrazione completata. Trovate {len(sequences)} sequenze valide (processati tutti i {len(df_regions):,} candidati).")
        print(f"    (Scartati {excluded_by_buffer:,} candidati per vicinanza ai positivi)")

    return sequences, chromosomes 


# ---------------------------------------------------------------------------------
# MODIFICA 3: Parametro cambiato da threshold_high a percentile_high
# ---------------------------------------------------------------------------------
def create_nucleosome_dataset(wig_files, genome_dict,  
                              max_sequences_per_file=1000,
                              percentile_high=99.5, # <--- NUOVO PARAMETRO (al posto di threshold_high)
                              threshold_low=1.0,
                              max_regions_per_wig=100000,
                              df_features=None,
                              sequence_length=201):
    """
    Crea dataset finale combinando dati WIG e genoma.
    Positivi: score > percentile_high (calcolato sui dati)
    Negativi: score == threshold_low
    """
    print("\n" + "="*60)
    print("CREAZIONE DATASET NUCLEOSOMI (Positivi vs Negativi in-loco)")
    print("="*60)
    print(f"Lunghezza sequenza: {sequence_length} bp")
    print(f"Soglia ALTA (percentile): {percentile_high}%") # <--- Log aggiornato
    print(f"Soglia BASSA (valore fisso): == {threshold_low:.2f}")
    if df_features is not None:
        print(f"Filtro regioni: ATTIVO (usando {len(df_features)} feature)")
    else:
        print("Filtro regioni: DISABILITATO (intero genoma)")
    
    all_sequences = []
    all_labels = []
    all_chromosomes = [] 
    
    # Processa TUTTI i file WIG
    for idx, wig_file in enumerate(wig_files, 1):
        print(f"\n[{idx}/{len(wig_files)}] Processando: {os.path.basename(wig_file)}")
        
        df_wig = parse_wig_file(wig_file, max_regions=max_regions_per_wig)
        
        if df_wig is None or len(df_wig) == 0:
            print("  ⚠ Nessun dato trovato, skip")
            continue
            
        # -----------------------------------------------------------------
        # 1. STREAM POSITIVO (Label 1)
        # -----------------------------------------------------------------
        print("\n  --- Stream Positivo (Label 1) ---")
        
        # Calcolo della soglia effettiva basata sul percentile
        calculated_threshold_high = 0
        if not df_wig.empty and 'score' in df_wig.columns:
            calculated_threshold_high = np.percentile(df_wig['score'], percentile_high)
            print(f"  ★ SOGLIA CALCOLATA ({percentile_high}%): {calculated_threshold_high:.4f}")
        else:
            print(f"  ⚠ Impossibile calcolare percentile.")
            continue

        # 1a. Trova TUTTI i positivi usando la soglia calcolata
        df_nucleosomes_high = identify_nucleosome_positions(
            df_wig,  
            threshold_value=calculated_threshold_high, # <-- Passa il valore calcolato
            sequence_length=sequence_length
        )
        
        # 1b. Filtra i positivi SOLO per i feature (se forniti)
        df_positive_filtered = filter_regions_by_feature(df_nucleosomes_high, df_features)

        if df_positive_filtered.empty:
            print("  ⚠ Nessuna regione positiva trovata [dopo il filtro], skip file")
            continue
            
        # 1c. Estrai sequenze
        sequences_pos, chroms_pos = extract_sequences_from_genome(
            df_positive_filtered,  
            genome_dict,  
            max_sequences=max_sequences_per_file  
        )
        
        # --- Crea l'albero dei positivi per la zona cuscinetto ---
        print("    Creazione albero intervalli positivi per zona cuscinetto...")
        positive_tree = {}
        for chrom in df_nucleosomes_high['chr'].unique():
            if chrom in CHROMS_TO_PROCESS: 
                positive_tree[chrom] = IntervalTree()
                for row in df_nucleosomes_high[df_nucleosomes_high['chr'] == chrom].itertuples():
                    positive_tree[chrom].addi(row.start, row.end)

        # -----------------------------------------------------------------
        # 2. STREAM NEGATIVO (Label 0)
        # -----------------------------------------------------------------
        print("\n  --- Stream Negativo (Label 0) ---")
        
        n_to_find = len(sequences_pos) # Numero di negativi da trovare
        if n_to_find == 0:
             print("  ⚠ Nessuna sequenza positiva trovata, skip negativi")
             continue
        
        # 2a. Trova TUTTI i candidati negativi (==1.0)
        df_nucleosomes_low = identify_non_nucleosome_positions(
            df_wig,  
            threshold_value=threshold_low,
            nucleosome_length=sequence_length
        )
        
        if df_nucleosomes_low.empty:
            print("  ⚠ Nessuna regione negativa (==1.0) trovata, skip negativi")
            continue
        
        # 2b. Estrazione negativi con filtro e albero
        sequences_neg, chroms_neg = extract_filtered_sequences_efficiently(
            df_nucleosomes_low,     
            df_features,            
            genome_dict,
            positive_tree,          
            max_sequences=n_to_find,
            buffer_zone=sequence_length 
        )
        
        if not sequences_neg:
             print("  ⚠ Nessuna sequenza negativa valida trovata (o tutte troppo vicine ai positivi).")
             continue
        
        # -----------------------------------------------------------------
        # 3. BILANCIAMENTO e AGGIUNTA AL DATASET
        # -----------------------------------------------------------------
        
        n_pos = len(sequences_pos)
        n_neg = len(sequences_neg)

        n_to_keep = min(n_pos, n_neg)
        
        if n_pos > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_pos} positivi a {n_to_keep}")
            data_pos = list(zip(sequences_pos, chroms_pos))
            sampled_data_pos = random.sample(data_pos, n_to_keep)
            sequences_pos_sampled, chroms_pos_sampled = zip(*sampled_data_pos)
            sequences_pos = list(sequences_pos_sampled)
            chroms_pos = list(chroms_pos_sampled)
        
        if n_neg > n_to_keep:
            print(f"  Bilanciamento: Sottocampiono {n_neg} negativi a {n_to_keep}")
            data_neg = list(zip(sequences_neg, chroms_neg))
            sampled_data_neg = random.sample(data_neg, n_to_keep)
            sequences_neg_sampled, chroms_neg_sampled = zip(*sampled_data_neg)
            sequences_neg = list(sequences_neg_sampled)
            chroms_neg = list(chroms_neg_sampled)

        # Aggiungi i dati positivi
        all_sequences.extend(sequences_pos)
        all_labels.extend([1] * len(sequences_pos))
        all_chromosomes.extend(chroms_pos) 
        
        # Aggiungi i dati negativi
        all_sequences.extend(sequences_neg)
        all_labels.extend([0] * len(sequences_neg))
        all_chromosomes.extend(chroms_neg) 
        
        print(f"\n  RISULTATO FILE: Aggiunti {len(sequences_pos)} positivi e {len(sequences_neg)} negativi (Bilanciati).")
        print(f"  ✓ Totale sequenze accumulate: {len(all_sequences):,}")

    
    if not all_sequences:
        print("⚠⚠ NESSUNA SEQUENZA TROVATA. Controlla i percorsi e i filtri.")
        return pd.DataFrame(columns=['sequenza', 'nucleosoma', 'chr'])

    # Crea DataFrame
    df = pd.DataFrame({
        'sequenza': all_sequences,
        'nucleosoma': all_labels,
        'chr': all_chromosomes 
    })
    
    # Rimuovi duplicati basati sulla sequenza (mantiene chr e label del primo)
    df = df.drop_duplicates(subset=['sequenza']).reset_index(drop=True)
    # Mescola il dataset finale
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    
    print("\n" + "="*60)
    print("DATASET COMPLETATO (Bilanciato per regione)")
    print("="*60)
    print(f"Totale sequenze: {len(df):,}")
    print(f"Nucleosomi (positivi): {df['nucleosoma'].sum():,}")
    print(f"Non-nucleosomi (negativi): {(1 - df['nucleosoma']).sum():,}")
    
    if len(df) > 0:
        print(f"Bilanciamento: {df['nucleosoma'].sum() / len(df) * 100:.1f}% positivi")
    
    return df

#################################################################################################
# ESECUZIONE PRINCIPALE (Modificata per hg19 / INTERO GENOMA)
#################################################################################################

def main():
    print("="*60)
    print("ELABORAZIONE DATI NUCLEOSOMI GSE36979 (HG19 - INTERO GENOMA)")
    print("="*60)

    # Imposta directory dati
    base_dir = os.getcwd()
    #data_dir = os.path.abspath(os.path.join(base_dir, "..", "data", "lymphoblasotid_data")) #os.path.abspath(os.path.join(base_dir, "..", "data", "cd4t_data"))
    data_dir = os.path.abspath(os.path.join(base_dir, "..", "data", "cd4t_data"))
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        print(f"Creata directory dati: {data_dir}")

    # STEP 1: Trova i file WIG hg19
    print("\n" + "="*60)
    print(f"Ricerca file *_147.hg19.wig in {data_dir}")
    
    #wig_search_path = os.path.join(data_dir, "*.hg19.wig")#
    wig_search_path = os.path.join(data_dir, "ActivatedNucleosomes-chr*.hg19.wig")
    wig_files = glob.glob(wig_search_path)
    
    if not wig_files:
        print(f"✗ Nessun file WIG 147.hg19.wig trovato in {data_dir}.")
        print("  Assicurati che i file .wig convertiti siano in questa cartella.")
        return
    else:
        print(f"✓ Trovati {len(wig_files)} file WIG hg19:")
        for f in wig_files:
            print(f"  - {os.path.basename(f)}")

    # STEP 2: Scarica/Carica genoma hg19 (TUTTI I CROMOSOMI)
    print("\n" + "="*60)
    genome_dir = os.path.join(data_dir, "hg19_genome") 

    check_fa_path = os.path.join(genome_dir, f"{CHROMS_TO_PROCESS[0]}.fa")
    
    if os.path.exists(genome_dir) and os.path.exists(check_fa_path):
        print(f"Genoma hg19 (cartella {genome_dir}) già presente, caricamento...")
        genome_dict = load_hg19_genome(genome_dir) 
    else:
        print("Scaricamento genoma hg19 (tutti i cromosomi standard)...")
        genome_dict = download_hg19_genome(genome_dir) 

    if not genome_dict:
        print("✗ Impossibile caricare genoma hg19")
        return

    # STEP 2.5: Filtro promotori DISABILITATO
    print("\n" + "="*60)
    print("Filtro per promotori DISABILITATO. Si processa l'intero genoma.")
    df_features_to_use = None 
        
    # STEP 3: Crea dataset
    df_final = create_nucleosome_dataset(
        wig_files,  
        genome_dict,
        max_sequences_per_file=1000000, 
        percentile_high=99.8, # <--- QUI ORA C'È SOLO IL PARAMETRO PERCENTILE
        threshold_low=1.0,            
        max_regions_per_wig=5000000000,
        df_features=df_features_to_use, 
        sequence_length=201
    )

    if df_final is None or df_final.empty:
        print("✗ Creazione dataset fallita o dataset vuoto.")
        return

    # STEP 4: Salva risultati
    #output_file = os.path.join(data_dir, "Lymphoblastoid_99_8_percentile.csv") #
    output_file = os.path.join(data_dir, "CD4T_h19_Act_tot_99_8_percentile.csv")
    try:
        df_final.to_csv(output_file, index=False)
        print(f"\n✓ Dataset (hg19, Intero Genoma) salvato: {output_file}")
    except Exception as e:
        print(f"✗ Errore salvataggio dataset: {e}")
        return

    # STEP 5: Statistiche finali
    print("\n" + "="*60)
    print("STATISTICHE FINALI (HG19 - INTERO GENOMA)")
    print("="*60)
    print(df_final.head(10))
    
    print(f"\nDistribuzione lunghezze:")
    print(df_final['sequenza'].str.len().describe())

    # GC content
    df_final['gc_content'] = df_final['sequenza'].apply(calc_gc)
    print(f"\nGC content per classe:")
    print(df_final.groupby('nucleosoma')['gc_content'].describe())

    # Plot
    print("\nGenerazione plot GC content...")
    try:
        plt.figure(figsize=(10, 6))
        sns.kdeplot(data=df_final, x='gc_content', hue='nucleosoma', fill=True, common_norm=False, palette="Set1")
        plt.title('Distribuzione GC Content (0=Bassa, 1=Alta Occupazione) - HG19 / Intero Genoma') 
        plot_file = os.path.join(data_dir, "gc_content_distribution_hg19_whole_genome.png")
        plt.savefig(plot_file)
        print(f"✓ Plot salvato: {plot_file}")
    except Exception as e:
        print(f"✗ Errore generazione plot: {e}")

    print("\n✓ ELABORAZIONE (HG19 - INTERO GENOMA) COMPLETATA!") 

if __name__ == "__main__":  
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import gzip
from io import StringIO
import warnings

#################################################################################################
# FUNZIONE PARSER (dalla pipeline precedente)
#################################################################################################

def parse_wig_file(wig_file_path, max_regions=100000):
    """
    Parsifica file WIG con dati di occupancy dei nucleosomi.
    Imposta max_regions=np.inf per leggere l'intero file.
    """
    print(f"  Parsing file WIG: {wig_file_path}")
    
    regions = []
    current_chrom = None
    current_start = None
    current_step = None
    current_span = 1
    position = 0
    lines_processed = 0
    
    try:
        # Gestisci file compressi o no
        opener = gzip.open if wig_file_path.endswith('.gz') else open
        
        with opener(wig_file_path, 'rt') as f:
            for line in f:
                line = line.strip()
                
                if not line or line.startswith('track') or line.startswith('browser'):
                    continue
                
                lines_processed += 1
                
                # Riduzione print (mostra ogni 50 milioni di righe)
                if lines_processed % 50000000 == 0:
                    print(f"    ... Processate {lines_processed:,} linee, trovate {len(regions):,} regioni...")
                
                # Stop se abbiamo raggiunto il limite (np.inf non verrà mai raggiunto)
                if len(regions) >= max_regions:
                    print(f"    ⚠ Raggiunto limite di {max_regions:,} regioni, stop parsing")
                    break
                
                # fixedStep format
                if line.startswith('fixedStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('start='):
                            current_start = int(part.split('=')[1])
                            position = current_start
                        elif part.startswith('step='):
                            current_step = int(part.split('=')[1])
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                
                # variableStep format (il tuo file è .vstep)
                elif line.startswith('variableStep'):
                    parts = line.split()
                    for part in parts:
                        if part.startswith('chrom='):
                            current_chrom = part.split('=')[1]
                        elif part.startswith('span='):
                            current_span = int(part.split('=')[1])
                    current_step = None # Flag per indicare variableStep
                
                # Data line
                else:
                    try:
                        if current_step is not None:
                            # fixedStep: solo valore
                            value = float(line)
                            if value > 0: # Solo regioni con segnale
                                regions.append({
                                    'chr': current_chrom,
                                    'start': position,
                                    'end': position + current_span,
                                    'score': value
                                })
                            position += current_step
                        else:
                            # variableStep: posizione e valore
                            pos, value = line.split()
                            pos = int(pos)
                            value = float(value)
                            if value > 0:
                                regions.append({
                                    'chr': current_chrom,
                                    'start': pos,
                                    'end': pos + current_span,
                                    'score': value
                                })
                    except Exception:
                        # Salta righe dati malformate
                        continue
            
        print(f"    ✓ Trovate {len(regions):,} regioni con segnale (Parsing completato)")
        return pd.DataFrame(regions)
    
    except FileNotFoundError:
        print(f"    ✗ ERRORE: File non trovato a {wig_file_path}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])
    except Exception as e:
        print(f"    ✗ Errore parsing WIG: {e}")
        return pd.DataFrame(columns=['chr', 'start', 'end', 'score'])

#################################################################################################
# ESECUZIONE ANALISI STATISTICA
#################################################################################################

def analyze_scores():
    print("="*60)
    print("ANALISI STATISTICA DELLO SCORE")
    print("="*60)

    # 1. Definisci il percorso del file (come nello script precedente)
    base_dir = os.getcwd()
    data_dir = os.path.abspath(os.path.join(base_dir, "..", "data/cd4t_data"))
    user_file_name = "RestingNucleosomes-chr3-scores.vstep.gz"#"RestingNucleosomes-chr1-scores.vstep.gz"
    user_file_path = os.path.join(data_dir, user_file_name)

    if not os.path.exists(user_file_path):
        print(f"✗ File non trovato: {user_file_path}")
        print(f"Assicurati di aver posizionato '{user_file_name}' nella directory: {data_dir}")
        return

    # 2. Carica i dati - LEGGI TUTTO IL FILE usando np.inf
    print(f"Caricamento di {user_file_name} in corso...")
    print("Questo processo leggerà l'INTERO file e potrebbe richiedere molto tempo e memoria.")
    df_wig = parse_wig_file(user_file_path, max_regions=np.inf) # np.inf = infinito
    
    if df_wig is None or df_wig.empty:
        print("✗ Dati non caricati o file vuoto.")
        return

    print(f"✓ Caricamento completato. {len(df_wig):,} regioni totali trovate.")
    
    # 3. Analisi Statistica Descrittiva
    print("\n" + "--- Statistiche Descrittive (Tutti gli score > 0) ---")
    # Usiamo .to_string() per una formattazione pulita
    print(df_wig['score'].describe(percentiles=[0.1, 0.3, 0.5, 0.7, 0.9, 0.99]).to_string())
    
    # 4. Analisi specifica per le soglie della pipeline
    print("\n" + "--- Conteggi per Soglie Chiave (per la pipeline) ---")
    
    total_count = len(df_wig)
    # Soglia per i negativi (score == 1.0)
    count_eq_1 = (df_wig['score'] == 1.0).sum()
    # Soglia per i positivi (usata a 10.0 nello script main)
    count_gt_10 = (df_wig['score'] > 50.0).sum()
    # Conteggio di tutti i possibili candidati positivi (> 1.0)
    count_gt_1 = (df_wig['score'] > 1.0).sum()
    
    if total_count > 0:
        print(f"Totale regioni (score > 0): {total_count:,}")
        print(f"Regioni con score == 1.0 (Negativi): {count_eq_1:,} ({count_eq_1/total_count*100:.2f}%)")
        print(f"Regioni con score > 1.0 (Candidati Pos): {count_gt_1:,} ({count_gt_1/total_count*100:.2f}%)")
        print(f"Regioni con score > 30.0 (Positivi > 30): {count_gt_10:,} ({count_gt_10/total_count*100:.2f}%)")
    else:
        print("Nessuna regione trovata.")

    # 5. Visualizzazione
    print("\nGenerazione plot in corso...")
    sns.set_theme(style="whitegrid")
    warnings.filterwarnings('ignore', category=UserWarning) # Ignora warning di seaborn su log scale

    try:
        # Grafico 1: Distribuzione completa (con scala log)
        # La scala log è essenziale perché la maggior parte dei valori è 1.0
        plt.figure(figsize=(14, 7))
        sns.histplot(df_wig['score'], bins=200, log_scale=True, kde=False)
        plt.title(f'Distribuzione Completa Score (Scala Log) - {user_file_name}', fontsize=16)
        plt.xlabel('Score (Scala Log)', fontsize=12)
        plt.ylabel('Conteggio (Frequenza)', fontsize=12)
        plt.show()

        # Grafico 2: Boxplot per vedere gli outliers (sempre in scala log)
        plt.figure(figsize=(14, 4))
        sns.boxplot(x=df_wig['score'])
        plt.xscale('log') # Fondamentale scala log
        plt.title('Boxplot Score (Scala Log) - Mostra outliers', fontsize=16)
        plt.xlabel('Score (Scala Log)', fontsize=12)
        plt.show()
        
        # Grafico 3: Zoom sui valori > 1.0 (i candidati "positivi")
        df_gt1 = df_wig[df_wig['score'] > 1.0]
        if not df_gt1.empty:
            plt.figure(figsize=(14, 7))
            sns.histplot(df_gt1['score'], bins=100, log_scale=True, kde=True)
            plt.title(f'Distribuzione Score > 1.0 (Candidati Positivi) - {len(df_gt1):,} regioni', fontsize=16)
            plt.xlabel('Score (Scala Log)', fontsize=12)
            plt.ylabel('Conteggio (Frequenza)', fontsize=12)
            plt.show()
        else:
            print("Nessuno score > 1.0 trovato per il plot di zoom.")

    except Exception as e:
        print(f"✗ Errore durante la generazione dei plot: {e}")
    
    warnings.filterwarnings('default') # Riabilita i warning
    print("\n✓ Analisi statistica completata.")

# Esegui l'analisi
analyze_scores()
```
